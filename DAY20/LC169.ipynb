# LeetCode 169 - Majority Element

## Problem

Given an array `nums` of size `n`, return the **majority element**.

A majority element is an element that appears **more than ⌊n / 2⌋ times**.

It is guaranteed that the majority element always exists.

---

## Pattern

**Boyer-Moore Voting Algorithm**

---

## Intuition

Think of every element as a vote.

- The majority element gains one vote.
- Every different element cancels one vote.

Since the majority element appears more than half the time,

it can never be completely cancelled.

Eventually,

it is the only candidate left.

---

## Key Observation

Whenever we see

- the same candidate → increase votes
- a different number → decrease votes

If the vote count becomes zero,

it means the current candidate has been completely cancelled.

Choose the current number as the new candidate.

---

## Algorithm

1. Initialize:
   - `candidate = None`
   - `count = 0`
2. Traverse the array.
3. If `count == 0`:
   - Set current number as the new candidate.
4. If current number equals candidate:
   - Increase `count`.
5. Otherwise:
   - Decrease `count`.
6. Return the candidate.

## Why Check `count == 0` First?

Suppose

```text
count = 0
current = 5
```

There is currently no active candidate.

So

```text
candidate = 5
```

Now process the same element.

```text
5 == candidate
```

Increase

```text
count = 1
```

Each element should contribute exactly one vote.

---

## Common Mistake

Wrong order

```python
if num == candidate:
    count += 1
else:
    count -= 1

if count == 0:
    candidate = num
```

The current element is processed before selecting the new candidate.

This causes incorrect counting.

---

## Correct Order

```python
if count == 0:
    candidate = num

if num == candidate:
    count += 1
else:
    count -= 1
```

---

## Why Doesn't This Lose the Majority Element?

Whenever a different element appears,

it cancels exactly one occurrence of the majority element.

Since the majority element appears more than

```text
n / 2
```

times,

it can never be completely cancelled.

Eventually,

it becomes the final candidate.

## Common Mistakes

### Mistake 1

Initializing

```python
count = 1
```

Always start with

```python
count = 0
```

---

### Mistake 2

Changing the candidate after processing the element.

Always choose the new candidate **before** counting the current vote.

---

### Mistake 3

Thinking the candidate is always the answer.

This works here because the problem guarantees that a majority element exists.

If that guarantee is removed,

a second verification pass is required.

````python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count=0
        value=nums[0]
        for i in range(len(nums)):
            if nums[i]==value:
                count+=1
            elif nums[i]!=value:
                count-=1
            if count==0:
                value=nums[i]
                count=1
        return value
````
