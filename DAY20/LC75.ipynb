# LeetCode 75 - Sort Colors

## Problem

Given an array `nums` containing only:

- `0`
- `1`
- `2`

Sort the array **in-place** without using the built-in sorting function.

---

## Pattern

**Dutch National Flag Algorithm (Three Pointers)**

---

## Intuition

Since the array contains only **three distinct values**, we do not need a general sorting algorithm.

Instead, divide the array into four regions:

```text
0s | 1s | Unknown | 2s
```

The algorithm gradually expands the sorted regions while shrinking the unknown region.

---

## Pointer Responsibilities

### `low`

Everything before `low` is guaranteed to be `0`.

```text
0 0 0 | ? ? ? ? ? ? ?
^
low
```

---

### `mid`

Current element being inspected.

Everything between `mid` and `high` is still unknown.

```text
0 0 | ? ? ? ? | 2 2
      ^
     mid
```

---

### `high`

Everything after `high` is guaranteed to be `2`.

```text
? ? ? ? ? | 2 2 2
          ^
        high
```

---

## Pointer Invariant

At every step:

```text
0 ........ low-1
```

All are `0`s.

---

```text
low ........ mid-1
```

All are `1`s.

---

```text
mid ........ high
```

Unknown region.

These elements still need processing.

---

```text
high+1 ........ end
```

All are `2`s.

---

## Algorithm

Initialize:

```text
low = 0
mid = 0
high = n-1
```

Process while:

```text
mid <= high
```

---

### Case 1

Current element is `0`.

It belongs on the left.

- Swap `low` and `mid`
- Move both pointers

```text
low++
mid++
```

---

### Case 2

Current element is `1`.

It is already in the correct region.

Move only

```text
mid++
```

---

### Case 3

Current element is `2`.

It belongs on the right.

- Swap `mid` and `high`
- Move only

```text
high--
```

Do **not** move `mid`.

The element that came from the right has not been processed yet.

---

## Why Don't We Move `mid` After Swapping with `high`?

Suppose

```text
2 1 0
^   ^
M   H
```

Swap:

```text
0 1 2
^
M
```

The new element at `mid` is `0`.

It still needs processing.

If we immediately do

```text
mid++
```

we skip this element completely.

Therefore,

after swapping with `high`:

```text
Only high moves.
```

---

## Dry Run

### Input

```text
[2,0,2,1,1,0]
```

Initially

```text
low = 0
mid = 0
high = 5
```

---

### Step 1

```text
2 0 2 1 1 0
^         ^
M         H
```

Current element is `2`.

Swap with `high`.

```text
0 0 2 1 1 2
^       ^
M       H
```

Move

```text
high--
```

---

### Step 2

Current element is now

```text
0
```

Swap with `low`.

Move

```text
low++
mid++
```

---

### Continue

Repeat until

```text
mid > high
```

Final array

```text
0 0 1 1 2 2
```

---

## Why Does This Work?

The algorithm never loses track of the sorted regions.

Each pointer has exactly one responsibility:

- `low` builds the `0` region.
- `mid` explores unknown elements.
- `high` builds the `2` region.

Every iteration reduces the unknown region.

Eventually,

no unknown elements remain.

---

## Time Complexity

### O(n)

Each element is processed at most once.

---

## Space Complexity

### O(1)

The sorting is performed completely in-place.

---

## Common Mistakes

### Mistake 1

Using a `for` loop.

Wrong

```python
for i in range(len(nums)):
```

A `for` loop always increments automatically.

After swapping with `high`, the new element at `mid` still needs processing.

Therefore,

use

```python
while mid <= high:
```

---

### Mistake 2

Incrementing `mid` after swapping with `high`.

Wrong

```text
swap(mid, high)
high--
mid++
```

The new element at `mid` has not been inspected.

Correct

```text
swap(mid, high)
high--
```

---

### Mistake 3

Using three independent `if` statements.

Wrong

```python
if ...
if ...
if ...
```

After the first case executes,

the value at `nums[mid]` may change.

Use

```python
if ...
elif ...
else ...
```

Only one case should execute in each iteration.

---

### Mistake 4

Treating this like Bubble Sort or Selection Sort.

This is **not** a comparison-based sorting algorithm.

It is a partitioning algorithm.

---

## Interview Takeaways

Whenever a problem contains only a small number of distinct values,

ask yourself:

> Can I partition the array instead of sorting it?

If the answer is yes,

the Dutch National Flag Algorithm may be applicable.

````python
class Solution(object):
    def sortColors(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        left=0
        right=len(nums)-1
        mid=0

        while mid<=right:
            if nums[mid]==0:
                nums[left],nums[mid]=nums[mid],nums[left]
                left+=1
                mid+=1
            elif nums[mid]==1:
                mid+=1
            else:
                nums[mid],nums[right]=nums[right],nums[mid]
                right-=1
                
        return nums
````
