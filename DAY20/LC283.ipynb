# LeetCode 238 - Product of Array Except Self

## Problem

Given an integer array `nums`, return an array `answer` such that:

```text
answer[i] = Product of all elements except nums[i]
```

### Constraints

- Do **not** use division.
- Must run in **O(n)** time.

---

## Pattern

**Prefix Product + Suffix Product**

---

## Intuition

Instead of multiplying every element except the current one repeatedly,

split the product into two parts.

```text
Product Except Self

=

(Product of Left Side)

×

(Product of Right Side)
```

For every index,

the answer is simply

```text
Prefix Product × Suffix Product
```

---

## Brute Force

For every index:

- Traverse the entire array.
- Multiply every element except itself.

```text
for every i
    for every j
```

### Time Complexity

```text
O(n²)
```

Too slow.

---

## Observation

Example

```text
nums = [1,2,3,4]
```

At index `2`

```text
1 × 2 × 4
```

can be written as

```text
(1 × 2)

×

(4)
```

Instead of recomputing these products,

store them once.

---

## Prefix Array

`prefix[i]`

stores

```text
Product of all elements BEFORE index i
```

Example

```text
nums = [1,2,3,4]
```

| Index | Prefix Product |
|------:|---------------:|
|0|1|
|1|1|
|2|2|
|3|6|

Result

```text
prefix = [1,1,2,6]
```

---

### Formula

```python
prefix[i] = prefix[i-1] * nums[i-1]
```

Build from

```text
Left → Right
```

---

## Suffix Array

`suffix[i]`

stores

```text
Product of all elements AFTER index i
```

Example

```text
nums = [1,2,3,4]
```

| Index | Suffix Product |
|------:|---------------:|
|0|24|
|1|12|
|2|4|
|3|1|

Result

```text
suffix = [24,12,4,1]
```

---

### Formula

```python
suffix[i] = suffix[i+1] * nums[i+1]
```

Build from

```text
Right → Left
```

---

## Final Answer

For every index

```python
answer[i] = prefix[i] * suffix[i]
```

Example

```text
Prefix

[1,1,2,6]

Suffix

[24,12,4,1]
```

Multiply

```text
1×24 = 24

1×12 = 12

2×4 = 8

6×1 = 6
```

Final

```text
[24,12,8,6]
```

---

## Algorithm

### Step 1

Create

```python
prefix = [1] * n
```

---

### Step 2

Build Prefix

```python
for i in range(1, n):
    prefix[i] = prefix[i-1] * nums[i-1]
```

---

### Step 3

Create

```python
suffix = [1] * n
```

---

### Step 4

Build Suffix

```python
for i in range(n-2, -1, -1):
    suffix[i] = suffix[i+1] * nums[i+1]
```

---

### Step 5

Build Answer

```python
for i in range(n):
    answer[i] = prefix[i] * suffix[i]
```

## Why Does This Work?

At every index,

everything to the left has already been multiplied.

Everything to the right has already been multiplied.

So

```text
Left Product

×

Right Product
```

gives

```text
Product Except Self
```

without recomputing anything.

---

## Complexity

### Time

Building Prefix

```text
O(n)
```

Building Suffix

```text
O(n)
```

Building Answer

```text
O(n)
```

Overall

```text
O(3n)

=

O(n)
```

---

### Space

```text
O(n)
```

for

- Prefix
- Suffix
- Answer

---

## Why Is This Better Than Brute Force?

Brute Force repeatedly calculates the same products.

Example

```text
1 × 2 × 3
```

is recomputed multiple times.

Prefix & Suffix compute it **once** and reuse it.

This removes redundant work.

---

## Follow-Up (Optimal Interview Solution)

The interview follow-up asks for

```text
O(1)
```

extra space.

Idea:

- Store Prefix Products directly inside the answer array.
- Use one variable

```text
suffix_product
```

while traversing from right to left.

No separate suffix array is needed.

## Common Mistakes

### Mistake 1

Thinking Prefix stores elements.

Wrong

```text
[1,2,3]
```

Correct

```text
[1,1,2]
```

Prefix stores **products**, not values.

---

### Mistake 2

Thinking Suffix stores elements.

Wrong

```text
[4,3,2]
```

Correct

```text
[24,12,4,1]
```

Suffix stores **products**.

---

### Mistake 3

Trying

```python
prefix * suffix
```

Lists cannot be multiplied element-wise.

Need

```python
answer[i] = prefix[i] * suffix[i]
```

---

### Mistake 4

Building Prefix from right to left.

Prefix always depends on

```python
prefix[i-1]
```

so it must be built

```text
Left → Right
```

---

### Mistake 5

Building Prefix, Suffix and Answer in one loop.

These are three independent stages.

Build them separately.

---

## Interview Takeaways

Whenever a problem asks

- Left Product
- Right Product
- No Division

Think immediately:

> **Prefix Product + Suffix Product**

## Pattern Recognition

| Keyword | Pattern |
|----------|---------|
| Product Except Self | Prefix + Suffix |

| Right Product | Suffix |
| No Division | Prefix & Suffix |

- Understood that Prefix stores the product of all elements before the current index.
- Understood that Suffix stores the product of all elements after the current index.
- Learned how to combine Prefix and Suffix to get the final answer.
- Understood why this is **O(n)** instead of **O(n²)**.
- Learned there is an interview optimization to reduce space from **O(n)** to **O(1)** extra space.
- Added a new DSA pattern:
  **Prefix & Suffix Products**.

````python
class Solution(object):
    def productExceptSelf(self, nums):
        n = len(nums)

        prefix = [1] * n
        suffix = [1] * n
        answer = [1] * n

        for i in range(1, n):
            prefix[i] = prefix[i - 1] * nums[i - 1]

        for i in range(n - 2, -1, -1):
            suffix[i] = suffix[i + 1] * nums[i + 1]

        for i in range(n):
            answer[i] = prefix[i] * suffix[i]

        return answer
````
