# LeetCode 53 - Maximum Subarray

## Problem

Given an integer array `nums`, find the **contiguous subarray** that has the largest sum and return its sum.

---

## Pattern

**Kadane's Algorithm**

---

## Intuition

While traversing the array:

- Keep track of the **current subarray sum**.
- Keep track of the **maximum subarray sum** found so far.

If the current sum becomes **negative**, discard it.

Why?

A negative sum will only decrease the sum of any future subarray.

---

## Key Observation

If

```text
Current Sum < 0
```

Then carrying it forward is harmful.

Instead,

start a new subarray from the next element.

---

## Running State Pattern

This problem follows the same pattern as **LC 485 (Max Consecutive Ones)**.

| LC 485 | LC 53 |
|---------|--------|
| Current Streak | Current Sum |
| Maximum Streak | Maximum Sum |

General Pattern:

```text
Current State
↓

Best State
```

---

## Algorithm

1. Initialize:
   - `current_sum = 0`
   - `max_sum = nums[0]`
2. Traverse the array.
3. Add the current element to `current_sum`.
4. Update `max_sum`.
5. If `current_sum` becomes negative:
   - Reset it to `0`.
6. Return `max_sum`.

---

## Why Do We Reset?

Suppose

```text
Current Sum = -8
```

Next number

```text
6
```

Option 1

```text
-8 + 6 = -2
```

Option 2

```text
6
```

Starting fresh always gives the better result.

Therefore,

whenever

```text
Current Sum < 0
```

discard it.

---

## Why Not Reset at Zero?

Suppose

```text
Current Sum = 0
```

Next number

```text
8
```

Continuing

```text
0 + 8 = 8
```

Starting fresh

```text
8
```

Both are identical.

Zero does not hurt future sums.

Only negative sums do.

---

## Why Initialize `max_sum = nums[0]`?

Consider

```text
[-5,-2,-8]
```

Correct answer

```text
-2
```

If

```python
max_sum = 0
```

the algorithm incorrectly returns

```text
0
```

which is not even present in the array.

Initializing

```python
max_sum = nums[0]
```

handles arrays containing only negative numbers.

---

## Time Complexity

### O(n)

Each element is processed exactly once.

---

## Space Complexity

### O(1)

Only two variables are maintained.

---

## Common Mistakes

### Mistake 1

Initializing

```python
max_sum = 0
```

Fails when all elements are negative.

---

### Mistake 2

Resetting when

```text
current_sum <= 0
```

Reset only when

```text
current_sum < 0
```

Zero does not reduce future sums.

---

### Mistake 3

Checking every possible subarray.

Brute Force

```text
O(n²)
```

Kadane's Algorithm

```text
O(n)
```

## Interview Takeaways

Whenever a problem asks for:

- Maximum Sum
- Contiguous Subarray
- Largest Running Sum

Think:

> **Kadane's Algorithm**

Ask yourself:

> "Is my current running sum helping future sums or hurting them?"

If it becomes negative,

discard it.

````python
class Solution(object):
    def maxSubArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        cs=0
        max_sum=nums[0]
        for i in nums:
            cs+=i
            max_sum=max(max_sum,cs)
            if cs<0:
                cs=0
        return max_sum
````
