# RECURSIVE BUBBLE SORT

```python
def bubble(arr,n): # for ascending 
    if n==1:
        return 
    for i in range(n-1):
        if arr[i]>arr[i+1]:
            arr[i],arr[i+1]=arr[i+1],arr[i]
    bubble(arr,n-1)

arr=list(map(int,input().split()))
n=len(arr)
bubble(arr,n)
print(arr)
```

```python
def bubble(arr,n): # for descending
    if n==1:
        return 
    for i in range(n-1):
        if arr[i]<arr[i+1]:
            arr[i],arr[i+1]=arr[i+1],arr[i]
    bubble(arr,n-1)
arr=list(map(int,input().split()))
n=len(arr)
bubble(arr,n)
print(arr)
```
