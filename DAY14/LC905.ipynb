# LC 905

# LeetCode 905 - Sort Array By Parity

## Problem

Given an integer array `nums`, move all the even integers to the beginning of the array followed by all the odd integers.

The relative order of even or odd numbers does **not** matter.

---

## Intuition

- We only need to partition the array into two groups:
  - Even numbers on the left.
  - Odd numbers on the right.
- Since the order is not important, we can swap misplaced elements instead of creating another array.

---

## Pattern

**Two Pointers + In-place Swapping**

---

## Algorithm

1. Initialize:
   - `left = 0`
   - `right = len(nums) - 1`
2. Continue while `left < right`.
3. If the left element is odd and the right element is even, swap them.
4. If the left element is even, move `left` forward.
5. If the right element is odd, move `right` backward.
6. Return the modified array.

---

## Why does this work?

The left pointer searches for an odd number that should move to the right.

The right pointer searches for an even number that should move to the left.

When both pointers find misplaced elements, swapping fixes two positions with one operation.

---

## Dry Run

Input:

```text
[3,1,2,4]
```

| Left | Right | Action | Array |
|------|-------|--------|-------|
|3|4|Swap|[4,1,2,3]|
|1|2|Swap|[4,2,1,3]|

Output:

```text
[4,2,1,3]
```

---

## Time Complexity

**O(n)**

### Reason

- Each pointer moves toward the center.
- Every element is visited at most once.

---

## Space Complexity

**O(1)**

### Reason

The rearrangement is done in-place without using any extra array.

---

## Edge Cases

- Single element
- All even numbers
- All odd numbers
- Empty array (if allowed)
- Duplicate values

---

## Common Mistakes

### Mistake 1

Created an extra array.

**Learning:**

Since order doesn't matter, swapping is enough.

---

### Mistake 2

Moved the wrong pointer.

The right pointer should only move when the right element is already an **odd** number.

---

### Mistake 3

Got a Time Limit Exceeded (Infinite Loop)

Reason:

Neither `left` nor `right` moved in some cases.

Always make sure at least one pointer moves during each iteration.

---

## Interview Takeaway

When a problem says:

- Move one category to the left.
- Move another category to the right.
- Order does not matter.

Think:

**Can I solve this using two pointers and in-place swapping instead of creating another array?**

## Pattern Recognition

| Problem | Pattern |
|---------|---------|
| LC 88 | Two pointers from the end |
| LC 977 | Two pointers + result array |
| LC 905 | Two pointers + in-place partition |

````python
class Solution(object):
    def sortArrayByParity(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        left=0
        right=len(nums)-1
        while left<right:
            if nums[left]%2>nums[right]%2:
                nums[left],nums[right]=nums[right],nums[left]
            if nums[left]%2==0:
                left+=1
            if nums[right]%2==1:
                right-=1
        return nums
````
