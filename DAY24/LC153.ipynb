# LeetCode 153 - Find Minimum in Rotated Sorted Array

## Problem

Given a rotated sorted array, return the **minimum element**.

The array contains **unique elements**.

Example

```text
Input

[3,4,5,1,2]
```

Output

```text
1
```

Another Example

```text
Input

[4,5,6,7,0,1,2]
```

Output

```text
0
```

---

# Pattern

**Modified Binary Search**

---

# Intuition

Originally the array was sorted.

```text
0 1 2 4 5 6 7
```

After rotation

```text
4 5 6 7 0 1 2
```

We don't know where the rotation happened.

Instead of searching for a target,

we search for the **minimum element**.

---

# Key Observation

Unlike LC 33,

we **do not** ask

```text
Which half is sorted?
```

Instead we compare

```text
nums[mid]

with

nums[high]
```

This comparison tells us where the minimum lies.

---

# Case 1

Suppose

```text
4 5 6 7 | 0 1 2
```

```text
mid = 7

high = 2
```

Compare

```text
7 > 2
```

Meaning

the minimum **cannot** be on the left.

It must be on the right.

Move

```python
low = mid + 1
```

---

# Case 2

Suppose

```text
0 1 2
```

```text
mid = 1

high = 2
```

Compare

```text
1 < 2
```

The right half is already sorted.

The minimum could still be

```text
0
```

or even

```text
1
```

Therefore

keep `mid`.

Move

```python
high = mid
```

Notice

NOT

```python
high = mid - 1
```

because `mid` itself could be the minimum.

---

# Binary Search Skeleton

```python
low = 0
high = len(nums) - 1

while low < high:

    mid = (low + high) // 2
```

---

# Decision

## Case 1

```python
if nums[mid] > nums[high]:
```

Minimum is on the right.

```python
low = mid + 1
```

---

## Case 2

```python
else:
```

Minimum is on the left (including mid).

```python
high = mid
```

---

# Complete Algorithm

```python
low = 0
high = len(nums) - 1

while low < high:

    mid = (low + high) // 2

    if nums[mid] > nums[high]:
        low = mid + 1
    else:
        high = mid

return nums[low]
```

# Why `high = mid`?

Suppose

```text
0 1 2
```

If

```python
high = mid - 1
```

you might discard

```text
0
```

which is actually the minimum.

Therefore

always keep `mid`.

```python
high = mid
```

---

# Why `while low < high`?

When

```text
low == high
```

only one candidate remains.

That candidate must be the minimum.

Using

```python
while low <= high
```

can lead to an infinite loop because

```python
high = mid
```

does not always move the pointer.

---

# Complexity

## Time

Every iteration removes half the search space.

```text
O(log n)
```

---

## Space

Only pointers are used.

```text
O(1)
```

---

# Common Mistakes

## Mistake 1

Using

```python
while low <= high
```

Correct

```python
while low < high
```

---

## Mistake 2

Using

```python
high = mid - 1
```

Correct

```python
high = mid
```

Keep `mid` because it could be the minimum.

---

## Mistake 3

Comparing

```python
nums[low]

with

nums[mid]
```

That's the logic for **LC 33**.

For **LC 153**, compare

```python
nums[mid]

with

nums[high]
```

---

## Mistake 4

Returning

```python
nums[mid]
```

The answer is obtained only after the loop finishes.

Return

```python
nums[low]
```

# Personal Learning Notes

- Learned another **Modified Binary Search** pattern.
- Understood that this problem does **not** search for a target.
- Learned to compare `nums[mid]` with `nums[high]`.
- If `nums[mid] > nums[high]`, the minimum lies on the right.
- Otherwise, the minimum lies on the left, including `mid`.
- Learned why `high = mid` is necessary instead of `mid - 1`.
- Learned why the loop condition is `while low < high`.
- Added a new DSA pattern:

**Modified Binary Search (Find Minimum in Rotated Sorted Array)**.

````python
class Solution(object):
    def findMin(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        low=0
        high=len(nums)-1
        
        while low<high:
            mid=(low+high)//2
            if nums[mid]>nums[high]:
                low=mid+1
            else:
                high=mid
        return nums[low]

````
