# LeetCode 33 - Search in Rotated Sorted Array

## Problem

Given a rotated sorted array and a target value, return the index of the target.

If the target does not exist, return

```text
-1
```

Example

```text
Input

nums = [4,5,6,7,0,1,2]

target = 0
```

Output

```text
4
```

---

# Pattern

**Modified Binary Search**

---

# Intuition

The array was originally sorted.

Example

```text
0 1 2 4 5 6 7
```

After rotation

```text
4 5 6 7 0 1 2
```

The entire array is no longer sorted.

However,

**at least one half is always sorted.**

Our job is to identify the sorted half and determine whether the target can exist inside it.

---

# Key Observation

Every iteration,

one of these is always sorted.

```text
Left Half

or

Right Half
```

Example

```text
4 5 6 7 | 0 1 2
```

Left half

```text
4 5 6 7
```

is sorted.

---

Another example

```text
6 7 0 1 | 2 4 5
```

Right half

```text
2 4 5
```

is sorted.

# Complete Algorithm

```python
low = 0
high = len(nums) - 1

while low <= high:

    mid = (low + high) // 2

    if nums[mid] == target:
        return mid

    if nums[low] <= nums[mid]:

        if nums[low] <= target < nums[mid]:
            high = mid - 1
        else:
            low = mid + 1

    else:

        if nums[mid] < target <= nums[high]:
            low = mid + 1
        else:
            high = mid - 1

return -1
```

# Complexity

## Time

Every iteration removes half the array.

```text
O(log n)
```

---

## Space

Only pointers are used.

```text
O(1)
```

# Common Mistakes

## Mistake 1

Checking

```python
if mid == target
```

Wrong.

Correct

```python
if nums[mid] == target
```

---

## Mistake 2

Using

```python
nums[low] < nums[mid]
```

Correct

```python
nums[low] <= nums[mid]
```

Single-element halves must also be considered sorted.

---

## Mistake 3

Wrong right-half condition.

Wrong

```python
nums[mid] <= target < nums[high]
```

Correct

```python
nums[mid] < target <= nums[high]
```

Reason

`nums[mid] == target` has already been checked.

---

## Mistake 4

Checking where the target is before checking which half is sorted.

Always do:

```text
1. Which half is sorted?

↓

2. Can target exist there?
```

---

## Mistake 5

Treating the array as completely sorted.

Remember

```text
Entire Array

❌ Not Sorted

One Half

✅ Always Sorted
```

**Modified Binary Search (Rotated Sorted Array)**.

````python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        low=0
        high=len(nums)-1
        while low<=high:
            mid=(low+high)//2
            if nums[mid]==target:
                return mid
            if nums[low]<=nums[mid]:
                if nums[low]<=target<=nums[mid]:
                    high=mid-1
                else:
                    low=mid+1
            else:
                if nums[mid]<=target<=nums[high]:
                    low=mid+1
                else:
                    high=mid-1
        return -1
````
