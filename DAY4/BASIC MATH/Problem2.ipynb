# =========================================
# 1. Problem Understanding
# =========================================

Given a signed 32-bit integer x,
return x with its digits reversed.

If reversing x causes the value to go
outside the signed 32-bit integer range,
return 0.

Examples:
123 -> 321
-123 -> -321
120 -> 21

# =========================================
# 2. Key Observation
# =========================================


Extract the last digit using %10
and build reversed number step-by-step.

Core formula:

result = result * 10 + digit

This shifts existing digits left
and appends the new digit.

Need overflow check BEFORE multiplying by 10.


# =========================================
# 3. Brute Force Approach
# =========================================


Convert integer into string,
reverse it,
then convert back to integer.

Need extra handling for:
- negative sign
- overflow


# Time Complexity: O(k)
# Space Complexity: O(k)

# =========================================
# 4. Optimal Approach
# =========================================


Use mathematical digit extraction.

Steps:
- Extract last digit using %10
- Remove digit using //10
- Rebuild reversed number
- Handle negative numbers separately
- Prevent overflow before multiplication

This avoids extra space used by strings.

# =========================================
# 5. Algorithm Steps
# =========================================

1. Store sign separately
2. Convert number to positive using abs()
3. Extract last digit using %10
4. Append digit into result
5. Remove last digit using //10
6. Check overflow before multiplication
7. Restore sign at end

# 7. Dry Run
# =========================================


Input:
x = 123

Initial:
result = 0

Iteration 1:
digit = 3
result = 0*10 + 3 = 3

Iteration 2:
digit = 2
result = 3*10 + 2 = 32

Iteration 3:
digit = 1
result = 32*10 + 1 = 321

Output:
321

# =========================================
# 8. Final Code
# =========================================

class Solution(object):

    def reverse(self, x):

        result = 0
        sign = 1

        if x < 0:
            sign = -1

        value = abs(x)

        while value > 0:

            digit = value % 10

            # Overflow check
            if result > (2**31 - 1) // 10:
                return 0

            result = result * 10 + digit

            value //= 10

        result *= sign

        # Final range check
        if result < -2**31 or result > 2**31 - 1:
            return 0

        return result
