# =========================================
# 1. Problem Understanding
# =========================================


Given an integer N,
check whether it is an Armstrong number.

An Armstrong number is a number where:

sum of each digit raised to the power
of total number of digits
equals original number.

Examples:
153 -> True
371 -> True
123 -> False

# =========================================
# 2. Key Observation
# =========================================


For every digit:
- extract digit using %10
- raise digit to power of total digits
- add into running sum

If final sum equals original number,
then number is Armstrong.

# =========================================
# 3. Brute Force Approach
# =========================================

Convert number into string
to calculate digit count.

Then:
- extract digits one-by-one
- compute digit^count
- add into result
- compare result with original number


# Time Complexity: O(log10(n))
# Space Complexity: O(1)

# =========================================
# 4. Optimal Approach
# =========================================

Use mathematical digit extraction.

Steps:
- count total digits
- extract digits using %10
- remove digits using //10
- add digit^count into result
- compare final result with original number

# =========================================
# 5. Algorithm Steps
# =========================================

1. Take input number
2. Count total digits
3. Preserve original number
4. Extract digits one-by-one
5. Raise digit to power of total digits
6. Add into running sum
7. Compare final sum with original number

# =========================================
# 6. Mistakes I Made

# =========================================
# 7. Dry Run
# =========================================


Input:
n = 153

digit = 3

Iteration 1:
digit = 3
3^3 = 27
result = 27

Iteration 2:
digit = 5
5^3 = 125
result = 152

Iteration 3:
digit = 1
1^3 = 1
result = 153

Compare:
153 == 153

Output:
True


# =========================================
# 8. Final Code
# =========================================

n = int(input())

digit = len(str(n))

result = 0

num = n

while num > 0:

    result = result + (num % 10) ** digit

    num //= 10

if result == n:
    print("True")

else:
    print("False")
