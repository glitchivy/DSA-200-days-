# =========================================
# 1. Problem Understanding
# =========================================

Given two integers,
find the Greatest Common Divisor (GCD).

GCD is the largest number
that divides both numbers exactly.

Examples:
GCD(20,15) = 5
GCD(9,12) = 3

# =========================================
# 2. Key Observation
# =========================================

GCD cannot be larger than
the smaller number.

So checking beyond min(a,b)
is unnecessary.

If we traverse backward from min(a,b),
the first common divisor found
is automatically the GCD.


# =========================================
# 3. Brute Force Approach
# =========================================

Traverse from 1 to min(a,b).

For every number:
- check if divides both numbers
- store maximum common divisor

# Time Complexity: O(min(a,b))
# Space Complexity: O(1)

# =========================================
# 4. Optimal Brute Force Approach
# =========================================


Traverse backward from min(a,b).

The first number dividing both numbers
is the greatest common divisor.

This avoids unnecessary traversal
after finding answer.


# =========================================
# 5. Algorithm Steps
# =========================================


1. Take two numbers as input
2. Find minimum of both numbers
3. Traverse backward from minimum to 1
4. Check divisibility using %
5. Print first valid divisor
6. Break loop immediately


# =========================================
# 6. Mistakes I Made
# =========================================


- Initially looped from 1 to n
- Used max() unnecessarily
- Forgot to check 1 in range()
- Boundary mistake in reverse traversal


# =========================================
# 7. Dry Run
# =========================================


Input:
x = 20
y = 15

n = 15

i = 15 -> not common divisor
i = 14 -> not common divisor

i = 5 -> divides both

Output:
5

# =========================================
# 8. Final Code
# =========================================

x = int(input())
y = int(input())

n = min(x, y)

for i in range(n, 0, -1):

    if x % i == 0 and y % i == 0:

        print(i)

        break
