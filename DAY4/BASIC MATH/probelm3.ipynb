# =========================================
# 1. Problem Understanding
# =========================================
Given an integer x,
return True if x is palindrome,
otherwise return False.

A palindrome number remains same
even after reversing its digits.

Examples:
121 -> True
-121 -> False
10 -> False

# =========================================
# 2. Key Observation
# =========================================

If reversed number equals original number,
then the number is palindrome.

Core reversal formula:

result = result * 10 + digit

Need to preserve original value
before modifying the number.

# =========================================
# 3. Brute Force Approach
# =========================================

Convert number into string,
reverse the string,
compare with original string.

# Time Complexity: O(k)
# Space Complexity: O(k)

# =========================================
# 4. Optimal Approach
# =========================================

Reverse digits mathematically
using modulo and integer division.

Steps:
- Extract last digit using %10
- Build reversed number
- Remove last digit using //10
- Compare reversed number with original

Handle edge cases:
- Negative numbers are never palindrome
- Numbers ending with 0 are not palindrome
  except 0 itself

# =========================================
# 5. Algorithm Steps
# =========================================

1. Handle negative numbers
2. Store original value
3. Initialize reversed number
4. Extract digits one by one
5. Build reversed number
6. Compare reversed number with original

# =========================================
# 6. Mistakes I Made
# =========================================

- Used false/true instead of False/True
- Wrong loop condition (new < 0)
- Forgot final comparison
- Initially returned True directly

# =========================================
# 7. Dry Run
# =========================================

Input:
x = 121

Initial:
new = 121
result = 0

Iteration 1:
digit = 1
result = 1
new = 12

Iteration 2:
digit = 2
result = 12
new = 1

Iteration 3:
digit = 1
result = 121
new = 0

Compare:
121 == 121

Output:
True

# =========================================
# 8. Final Code
# =========================================

class Solution(object):

    def isPalindrome(self, x):

        if x < 0:
            return False

        new = x
        result = 0

        while new > 0:

            digit = new % 10

            result = result * 10 + digit

            new //= 10

        return result == x
