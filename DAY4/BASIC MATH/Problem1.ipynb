# =========================================
# 1. Problem Understanding
# =========================================


Given an integer N, return the total number of digits present in it.

Examples:
12345 -> 5 digits
7789 -> 4 digits

# =========================================
# 2. Key Observation
# =========================================


Every time we divide a number by 10,
one digit gets removed from the end.

Example:
12345
1234
123
12
1
0

Number of iterations = number of digits

# =========================================
# 3. Brute Force Approach
# =========================================


Convert number into string and return length.

Example:
len(str(n))


# Time Complexity: O(k)
# Space Complexity: O(k)

# =========================================
# 4. Optimal Approach
# =========================================


Use integer division (//10) repeatedly.

Each iteration removes one digit.
Increase count after every division.

Loop stops when number becomes 0.

This avoids extra space usage.

# =========================================
# 5. Algorithm Steps
# =========================================


1. Take input number
2. Handle edge case for n = 0
3. Initialize count = 0
4. Divide number by 10 repeatedly
5. Increase count after every iteration
6. Print final count

# =========================================
# 6. Mistakes I Made

# =========================================
# 7. Dry Run
# =========================================


Input:

Iteration 2:
n = 123
count = 2

Iteration 3:
n = 12
count = 3

Iteration 4:
n = 1

Iteration 5:

Output:

# =========================================
# 8. Final Code
# =========================================

n = int(input())

if n == 0:
    print(1)

else:

    new = n
    count = 0

    while new > 0:

        new = new // 10
        count += 1

    print(count)

# Space Complexity: O(1)
