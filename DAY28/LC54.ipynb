# LC 54 – Spiral Matrix

## Problem
Given an `m x n` matrix, return all elements of the matrix in **spiral order** (clockwise).

---

# Brute Force Approach

### Idea
- Keep track of visited cells using a separate `visited` matrix.
- Move in one direction until you hit a boundary or a visited cell.
- Change direction (Right → Down → Left → Up).
- Continue until all elements are visited.

### Time Complexity
O(m × n)

### Space Complexity
O(m × n)

---

# Optimal Approach (Boundary Traversal)

## Pattern
**Boundary Traversal / Matrix Simulation (Layer-by-Layer Traversal)**

### Key Observation
Instead of marking cells as visited, keep track of the **current unvisited rectangle** using four boundaries:

- Top
- Bottom
- Left
- Right

Traverse one boundary at a time, then shrink it.

---

# Algorithm

### Step 1
Initialize four boundaries.

- top = first row
- bottom = last row
- left = first column
- right = last column

Create an empty answer list.

---

### Step 2
Repeat while there is still an unvisited rectangle.

Condition:

top ≤ bottom AND left ≤ right

---

### Step 3
Traverse **Left → Right**

- Fixed row = top
- Columns = left → right

Append all elements.

Shrink:

top++

---

### Step 4
Traverse **Top → Bottom**

- Fixed column = right
- Rows = top → bottom

Append all elements.

Shrink:

right--

---

### Step 5
If there is still a valid row:

(top ≤ bottom)

Traverse **Right → Left**

- Fixed row = bottom
- Columns = right → left

Append all elements.

Shrink:

bottom--

---

### Step 6
If there is still a valid column:

(left ≤ right)

Traverse **Bottom → Top**

- Fixed column = left
- Rows = bottom → top

Append all elements.

Shrink:

left++

---

### Step 7
Continue until all layers are processed.

Return the answer.

---

# Why are the two `if` conditions needed?

After traversing:

- Left → Right
- Top → Bottom

Sometimes there may be **no remaining rows**.

Checking:

top ≤ bottom

prevents traversing the same row twice.

---

Similarly,

after traversing the bottom row,

there may be **no remaining columns**.

Checking:

left ≤ right

prevents traversing the same column twice.

These checks are especially important for:

- Single row matrices
- Single column matrices
- Rectangular matrices

---

# Time Complexity

O(m × n)

Every element is visited exactly once.

---

# Space Complexity

O(1)

(Excluding the output list.)

---

# Primary Pattern

**Boundary Traversal / Layer-by-Layer Matrix Simulation**

---

# Common Mistakes

- Forgetting `right = cols - 1`
- Using incorrect `range()` boundaries
- Forgetting `right + 1` while traversing left → right
- Using `right++` instead of `right--`
- Forgetting the two `if` conditions
- Visiting elements twice in single-row or single-column matrices
- Updating the wrong boundary after a traversal

---

# Interview Takeaway

The key idea is to **simulate peeling one layer of the matrix at a time**.

The four boundaries always represent the current unvisited rectangle.

After each traversal, shrink the corresponding boundary until no unvisited rectangle remains.

```python
class Solution(object):
    def spiralOrder(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: List[int]
        """
        top=0
        right=len(matrix[0])-1
        bottom=len(matrix)-1
        left=0
        ans=[]
        while top<=bottom and left<=right:
            for i in range(left,right+1):
                ans.append(matrix[top][i])
            top+=1
            for i in range(top,bottom+1):
                ans.append(matrix[i][right])
            right-=1
            if top<=bottom:
                for i in range(right,left-1,-1):
                    ans.append(matrix[bottom][i])
                bottom-=1
            if left<=right:
                for i in range(bottom,top-1,-1):
                    ans.append(matrix[i][left])
                left+=1
        return ans
```
