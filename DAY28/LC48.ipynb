# LC 48 – Rotate Image

## Problem
Given an `n × n` matrix representing an image, rotate the image **90° clockwise** **in-place**.

**In-place** means modifying the original matrix without using another `n × n` matrix.

---

# Brute Force Approach

### Idea
- Create a new matrix.
- Place each element into its rotated position.
- Copy the new matrix back.

### Time Complexity
O(n²)

### Space Complexity
O(n²)

---

# Optimal Approach

## Pattern

**Matrix Transformation + Triangle Traversal + Two Pointers**

The rotation can be achieved in two simple steps:

1. **Transpose the matrix**
2. **Reverse every row**

---

# Key Observation

A 90° clockwise rotation is equivalent to:

Original

1 2 3

4 5 6

7 8 9

↓

Transpose

1 4 7

2 5 8

3 6 9

↓

Reverse every row

7 4 1

8 5 2

9 6 3

---

# Algorithm

### Step 1

Transpose the matrix.

Swap only one triangle (upper or lower).

Swap:

(matrix[i][j], matrix[j][i])

for every element below (or above) the main diagonal.

---

### Step 2

Reverse every row.

For each row:

- left = 0
- right = last column

Swap:

row[left] ↔ row[right]

Move both pointers toward the center until they meet.

---

### Step 3

The matrix is now rotated 90° clockwise.

---

# Dry Run

Original Matrix

1 2 3

4 5 6

7 8 9

---

### After Transpose

Swap:

2 ↔ 4

3 ↔ 7

6 ↔ 8

Result

1 4 7

2 5 8

3 6 9

---

### Reverse Every Row

Row 1

1 4 7

↓

7 4 1

---

Row 2

2 5 8

↓

8 5 2

---

Row 3

3 6 9

↓

9 6 3

---

Final Matrix

7 4 1

8 5 2

9 6 3

---

# Time Complexity

Transpose → O(n²)

Reverse rows → O(n²)

Overall:

O(n²)

---

# Space Complexity

O(1)

No extra matrix is created.

---

# Primary Pattern

**Matrix Transformation**

---

# Secondary Patterns

- Triangle Traversal (Transpose)

---

# Common Mistakes

- Creating another matrix (violates O(1) space)
- Swapping the entire matrix during transpose (double swapping)
- Forgetting to transpose before reversing
- Reversing columns instead of rows for clockwise rotation
- Returning the matrix instead of modifying it in-place
- Incorrect loop bounds during transpose

---

# Interview Takeaway

Instead of memorizing index mappings, remember the transformation:

**90° Clockwise Rotation = Transpose + Reverse Every Row**

Break the problem into two familiar operations:

1. Matrix Transpose
2. Reverse using Two Pointers

This makes the solution simple, efficient, and easy to derive during interviews.|

```python
class Solution(object):
    def rotate(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        n=len(matrix)
        for i in range(n):
            for j in range(i):
                matrix[i][j],matrix[j][i]=matrix[j][i],matrix[i][j]
        
        for i in range(n):
            left=0
            right=n-1
            while left<right:
                matrix[i][left], matrix[i][right] = matrix[i][right], matrix[i][left]
                right-=1
                left+=1
       
```
