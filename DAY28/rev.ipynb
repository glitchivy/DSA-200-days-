# LC 2149

```python
class Solution(object):
    def rearrangeArray(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        ans=[0]*len(nums)
        neg=1
        pos=0
        for i in range(len(nums)):
            if nums[i]<0:
                ans[neg]=nums[i]
                neg+=2
            else:
                ans[pos]=nums[i]
                pos+=2
        return ans
```

# lc31

```python
class Solution(object):
    def nextPermutation(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        pivot=-1
        for i in range(len(nums)-2,-1,-1):
            if nums[i]<nums[i+1]:
                pivot=i
                break
        if pivot==-1:
            nums.reverse()
            return 
        for i in range(len(nums)-1, -1, -1):
            if nums[i]>nums[pivot]:
                nums[pivot],nums[i]=nums[i],nums[pivot]
                break
        left = pivot + 1
        right = len(nums) - 1
        while left<right:
            nums[right],nums[left]=nums[left],nums[right]
            left+=1
            right-=1
        return 
```

# 128

```python
class Solution(object):
    def longestConsecutive(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        num_set=set(nums)
        longest=0
        count=0
        num_set=set(nums)
        for i in num_set:
            if i-1 not in num_set:
                current=i
                count=1
                while current+1 in num_set:
                    current+=1
                    count+=1
                longest=max(count,longest)
        return longest
```
