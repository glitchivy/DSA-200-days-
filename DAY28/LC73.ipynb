# LC 73 – Set Matrix Zeroes

## Problem
Given an `m × n` matrix, if an element is `0`, set its **entire row** and **entire column** to `0`.

The modification must be done **in-place**.

---

# Brute Force Approach

### Idea
- Whenever a `0` is found, immediately set its entire row and column to `0`.

### Problem
This creates **new zeros**, which are mistaken as original zeros while scanning.

This causes a **chain reaction**, making more rows and columns zero than required.

### Time Complexity
O((m × n) × (m + n))

### Space Complexity
O(1)

---

# Better Approach

## Pattern

**Hashing (Using Sets)**

---

# Key Observation

Instead of modifying the matrix immediately,

first **remember** which rows and columns originally contain a zero.

Use:

- One set for rows
- One set for columns

After collecting all original zero positions, update the matrix in a second pass.

This avoids the chain reaction problem.

---

# Algorithm

### Step 1

Create two sets:

- rows
- cols

---

### Step 2

Traverse the entire matrix.

Whenever a cell contains `0`:

- Add its row index to `rows`
- Add its column index to `cols`

---

### Step 3

Traverse the matrix again.

For every cell `(i, j)`:

If

- row `i` exists in `rows`

OR

- column `j` exists in `cols`

then set that cell to `0`.

---

### Step 4

The matrix is now correctly updated.

---

# Dry Run

Input

1 1 1

1 0 1

0 1 1

---

### First Pass

Found zero at (1,1)

rows = {1}

cols = {1}

Found zero at (2,0)

rows = {1,2}

cols = {0,1}

---

### Second Pass

Any cell whose row is in

{1,2}

or whose column is in

{0,1}

becomes zero.

---

Final Matrix

0 0 1

0 0 0

0 0 0

---

# Time Complexity

First Pass

O(m × n)

Second Pass

O(m × n)

Overall

O(m × n)

---

# Space Complexity

O(m + n)

Extra space is used for the two sets.

---

# Primary Pattern

**Hashing (Sets)**

---

# Recognition Clues

If the problem asks you to:

- Remember rows containing a property
- Remember columns containing a property
- Avoid modifying data while scanning
- Prevent chain reactions

Think:

**Use Hash Sets to record information first, then process later.**

---

# Common Mistakes

- Modifying the matrix immediately after finding a zero
- Causing chain reactions from newly created zeros
- Using the number of rows for both loops in a non-square matrix
- Forgetting to perform a second traversal
- Confusing row indices with column indices

---

# Interview Takeaway

The important insight is to **separate information collection from modification**.

First identify **which rows and columns must become zero**.

Only after collecting all original information should the matrix be modified.

This solution is simple and efficient, but it uses **O(m + n)** extra space.

A common interview follow-up is to reduce the space complexity to **O(1)** by using the matrix itself to store the row and column markers.

```python
class Solution(object):
    def setZeroes(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        rows=set()
        col=set()
        m=len(matrix)
        n=len(matrix[0])
        for i in range(m):
            for j in range(n):
                if matrix[i][j]==0:
                    rows.add(i)
                    col.add(j)
        for i in range(m):
            for j in range(n):
                if i in rows or j in col:
                    matrix[i][j]=0
```
