# 1838. Frequency of the Most Frequent Element

````python
class Solution(object):
    def maxFrequency(self, nums, k):
        nums.sort()
        left=0
        sum=0
        ans=0
        for right in range(len(nums)):
            sum+=nums[right]
            while (nums[right]*(right-left+1))-sum>k:
                sum-=nums[left] 
                left+=1
            ans=max(ans,right-left+1)
        return ans
````

# Leetcode 1838 — Frequency of the Most Frequent Element

---

# Problem Understanding

We are given:
- an array `nums`
- an integer `k`

We can perform at most `k` operations.

In one operation:
```python
nums[i] += 1
```

Goal:
Make the maximum number of elements equal.

Return:
```python
maximum possible frequency
```

---

# Important Observation

We can ONLY:
```python
increase numbers
```

We CANNOT:
```python
decrease numbers
```

This changes the entire approach.

---

# Why Sorting?

Suppose:

```python
[1,2,4]
```

It is easier to:
```python
1 → 4
2 → 4
```

than random operations.

After sorting:
- smaller numbers stay on left
- larger numbers stay on right

This helps create valid windows.

---

# Core Pattern

## Sliding Window + Sorting

We maintain:
```python
left
right
```

Current window:
```python
nums[left:right]
```

Goal:
Can ALL elements in this window become:
```python
nums[right]
```

?

Because after sorting:
```python
nums[right]
```

is the largest element in current window.

---

# MOST IMPORTANT FORMULA

Operations needed:

```python
(target * window_size) - window_sum
```

Where:

```python
target = nums[right]
```

```python
window_size = right-left+1
```

```python
window_sum = sum of current window
```

---

# Formula Derivation

Example:

```python
[1,2,4]
```

Target:
```python
4
```

Operations:

```python
(4-1) + (4-2) + (4-4)
```

Expand:

```python
4+4+4 - (1+2+4)
```

becomes:

```python
(4*3)-7
```

becomes:

```python
5
```

---

# Sliding Window Logic

## Step 1
Expand window using:
```python
right += 1
```

---

## Step 2
Add current element to:
```python
window_sum
```

---

## Step 3
Calculate operations needed.

If:
```python
operations > k
```

window becomes invalid.

---

## Step 4
Shrink window from left:

```python
left += 1
```

until window becomes valid again.

---

## Step 5
Track maximum valid window size.

---

# Final Code

```python
class Solution(object):

    def maxFrequency(self, nums, k):

        nums.sort()

        left = 0
        window_sum = 0
        ans = 0

        for right in range(len(nums)):

            window_sum += nums[right]

            while (nums[right] * (right - left + 1)) - window_sum > k:

                window_sum -= nums[left]
                left += 1

            ans = max(ans, right - left + 1)

        return ans
```
