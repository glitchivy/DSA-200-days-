# Hashing


hashing is basically prestored information for fast lookup 

it contains index which is the number and value which is the frequency 

````python
arr = [1,2,1,3,2]

hash_arr = [0] * 4

for num in arr:
    hash_arr[num] += 1

print(hash_arr)
````

````python
arr= list(map(int,input().split()))


hash_arr=[0]* (max(arr)+1)

for i in arr:
    hash_arr[i]+=1
print(hash_arr)
````

# Question 1 — Basic Number Hashing

## Problem Statement

Given an array:

```python
arr = [1,2,1,3,2,1]
```

Find:
- frequency of `1`
- frequency of `2`
- frequency of `3`

using hashing.

````python
arr=[1,2,1,3,2,1]

hash_list=[0]*(max(arr)+1)

for i in arr:
    hash_list[i]+=1
print(hash_list)
````
