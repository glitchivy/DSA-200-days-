# LeetCode 645 - Set Mismatch

## Problem

You are given an integer array `nums` of length `n` containing numbers from `1` to `n`.

One number appears **twice**, while another number is **missing**.

Return an array:

```
[duplicate, missing]
```

**Example:**

Input:

```
nums = [1,2,2,4]
```

Output:

```
[2,3]
```

---

# Brute Force Approach

For every number from `1` to `n`:

- Count its occurrences in the array.
- If the count is `2`, it is the duplicate.
- If the count is `0`, it is the missing number.

### Time Complexity

**O(n²)**

### Space Complexity

**O(1)**

---

# Better Approach (HashMap / Frequency Array)

Store the frequency of every number using a dictionary (or frequency array).

Traverse numbers from `1` to `n`:

- Frequency `2` → Duplicate
- Frequency `0` → Missing

### Time Complexity

**O(n)**

### Space Complexity

**O(n)**

---

# Optimal Approach (Cyclic Sort)

## Key Idea

Since every number belongs to the range **1 to n**, every value has a correct position.

```
1 → index 0
2 → index 1
3 → index 2
...
n → index n-1
```

Keep swapping every number until it reaches its correct position.

Because one number is duplicated, one position can never be filled correctly.

After cyclic sort, scan the array once.

If:

```
nums[i] != i + 1
```

then

- `nums[i]` is the duplicate.
- `i + 1` is the missing number.

---

# Key Observations

- Every number has a fixed correct index.
- Duplicate numbers prevent another number from reaching its correct position.
- After cyclic sort, only one index remains incorrect.
- That incorrect index directly reveals both answers.

---

# Algorithm

1. Traverse the array.
2. While the current number is not at its correct index:
   - Find its correct index.
   - Swap the two numbers.
3. Traverse the array again.
4. Find the index where:
   ```
   nums[i] != i + 1
   ```
5. Return:
   - Duplicate = `nums[i]`
   - Missing = `i + 1`

---

# Dry Run

Input:

```
nums = [1,2,2,4]
```

Initial:

```
Index : 0 1 2 3
Value : 1 2 2 4
```

Cyclic sort:

- 1 is already correct.
- 2 is already correct.
- Another 2 also wants to go to index 1.
- Since index 1 already contains 2, no swap occurs.
- 4 is already correct.

Final array:

```
[1,2,2,4]
```

Traverse again:

```
Index 0 → 1 ✓
Index 1 → 2 ✓
Index 2 → Expected 3, Found 2 ✗
Index 3 → 4 ✓
```

Answer:

```
Duplicate = 2
Missing = 3
```

Return:

```
[2,3]
```

---

# Time Complexity

- Cyclic Sort: **O(n)**
- Final Traversal: **O(n)**

Overall:

**O(n)**

---

# Space Complexity

Only a few variables are used.

**O(1)**

---

# Primary Pattern

**Cyclic Sort**

---

# Recognition Clues

Use this approach when:

- Numbers are in the range `1` to `n`.
- One or more numbers are misplaced.
- Need to find missing or duplicate elements.
- Constant extra space is required.
- Swapping elements into their correct positions is possible.

Keywords:

- Missing Number
- Duplicate Number
- Set Mismatch
- Correct Index
- Cyclic Sort

---

# Common Mistakes

- Forgetting that the correct index is `nums[i] - 1`.
- Using an `if` instead of a `while` during cyclic sort.
- Swapping even when the duplicate is already at its correct position, causing an infinite loop.
- Returning the wrong order (`[missing, duplicate]` instead of `[duplicate, missing]`).
- Forgetting the final traversal after cyclic sort.

---

# Interview Takeaway

The important observation is that every value belongs to a unique index.

By repeatedly placing each number at its correct position, the duplicate prevents the missing number's position from ever being filled.

After cyclic sort, the only mismatched index directly identifies:

- The duplicate number.
- The missing number.

This yields an optimal solution with **O(n)** time and **O(1)** extra space.

````python
class Solution(object):
    def findErrorNums(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        n=len(nums)
        for i in range(n):
            while nums[i]!=nums[nums[i]-1]:
                correct=nums[i]-1
                nums[i],nums[correct]=nums[correct],nums[i]
        for i in range(n):
            if nums[i] != i + 1:
                return [nums[i], i + 1]
````
