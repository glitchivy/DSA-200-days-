# LeetCode 88 - Merge Sorted Array

## Problem

You are given two sorted integer arrays:

- `nums1` of size `m + n`, where the first `m` elements are valid and the last `n` elements are `0`s.
- `nums2` of size `n`.

Merge `nums2` into `nums1` as one sorted array.

The merge must be performed **in-place**.

**Example:**

Input:

```
nums1 = [1,2,3,0,0,0]
m = 3

nums2 = [2,5,6]
n = 3
```

Output:

```
[1,2,2,3,5,6]
```

---

# Brute Force Approach

Copy all elements of `nums2` into `nums1`.

Sort the entire array.

### Time Complexity

- Copying: **O(n)**
- Sorting: **O((m+n) log(m+n))**

Overall:

**O((m+n) log(m+n))**

### Space Complexity

**O(1)**

---

# Better Approach

Use an extra array.

Maintain two pointers:

- One for `nums1`
- One for `nums2`

Compare the smaller element and insert it into a new array.

Finally copy the merged array back to `nums1`.

### Time Complexity

**O(m+n)**

### Space Complexity

**O(m+n)**

---

# Optimal Approach (Three Pointers)

## Key Idea

Instead of merging from the front, merge **from the back**.

Since `nums1` already has empty spaces at the end, placing the largest remaining element there prevents overwriting useful values.

Maintain three pointers:

- `i` → Last valid element in `nums1`
- `j` → Last element in `nums2`
- `k` → Last index of `nums1`

Compare the larger element between `nums1[i]` and `nums2[j]`.

Place it at `nums1[k]`.

Move the corresponding pointer backward.

---

# Key Observations

- Both arrays are already sorted.
- The largest remaining element always belongs at the last available position.
- Filling from the back avoids shifting elements.
- Continue until all elements of `nums2` have been placed.

---

# Algorithm

1. Initialize three pointers:
   - `i = m - 1`
   - `j = n - 1`
   - `k = m + n - 1`
2. While `j >= 0`:
   - If `i >= 0` and `nums1[i] > nums2[j]`:
     - Place `nums1[i]` at `nums1[k]`.
     - Decrement `i`.
   - Otherwise:
     - Place `nums2[j]` at `nums1[k]`.
     - Decrement `j`.
   - Decrement `k`.
3. `nums1` now contains the merged sorted array.

# Common Mistakes

- Merging from the front, which overwrites useful elements.
- Forgetting to check `i >= 0` before accessing `nums1[i]`.
- Using `while i >= 0` instead of `while j >= 0`.
- Forgetting to decrement `k` after every insertion.
- Returning `nums1` instead of modifying it in-place.

---

# Interview Takeaway

The important observation is that the extra space already exists at the end of `nums1`.

By filling the array from the back, the largest remaining element is always placed in its correct position without overwriting unprocessed elements.

This results in an optimal **O(m+n)** time solution with **O(1)** extra space.

````python
class Solution(object):
    def merge(self, nums1, m, nums2, n):
        """
        :type nums1: List[int]
        :type m: int
        :type nums2: List[int]
        :type n: int
        :rtype: None Do not return anything, modify nums1 in-place instead.
        """
        i=m-1
        j=n-1
        k=m+n-1
        while j>=0:
            if i>=0 and nums1[i]>nums2[j]:
                nums1[k]=nums1[i]
                i-=1
                k-=1
            else:
                nums1[k]=nums2[j]
                j-=1
                k-=1
        return nums1
````
