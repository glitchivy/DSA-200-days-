# LeetCode 56 - Merge Intervals

## Problem

You are given an array of intervals where `intervals[i] = [start, end]`.

Merge all overlapping intervals and return an array of the non-overlapping intervals that cover all the intervals.

**Example:**

Input:
```
[[1,3],[2,6],[8,10],[15,18]]
```

Output:
```
[[1,6],[8,10],[15,18]]
```

---

# Brute Force Approach

Compare every interval with every other interval and merge overlapping ones repeatedly until no more merges are possible.

### Time Complexity
- **O(n²)**

### Space Complexity
- **O(n)**

---

# Better Approach

Sort the intervals based on their starting time.

Maintain a separate answer list.

For every interval:
- Compare it with the last interval stored in the answer.
- If they overlap, merge them.
- Otherwise, add the current interval to the answer.

This avoids comparing every pair of intervals.

---

# Optimal Approach (Sorting + Greedy)

### Key Idea

After sorting by the starting point:

- If the current interval overlaps with the last merged interval, extend the ending point.
- Otherwise, start a new interval.

Since the intervals are sorted, only the **last merged interval** needs to be checked.

---

# Key Observations

- Sorting brings all possible overlapping intervals together.
- We only compare the current interval with the last merged interval.
- The merged interval's ending point should always be the maximum ending value.
- Touching intervals also overlap.

Example:

```
[1,4]
[4,5]
```

becomes

```
[1,5]
```

---

# Algorithm

1. Sort the intervals by their starting point.
2. Add the first interval to the answer.
3. Traverse the remaining intervals.
4. Compare the current interval with the last interval in the answer.
5. If they overlap:
   - Update the ending point using the maximum ending value.
6. Otherwise:
   - Append the current interval to the answer.
7. Return the answer.

# Common Mistakes

- Forgetting to sort the intervals.
- Using `>` instead of `>=`, which misses touching intervals.
- Updating the end as `curr[1]` instead of `max(last[1], curr[1])`.
- Comparing with the previous interval instead of the last merged interval.
- Forgetting to add non-overlapping intervals to the answer.

---

# Interview Takeaway

The crucial observation is that after sorting, an interval can only overlap with the **last merged interval**.

Instead of checking all previous intervals, maintain a merged answer list and greedily extend the ending point whenever an overlap occurs.

This reduces the problem to a single linear scan after sorting, giving an optimal complexity of **O(n log n)**.

````python
class Solution(object):
    def merge(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: List[List[int]]
        """
        intervals.sort()
        ans=[]
        ans.append(intervals[0])
        for i in range(1,len(intervals)):
            last=ans[-1]
            curr=intervals[i]
            if last[1] >= curr[0]:
                last[1] = max(last[1], curr[1])
            else:
                ans.append(curr)
        return ans
````
