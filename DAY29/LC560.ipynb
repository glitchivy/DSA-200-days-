# LeetCode 560 - Subarray Sum Equals K

## Problem

Given an integer array `nums` and an integer `k`, return the **total number of continuous subarrays** whose sum equals `k`.

---

## Brute Force

Generate every possible subarray using two nested loops, calculate its sum, and increment the answer whenever the sum equals `k`.

### Time Complexity
- **O(n²)**

### Space Complexity
- **O(1)**

---

## Better

Compute prefix sums in an array first. Then for every pair of indices, calculate the subarray sum using the prefix sum array.

Subarray Sum = `prefix[j] - prefix[i-1]`

### Time Complexity
- **O(n²)**

### Space Complexity
- **O(n)**

---

## Optimal

Maintain a running prefix sum while traversing the array.

Use a HashMap to store:

- **Key:** Prefix Sum
- **Value:** Frequency of that prefix sum

For every element:

1. Update the current prefix sum.
2. Calculate the required previous prefix:
   ```
   required = currentPrefix - k
   ```
3. If `required` exists in the HashMap, add its frequency to the answer.
4. Store the current prefix sum in the HashMap.

Initialize the HashMap with:

```
{0 : 1}
```

to handle subarrays starting from index `0`.

---

## Key Observation

If:

```
Current Prefix - Previous Prefix = k
```

Then,

```
Previous Prefix = Current Prefix - k
```

Instead of checking every subarray, simply check whether the required previous prefix sum has already been seen.

Every occurrence of that prefix sum represents one valid subarray.

---

## Algorithm

1. Create a HashMap with `{0:1}`.
2. Initialize:
   - `prefixSum = 0`
   - `count = 0`
3. Traverse the array.
4. Update the running prefix sum.
5. Compute:
   ```
   required = prefixSum - k
   ```
6. If `required` exists in the HashMap, add its frequency to the answer.
7. Insert/update the current prefix sum in the HashMap.
8. Return the final count.

## Recognition Clues

Use this approach when:

- Asked to count subarrays.
- Subarray sum equals a target.
- Negative numbers are present.
- Sliding window does **not** work.
- Prefix sums can help eliminate repeated calculations.

---

## Common Mistakes

- Updating the HashMap **before** checking for the required prefix.
- Forgetting to initialize the map with `{0:1}`.
- Storing the required value instead of the current prefix sum.
- Using `k - prefixSum` instead of `prefixSum - k`.
- Forgetting that the HashMap stores **frequencies**, not just existence.

---

## Interview Takeaway

The key insight is converting the subarray sum equation into a prefix sum equation:

```
Current Prefix - Previous Prefix = k
```

By storing previously seen prefix sums in a HashMap, every lookup instantly tells us how many valid subarrays end at the current index, reducing the complexity from **O(n²)** to **O(n)**.

````python
class Solution(object):
    def subarraySum(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        prefix_count={0:1}
        sum_prefix=0
        count=0
        for i in nums:
            sum_prefix+=i
            value=sum_prefix-k
            if value in prefix_count:
                count+=prefix_count[value]
            if sum_prefix not in prefix_count:
                prefix_count[sum_prefix]=1
            else:
                prefix_count[sum_prefix]+=1
        return count
````
