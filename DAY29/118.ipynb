# LeetCode 118 - Pascal's Triangle

## Problem

Given an integer `numRows`, return the first `numRows` of Pascal's Triangle.

In Pascal's Triangle:

- Every row starts with `1`.
- Every row ends with `1`.
- Every middle element is the sum of the two elements directly above it.

Example:

```
Input:
numRows = 5

Output:
[
     [1],
    [1,1],
   [1,2,1],
  [1,3,3,1],
 [1,4,6,4,1]
]
```

---

## Brute Force

There is no meaningful brute-force approach for this problem since every row depends on the previous row.

The natural way is to construct the triangle row by row.

---

## Better

Generate each row using the previous row.

For every new row:

- Start with `1`.
- Compute the middle elements by adding adjacent elements from the previous row.
- End with `1`.

---

## Optimal

Store all generated rows in a list.

For each new row:

1. Take the previous row.
2. Create a new row beginning with `1`.
3. Add every adjacent pair from the previous row.
4. Append the ending `1`.
5. Store the new row.

Repeat until all rows are generated.

---

## Key Observation

Every row depends **only on the previous row**.

If:

```
Previous Row:
[1,3,3,1]
```

Then:

```
Current Row:
[1,4,6,4,1]
```

because:

```
4 = 1 + 3
6 = 3 + 3
4 = 3 + 1
```

So every middle element is:

```
current[i] = previous[i] + previous[i+1]
```

---

## Algorithm

1. Initialize the answer with the first row:
   ```
   [[1]]
   ```
2. Repeat `numRows - 1` times.
3. Get the previous row.
4. Create a new row starting with `1`.
5. Traverse the previous row and append the sum of adjacent elements.
6. Append the ending `1`.
7. Add the completed row to the answer.
8. Return the answer.

## Time Complexity

- **O(numRows²)**

Reason:

The total number of elements generated is:

```
1 + 2 + 3 + ... + numRows
```

which equals:

```
O(numRows²)
```

---

## Space Complexity

- **O(numRows²)**

Since the entire Pascal's Triangle is stored.

## Recognition Clues

Use this approach when:

- A structure is built row by row.
- The current row depends only on the previous row.
- Each state is generated from the immediately preceding state.
- The output itself is a matrix or list of lists.

---

## Common Mistakes

- Forgetting to initialize the first row as `[1]`.
- Using the current row instead of the previous row to calculate values.
- Incorrect loop range for middle elements.
- Forgetting to append the ending `1`.
- Forgetting to append the completed row to the final answer.

---

## Interview Takeaway

The key insight is that every row can be generated using only the previous row.

Instead of hardcoding Pascal's Triangle, repeatedly apply these three rules:

1. Start with `1`.
2. Add adjacent pairs from the previous row.
3. End with `1`.

This produces the entire triangle in **O(numRows²)** time while keeping the implementation simple and intuitive.

````python
class Solution(object):
    def generate(self, numRows):
        """
        :type numRows: int
        :rtype: List[List[int]]
        """
        final=[]
        final.append([1])
        for k in range(numRows-1):
            prev=final[-1]
            row=[1]
            for i in range(len(prev)-1):
                row.append(prev[i]+prev[i+1])
            row.append(1)
            final.append(row)
        return final
````
