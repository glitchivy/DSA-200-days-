# Find Second Smallest and Second Largest Element in an array

Example 1:
Input:
 [1, 2, 4, 7, 7, 5]  
Output:
  
Second Smallest : 2  
Second Largest : 5  
Explanation:
  The elements are sorted as 1, 2, 4, 5, 7, 7.  
Hence, the second smallest element is 2, and the second largest element is 5.

Example 2:
Input:
 [1]  
Output:
  
Second Smallest : -1  
Second Largest : -1  
Explanation:
  Since there is only one element in the array, it is both the largest and smallest element.  
Therefore, there is no second smallest or second largest element present.

```python
arr=list(map(int,input().split()))

first_largest=arr[0]
second_largest=float("-inf")

first_small=arr[0]
second_small=float("inf")

n=len(arr)

for i in range(n):
    if arr[i]>first_largest and arr[i]!=first_largest:
        second_largest=first_largest
        first_largest=arr[i]
    if arr[i]>second_largest and arr[i]<first_largest and arr[i]!=first_largest:
        second_largest=arr[i]
    if arr[i]<first_small and arr[i]!=first_small:
        second_small=first_small
        first_small=arr[i]
    if arr[i]<second_small and arr[i]>first_small and arr[i]!=first_small:
        second_small=arr[i]
print(arr)
print(second_largest)
print(second_small)
```
