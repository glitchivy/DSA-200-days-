# Find the Largest element in an array

Example 1:
Input:
 arr[] = {2, 5, 1, 3, 0}  
Output:
 5  
Explanation:
  
5 is the largest element in the array.

Example 2:
Input:
 arr[] = {8, 10, 5, 7, 9}  
Output:
 10  
Explanation:
  
10 is the largest element in the array.

```python
arr=list(map(int,input().split()))
n=len(arr)
value=arr[0]
for i in range(1,n):
    if value<arr[i]:
        value=arr[i]

print(value)
```

# TC- o(N)
# sc - O(1)
