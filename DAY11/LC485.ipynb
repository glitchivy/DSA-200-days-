# Problem:
### LC 485 - Max Consecutive Ones



```python
class Solution(object):
    def findMaxConsecutiveOnes(self, nums):
        """
        :type nums: List[int]
        :rtype: int

        """
        n=len(nums)
        count=0
        value=0
        for i in range(n):
            if nums[i]==1:
                count+=1
                value=max(value,count)
            else:
                count=0
        return value 
            
        
        
```

Pattern:
Array Traversal

Approach:
Maintain two variables:
1. current_count
2. maximum_count

If element == 1:
    increase current_count
    update maximum

Else:
    reset current_count

Time Complexity:
O(n)

Space Complexity:
O(1)
