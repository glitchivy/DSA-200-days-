# Print Name N times using Recursion

```python
# before that lets do an example

def func(count):
    if count==4:
        return 
    print(count)
    func(count+1)
func(1)
```

```python
def name(count,n):
    if count>n:
        return 
    print("Pikachu")
    name(count+1,n)
n=int(input())
name(1,n)
```
