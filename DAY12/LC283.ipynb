# LC 283

Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.

 

Example 1:

Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]
Example 2:

Input: nums = [0]
Output: [0]
 

Constraints:

1 <= nums.length <= 104
-231 <= nums[i] <= 231 - 1

````python
class Solution(object):
    def moveZeroes(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        j=0
        if not nums:
            return nums
        n=len(nums)
        for i in range(n):
            if nums[i]!=0:
                nums[j], nums[i] = nums[i], nums[j]
                j+=1
        return nums
````

# LeetCode 283 - Move Zeroes

## Pattern
- Two Pointers (Read & Write)
- In-place Array Manipulation

## Approach
- Maintain two pointers:
  - `i` → Read Pointer (traverses the entire array).
  - `j` → Write Pointer (points to the next position where a non-zero element should be placed).
- Traverse the array once.
- If `nums[i]` is non-zero:
  - Swap `nums[i]` with `nums[j]`.
  - Increment `j`.
- By the end of the traversal, all non-zero elements are shifted to the front while all zeros automatically move to the end.

## Invariant
At every iteration:
- `nums[0...j-1]` contains all non-zero elements in their original relative order.
- `nums[j...i]` contains elements that are yet to be processed or zeros.

## Why Swapping?
Using

```python
nums[j], nums[i] = nums[i], nums[j]
```

prevents overwriting values when `i == j`.

Example:

Input:
[1]

If we write:

```python
nums[j] = nums[i]
nums[i] = 0
```

the array becomes:

```text
[0]
```

which is incorrect.

Swapping safely handles both cases:
- `i == j`
- `i != j`

## Time Complexity
- **O(n)**
- Every element is visited exactly once.

## Space Complexity
- **O(1)**
- Only two pointers (`i` and `j`) are used.

## Mistakes I Made
- Initially tried to overwrite the element and manually assign `0`.
- This failed when `i == j`, because the current non-zero element was overwritten.
- Learned that swapping is the correct approach for in-place movement while preserving the array.
