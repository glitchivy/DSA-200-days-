# LeetCode 18 - 4Sum

## Problem

Given an array `nums` of `n` integers, return all unique quadruplets `[nums[a], nums[b], nums[c], nums[d]]` such that:

- `a`, `b`, `c`, and `d` are distinct indices.
- `nums[a] + nums[b] + nums[c] + nums[d] == target`.

The solution should not contain duplicate quadruplets.

---

# Brute Force

Generate every possible combination of four numbers using four nested loops.

For every combination:

- Calculate the sum.
- If the sum equals the target, store the quadruplet.
- Use a set to remove duplicates.

### Time Complexity

**O(n⁴)**

### Space Complexity

**O(1)** (excluding output)

---

# Better

Sort the array.

Instead of fixing all four numbers:

- Fix the first number.
- Fix the second number.
- Use two pointers (`left` and `right`) to find the remaining two numbers.

If the sum is:

- Less than target → move `left`.
- Greater than target → move `right`.
- Equal to target → store the quadruplet.

A set can be used to remove duplicate quadruplets.

### Time Complexity

**O(n³)**

### Space Complexity

**O(n)** (because of the set)

---

# Optimal

Sort the array.

Instead of using a set:

- Skip duplicate values for `i`.
- Skip duplicate values for `j`.
- After finding a valid quadruplet, skip duplicate values for both `left` and `right`.

This prevents duplicate quadruplets from being generated in the first place.

### Time Complexity

**O(n³)**

### Space Complexity

**O(1)** (excluding output)

---

# Key Observations

- Sorting enables the two-pointer approach.
- 4Sum is simply an extension of 3Sum.
- After fixing two numbers, the remaining problem becomes a 2Sum solved using two pointers.
- The search space reduces from four nested loops to two loops and one two-pointer traversal.
- Duplicate skipping is better than generating duplicates and removing them later.

---

# Algorithm

1. Sort the array.
2. Iterate through every possible first element.
3. Iterate through every possible second element.
4. Initialize:
   - `left = j + 1`
   - `right = n - 1`
5. While `left < right`:
   - Compute the sum of the four numbers.
   - If the sum is smaller than the target, move `left`.
   - If the sum is larger than the target, move `right`.
   - Otherwise:
     - Store the quadruplet.
     - Move both pointers.
     - Skip duplicate values (optimal solution).
6. Return all unique quadruplets.


Continue moving pointers until they cross.

Repeat for all valid choices of `i` and `j`.

- **O(n)** (set + output)

### Optimal

Time:

- **O(1)** (excluding output)

# Primary Pattern

- Sorting
- Two Pointers
- k-Sum Reduction

---

# Recognition Clues

Look for this approach when:

- The problem asks for four numbers satisfying a condition.
- The array can be sorted.
- Duplicate answers must be avoided.
- Multiple elements must sum to a target.
- The problem can be reduced recursively:
  - 4Sum → 3Sum → 2Sum

---

# Common Mistakes

- Starting `j` from `i` instead of `i + 1`.
- Setting `left = i + 1` instead of `j + 1`.
- Forgetting to sort the array.
- Forgetting duplicate handling.
- Using the same index more than once.
- Returning tuples instead of lists.
- Forgetting to move both pointers after finding a valid quadruplet.

---

# Interview Takeaway

The natural progression of k-Sum problems is:

- 2Sum → Two Pointers → **O(n)**
- 3Sum → One fixed loop + Two Pointers → **O(n²)**
- 4Sum → Two fixed loops + Two Pointers → **O(n³)**

General rule:

- Fix one element repeatedly until the problem becomes a 2Sum.
- Solve the final 2Sum using the two-pointer technique.

If an interviewer asks for **O(n²)**, discuss the **pair-sum hash map** approach, which reduces the time complexity to **O(n²)** at the cost of **O(n²)** extra space.

````python
class Solution(object):
    def fourSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        nums.sort()
        n=len(nums)
        arr=set()
        for i in range(n-3):
            for j in range(i+1,n-2):
                left=j+1
                right=n-1
                while left<right:
                    total=nums[i]+nums[j]+nums[left]+nums[right]
                    if total<target:
                        left+=1
                    elif total>target:
                        right-=1
                    else:
                        arr.add((nums[i],nums[j],nums[left],nums[right]))
                        left+=1
                        right-=1
        final=[]
        for i in arr:
            final.append(list(i))
        return final
````
