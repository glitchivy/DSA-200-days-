```python
# 41
class Solution(object):
    def firstMissingPositive(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n=len(nums)
        for i in range(n):
            while nums[i]>0 and nums[i]<=len(nums):
                actual=nums[i]-1
                if nums[i]!=nums[actual]:
                    nums[i],nums[actual]=nums[actual],nums[i]
                else:
                    break
        for i in range(n):
            if nums[i]!=i+1:
                return i+1
        return n+1
```

```python
# 560
class Solution(object):
    def subarraySum(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        prefix_count={0:1}
        sum_prefix=0
        count=0
        for i in nums:
            sum_prefix+=i
            target=sum_prefix-k
            if target in prefix_count:
                count+=prefix_count[target]
            if sum_prefix in prefix_count:
                prefix_count[sum_prefix]+=1
            else:
                prefix_count[sum_prefix]=1
        return count
```
