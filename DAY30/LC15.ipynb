# LeetCode 15 – 3Sum

## Problem

Given an integer array `nums`, return all the unique triplets `[nums[i], nums[j], nums[k]]` such that:

- `i != j`
- `j != k`
- `i != k`
- `nums[i] + nums[j] + nums[k] == 0`

The solution set must not contain duplicate triplets.

---

## Brute Force Approach

Try every possible combination of three elements using three nested loops.

For each triplet:

- Check whether its sum equals `0`.
- Store the triplet if valid while avoiding duplicates.

### Time Complexity

- **O(n³)**

### Space Complexity

- **O(1)**

---

## Better Approach (Sorting + Two Pointers + Set)

Sort the array.

Fix one element at a time.

For every fixed element, use the Two Pointer technique to find the remaining two numbers whose sum equals the negative of the fixed element.

Store every valid triplet inside a **set** as a tuple to automatically remove duplicates.

Finally, convert the tuples back into lists.

### Time Complexity

- Sorting: **O(n log n)**
- Two Pointer Search: **O(n²)**

Overall:

- **O(n²)**

### Space Complexity

- **O(n)**

(Set used to remove duplicates.)

---

## Optimal Approach (Sorting + Two Pointers + Duplicate Skipping)

Sort the array first.

Fix one element at index `i`.

Now the problem becomes finding two numbers whose sum equals `-nums[i]`.

Use two pointers:

- `left = i + 1`
- `right = n - 1`

Calculate the sum of the three elements.

- If the sum is too small, move the left pointer.
- If the sum is too large, move the right pointer.
- If the sum is zero, store the triplet and move both pointers.

Instead of storing duplicates inside a set, skip consecutive duplicate values for the fixed element, left pointer, and right pointer. This avoids generating duplicate triplets while keeping the extra space constant.

---

## Key Observations

- Sorting allows the Two Pointer technique.
- Fixing one element reduces the problem to **Two Sum**.
- Moving the left pointer increases the sum.
- Moving the right pointer decreases the sum.
- Duplicate triplets can be avoided by skipping consecutive equal elements.
- No hash map or set is required in the optimal solution.

---

## Algorithm

1. Sort the array.
2. Traverse the array and fix one element.
3. Skip duplicate fixed elements.
4. Initialize two pointers:
   - `left = i + 1`
   - `right = n - 1`
5. While `left < right`:
   - Compute the total sum.
   - If the sum is less than `0`, move `left`.
   - If the sum is greater than `0`, move `right`.
   - Otherwise:
     - Store the triplet.
     - Move both pointers.
     - Skip duplicate values on both sides.
6. Return the answer.

## Dry Run

**Input**

`nums = [-1,0,1,2,-1,-4]`

### Step 1

Sort the array.

`[-4,-1,-1,0,1,2]`

---

### Fix `-4`

Need two numbers whose sum is `4`.

No valid pair exists.

---

### Fix first `-1`

Need two numbers whose sum is `1`.

```
left = -1
right = 2

-1 + (-1) + 2 = 0
```

Store:

`[-1,-1,2]`

Move both pointers.

```
left = 0
right = 1

-1 + 0 + 1 = 0
```

Store:

`[-1,0,1]`

Move both pointers.

---

### Fix second `-1`

Skip because it is the same as the previous fixed element.

---

### Remaining elements

No more valid triplets.

```
[
 [-1,-1,2],

]
```

## Recognition Clues

Use this pattern when:

- The array can be sorted.
- You need to find triplets satisfying a condition.
- The problem asks for unique combinations.
- You need an optimal solution with constant extra space.
- A fixed element reduces the problem to Two Sum.

---

## Common Mistakes

- Forgetting to sort the array.
- Moving the wrong pointer when the sum is too small or too large.
- Not moving both pointers after finding a valid triplet.
- Not skipping duplicate fixed elements.
- Not skipping duplicate left and right values.
- Using a set when the interviewer expects **O(1)** extra space.
- Iterating until `n` instead of `n - 2`.

---

## Interview Takeaway

The key insight is to sort the array and reduce the problem to **Two Sum** by fixing one element. The Two Pointer technique efficiently searches for the remaining two numbers in linear time. Skipping duplicate values ensures that only unique triplets are generated, eliminating the need for extra data structures. This achieves the optimal solution with **O(n²)** time and **O(1)** extra space.

````python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        n=len(nums)
        nums.sort()
        ans=set()
        for i in range(n):
            left=i+1
            right=n-1
            while left<right:
                total=nums[i]+nums[left]+nums[right]
                if total<0:
                    left+=1
                elif total>0:
                    right-=1
                else: 
                    ans.add((nums[i],nums[left],nums[right]))
                    left+=1
                    right-=1
        final=[]
        for triplet in ans:
            final.append(list(triplet))
        return final
````
