# LeetCode 229 – Majority Element II

## Problem

Given an integer array `nums` of size `n`, return all elements that appear **more than ⌊n/3⌋ times**.

You may return the answer in any order.

---

## Brute Force Approach

For every unique element, traverse the entire array and count its occurrences.

If the count is greater than `n/3`, add it to the answer.

### Time Complexity
- **O(n²)**

### Space Complexity
- **O(1)**

---

## Better Approach (Hash Map)

Use a hash map to store the frequency of every element.

Traverse the map and collect all elements whose frequency is greater than `n/3`.

### Time Complexity
- **O(n)**

### Space Complexity
- **O(n)**

---

## Optimal Approach (Extended Boyer-Moore Voting Algorithm)

The optimal approach uses the **Extended Boyer-Moore Voting Algorithm**. Since an array can have at most **two elements** occurring more than **n/3** times, maintain two candidates and their counts. During the first pass, different elements cancel each other out, eliminating values that cannot be majority elements while preserving possible candidates. After this pass, the remaining candidates are only potential majority elements. Perform a second pass to count their actual frequencies. Any candidate appearing more than **n/3** times is added to the answer. This achieves **O(n)** time and **O(1)** extra space.

---

## Key Observations

- An array can have **at most two** elements occurring more than `n/3` times.
- Three different elements cannot each appear more than `n/3` times because their total frequency would exceed `n`.
- The first pass does **not** find the final answer.
- It only identifies **possible candidates**.
- A second pass is necessary to verify their actual frequencies.

---

## Algorithm

1. Initialize two candidates and two counters.
2. Traverse the array:
   - If the current element matches a candidate, increase its count.
   - Else if a counter is zero, assign the current element as the new candidate.
   - Otherwise, decrease both counters.
3. Reset both counters.
4. Traverse the array again and count the actual occurrences of both candidates.
5. Add any candidate whose frequency is greater than `n/3` to the answer.
6. Return the answer.

---

## Dry Run

**Input:**

`nums = [3,2,3,1,3,2,2]`

### First Pass (Voting)

| Element | Candidate 1 | Count 1 | Candidate 2 | Count 2 |
|---------|-------------|---------|-------------|---------|
| 3 | 3 | 1 | - | 0 |
| 2 | 3 | 1 | 2 | 1 |
| 3 | 3 | 2 | 2 | 1 |
| 1 | 3 | 1 | 2 | 0 |
| 3 | 3 | 2 | 2 | 0 |
| 2 | 3 | 2 | 2 | 1 |
| 2 | 3 | 2 | 2 | 2 |

Potential candidates:

- Candidate 1 = 3
- Candidate 2 = 2

### Second Pass

Count actual frequencies:

- 3 → 3 times
- 2 → 3 times

Since `n = 7`

`n/3 = 2`

Both frequencies are greater than `2`.

Answer:

`[3,2]`

---

## Time Complexity

- First Pass: **O(n)**
- Second Pass: **O(n)**

Overall:

**O(n)**

---

## Space Complexity

Only two candidates, two counters, and an answer list (maximum size 2) are used.

**O(1)**

---

## Primary Pattern

- Boyer-Moore Voting Algorithm
- Majority Voting
- Candidate Elimination

---

## Recognition Clues

This pattern is useful when:

- The problem asks for a majority element.
- The threshold is based on the array size (`> n/2`, `> n/3`, etc.).
- The interviewer asks for **constant extra space**.
- Frequency counting with a hash map is not allowed.

---

## Common Mistakes

- Forgetting the second verification pass.
- Assuming the first-pass counts are actual frequencies.
- Returning candidates immediately after the first pass.
- Performing the verification inside the first loop.
- Returning a single value instead of a list.
- Using `if` instead of `elif` while updating candidates.

---

## Interview Takeaway

The Extended Boyer-Moore Voting Algorithm works by repeatedly cancelling different elements. Since there can be at most two elements occurring more than `n/3` times, only two candidates need to be maintained. The first pass identifies potential candidates through elimination, while the second pass verifies their actual frequencies. This provides the optimal solution with **O(n)** time and **O(1)** extra space.

```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        can1= None
        can2= None
        count1=0
        count2=0
        for i in nums:
            if i==can1:
                count1+=1
            elif i == can2:
                count2+=1
            elif count1==0:
                can1=i
                count1=1
            elif count2==0:
                can2=i
                count2=1
            else:
                count1-=1
                count2-=1
        count1=0
        count2=0
        ans=[]
        for i in nums:
            if i ==can1:
                count1+=1
            elif i==can2:
                count2+=1
        if count1>len(nums)//3:
            ans.append(can1)
        if count2>len(nums)//3:
            ans.append(can2)
        return ans
```
