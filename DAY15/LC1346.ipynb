# LeetCode 1346 - Check If N and Its Double Exist

## Problem

Given an integer array `arr`, return `True` if there exist two different indices `i` and `j` such that:

```text
arr[i] = 2 × arr[j]
```

Otherwise, return `False`.

---

## Pattern

**Hash Set (One-Pass Lookup)**

---

## Intuition

For every number `x`, we need to know:

- Has its **double (`2*x`)** already appeared?
- If `x` is even, has its **half (`x//2`)** already appeared?

A **Hash Set** is perfect because it allows **O(1)** average lookup.

Instead of searching the entire array every time, we only search inside the set of numbers we've already seen.

---

## Algorithm

1. Create an empty set called `seen`.
2. Traverse the array one element at a time.
3. For every number:
   - Check if its double exists in `seen`.
   - If the number is even, check if its half exists in `seen`.
4. If either condition is true, return `True`.
5. Otherwise, add the current number to `seen`.
6. If the loop finishes, return `False`.

## Why Do We Check Half Only For Even Numbers?

Suppose

```text
num = 3
```

Half would be

```text
3 // 2 = 1
```

But

```text
2 × 1 = 2
```

Not `3`.

So only even numbers can have an integer half.

Correct check:

```python
if num % 2 == 0 and num // 2 in seen:
```

## Time Complexity

### O(n)

Reason:

- Traverse the array once.
- Hash Set lookup is **O(1)** on average.

Overall

```text
O(n)
```

---

## Space Complexity

### O(n)

Reason:

The `seen` set may store every element.

---

## Common Mistakes

### Mistake 1

Using nested loops.

```python
for i:
    for j:
```

Time Complexity

```text
O(n²)
```

Very inefficient.

---

### Mistake 2

Checking

```python
num // 2
```

for every number.

Wrong because odd numbers do not have an integer half.

Always check

```python
num % 2 == 0
```

first.

---

### Mistake 3

Adding the current number before checking.

Wrong order:

```python
seen.add(num)
```

then checking.

Correct order:

1. Check
2. Add

Otherwise, one element could incorrectly match itself.

# code

````python
class Solution(object):
    def checkIfExist(self, arr):
        """
        :type arr: List[int]
        :rtype: bool
        """
        seen=set()
        for num in arr:
            if num * 2 in seen:
                return True
            elif num % 2 == 0 and num // 2 in seen:
                return True
            else:
                seen.add(num)
        return False
````
