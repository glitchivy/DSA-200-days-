# LeetCode 268 - Missing Number

## Problem

Given an array `nums` containing `n` distinct numbers in the range `[0, n]`, return the only number in the range that is missing from the array.

---

## Pattern

**Mathematical Formula (Expected vs Actual)**

---

## Intuition

If no number were missing, the numbers would be:

```
0, 1, 2, 3, ..., n
```

Calculate:

- The **expected sum** of all numbers from `0` to `n`.
- The **actual sum** of the given array.

The difference between these two sums is the missing number.

---

## Algorithm

1. Find the length of the array `n`.
2. Calculate the expected sum using:

```python
expected = n * (n + 1) // 2
```

3. Calculate the actual sum of the array.

```python
actual = sum(nums)
```

4. Return:

```python
expected - actual
```

---

## Why Does This Work?

The array should contain every number from `0` to `n`.

If one number is missing:

```
Expected Sum = Actual Sum + Missing Number
```

Therefore,

```
Missing Number = Expected Sum - Actual Sum
```

## Formula Used

Sum of first `n` natural numbers:

```text
1 + 2 + 3 + ... + n
```

Formula:

```text
n × (n + 1)
------------
      2
```

Since adding `0` does not change the sum, this also represents the sum from `0` to `n`.

---

## Code

```python
class Solution(object):
    def missingNumber(self, nums):
        n = len(nums)

        expected = n * (n + 1) // 2
        actual = sum(nums)

        return expected - actual
```

---

## Time Complexity

### O(n)

Reason:

- Calculating the expected sum is **O(1)**.
- Calculating `sum(nums)` takes **O(n)**.

Overall:

```
O(1) + O(n) = O(n)
```

---

## Space Complexity

### O(1)

Reason:

Only two integer variables are used.

---

## Alternative Approaches

| Approach | Time | Space |
|----------|------|-------|
| Brute Force (`i in nums`) | O(n²) | O(1) |
| Hash Set | O(n) | O(n) |
| Sum Formula | **O(n)** | **O(1)** |
| XOR | **O(n)** | **O(1)** |

---

## Common Mistakes

### Mistake 1

Checked:

```python
if nums[i] not in nums
```

Problem:

Every element in the array is obviously present in the array.

Correct idea:

```python
if i not in nums
```

---

### Mistake 2

Looped only till:

```python
range(n)
```

Problem:

The missing number can also be `n`.

Correct:

```python
range(n + 1)
```

---

### Mistake 3

Forgot that `sum(nums)` is **O(n)**.

The mathematical formula is constant time, but computing the array sum still requires traversing the array.

---

## Interview Takeaways

- When the problem gives numbers in a fixed range (`0` to `n`), compare **what should exist** with **what actually exists**.
- Always think about:
  - Mathematical formulas
  - XOR
  - Hashing
- Learn to derive the formula instead of memorizing it.

---

## Recognition Trigger

If a problem says:

- Numbers are from a fixed range.
- Exactly one value is missing.

Think:

> **Can I compare the expected total with the actual total?**

This often leads to either the **Sum Formula** or **XOR** solution.

# brute

````python
class Solution(object):
    def missingNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n=len(nums)
        for i in range(n+1):
            if i not in nums:
                return i
               
        return i
````

# optimal

````python
class Solution(object):
    def missingNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n=len(nums)
        expected=n*(n+1)//2
        actual=sum(nums)
        return expected-actual
````
