# LeetCode 941 - Valid Mountain Array

## Problem

Given an integer array `arr`, return `True` if it is a valid mountain array, otherwise return `False`.

A valid mountain array must:
- Have at least 3 elements.
- Strictly increase to a single peak.
- Strictly decrease after the peak.
- The peak cannot be the first or last element.

---

## Pattern

**Array Traversal (Single Pointer)**

Unlike previous two-pointer problems, this problem only requires one pointer that traverses the array.

---

## Intuition

Imagine climbing a mountain.

1. Start from the beginning.
2. Keep climbing while the next element is larger.
3. Once climbing stops, you've reached the peak.
4. The peak cannot be the first or last element.
5. From the peak, keep walking downhill.
6. If you reach the last element while only moving downhill, it's a valid mountain.

The key insight is that **we don't search for the maximum element**. The pointer naturally stops at the peak.

---

## Algorithm

1. If the array length is less than 3, return `False`.
2. Initialize `i = 0`.
3. Move `i` forward while the array is strictly increasing.
4. If `i` is at the first or last index, return `False`.
5. Continue moving `i` while the array is strictly decreasing.
6. Return `True` only if `i` reaches the last index.

---

## Dry Run

### Input

```text
[0,2,4,6,5,3,1]
```

### Climbing Phase

| i | Compare | Action |
|---|---------|--------|
|0|0 < 2|Move|
|1|2 < 4|Move|
|2|4 < 6|Move|
|3|6 < 5|Stop|

Peak found at value **6**.

---

### Peak Validation

Peak is neither the first nor the last element.

Continue.

---

### Descending Phase

| i | Compare | Action |
|---|---------|--------|
|3|6 > 5|Move|
|4|5 > 3|Move|
|5|3 > 1|Move|

Reached the last index.

Return **True**.

---

## Edge Cases

### Too Small

```text
[1,2]
```

False

---

### Only Increasing

```text
[1,2,3]
```

False

---

### Only Decreasing

```text
[3,2,1]
```

False

---

### Equal Adjacent Elements

```text
[1,2,2,1]
```

False

Mountain arrays must be **strictly** increasing and **strictly** decreasing.

---

### Multiple Peaks

```text
[1,3,2,4]
```

False

Once the array starts decreasing, it must never increase again.

## Time Complexity

**O(n)**

### Reason

Each element is visited at most once.

---

## Space Complexity

**O(1)**

### Reason

Only one pointer is used.

---

## Common Mistakes

### Mistake 1

Trying to find the maximum element.

**Learning**

The pointer naturally stops at the peak.

No need for `max()` or a `peak` variable.

---

### Mistake 2

Using `for` instead of `while`.

**Learning**

We don't want to visit every element.

We only want to continue while the mountain keeps increasing or decreasing.

---

### Mistake 3

Checking

```python
arr[i] < arr[i+1] and i+1 < len(arr)
```

This may cause an `IndexError`.

Always check the boundary first.

Correct:

```python
i + 1 < len(arr) and arr[i] < arr[i+1]
```

---

### Mistake 4

Using `>=` during the descending phase.

Wrong:

```python
arr[i] >= arr[i+1]
```

Correct:

```python
arr[i] > arr[i+1]
```

The mountain must be **strictly** decreasing.

---

## Interview Takeaways

- Don't force every array problem into a two-pointer solution.
- Sometimes a single traversal is simpler.
- Let the traversal naturally identify important positions (like the peak) instead of explicitly searching for them.
- Always perform boundary checks before accessing an array index.
- Translate the problem into a real-world analogy. In this case:
  - Walk uphill.
  - Reach the peak.
  - Walk downhill.

## Personal Learning Notes

- Initially tried solving it using two pointers because of previous problems.
- Realized not every array problem uses the same pattern.
- Learned to think about the **shape of the array** instead of finding the maximum element.
- Understood the difference between comparing **indices** (`i`) and **values** (`arr[i]`).
- Learned why boundary conditions must always be checked before accessing `arr[i+1]`.

````python
class Solution(object):
    def validMountainArray(self, arr):
        """
        :type arr: List[int]
        :rtype: bool
        """
        if len(arr)<3:
            return False
        i=0
        while i + 1 < len(arr) and arr[i]<arr[i+1]:
            i+=1
        if i == 0 or i == len(arr) - 1:
            return False
        while i + 1 < len(arr) and arr[i] > arr[i+1]:
            i+=1
        return i == len(arr)-1
````
