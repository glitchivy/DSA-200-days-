# LeetCode 349 - Intersection of Two Arrays

## Problem

Given two integer arrays `nums1` and `nums2`, return an array containing **only the unique elements** that appear in **both** arrays.

The order of the output does **not** matter.

---

## Pattern

**Hash Set**

---

## Intuition

The question asks for:

- Common elements
- No duplicates

Whenever you hear:

- **Unique**
- **Intersection**
- **Membership Checking**

Think:

> **Set**

A set automatically removes duplicates and provides fast lookup.

---

## Algorithm

1. Convert `nums1` into a set.
2. Convert `nums2` into a set.
3. Create an empty result set.
4. Iterate through one set.
5. If the current element exists in the other set, add it to the result.
6. Convert the result set into a list and return it.

---

## Code

```python
class Solution(object):
    def intersection(self, nums1, nums2):
        set1 = set(nums1)
        set2 = set(nums2)

        result = set()

        for num in set1:
            if num in set2:
                result.add(num)

        return list(result)
```

---

## Why Does This Work?

A set:

- Stores only unique elements.
- Removes duplicates automatically.
- Allows fast membership checking.

Since duplicates are already removed, every common element is added only once.

---

## Time Complexity

Let:

- `n = len(nums1)`
- `m = len(nums2)`

Creating sets:

```text
O(n) + O(m)
```

Traversing one set:

```text
O(n)
```

Membership check:

```text
O(1) average
```

### Overall

```text
O(n + m)
```

---

## Space Complexity

Two sets + result set

```text
O(n + m)
```

---

## Common Mistakes

### Mistake 1

Using lists instead of sets.

Problem:

Duplicates remain.

---

### Mistake 2

Checking every element using nested loops.

```python
for i in nums1:
    for j in nums2:
```

Time Complexity:

```text
O(n × m)
```

Very inefficient.

---

### Mistake 3

Returning a set.

```python
return result
```

The function expects:

```python
List[int]
```

Correct:

```python
return list(result)
```

---

## Alternative Python Solution

Python provides a built-in set intersection operator.

```python
return list(set(nums1) & set(nums2))
```

This is shorter but less expressive in interviews.

## Interview Takeaways

- Whenever a problem asks for **common elements**, think about **set intersection**.
- Hash sets are useful because membership checking is approximately **O(1)**.

# code

````python
class Solution(object):
    def intersection(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        set1 = set(nums1)
        set2 = set(nums2)
        result=set()
        for i in set1:
            if i in set2:
                result.add(i)
        return list(result) 
````
