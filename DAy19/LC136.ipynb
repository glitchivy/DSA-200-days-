# LeetCode 136 - Single Number

## Problem

Given a non-empty integer array `nums`, every element appears **twice** except for one element.

Return the element that appears only once.

---

## Pattern

**Bit Manipulation (XOR)**

---

## Intuition

Every duplicate number cancels itself using XOR.

Only the element that appears once remains.

---

## XOR Properties

### Property 1

```text
a ^ a = 0
```

A number XOR with itself becomes zero.

Example

```text
5 ^ 5 = 0
```

---

### Property 2

```text
0 ^ a = a
```

Zero does not affect the number.

Example

```text
0 ^ 8 = 8
```

---

### Property 3 (Commutative)

```text
a ^ b = b ^ a
```

The order of XOR does not matter.

---

### Property 4 (Associative)

```text
(a ^ b) ^ c = a ^ (b ^ c)
```

Grouping does not matter.

Because of Properties 3 and 4, duplicate numbers can always be grouped together.

---

## Key Observation

Every duplicate pair becomes zero.

Example

```text
4 ^ 1 ^ 2 ^ 1 ^ 2
```

Rearrange

```text
4 ^ (1 ^ 1) ^ (2 ^ 2)
```

↓

```text
4 ^ 0 ^ 0
```

↓

```text
4
```

The duplicates disappear.

Only the unique number remains.

---

## Algorithm

1. Initialize `ans = 0`.
2. Traverse every element in the array.
3. XOR the current element with `ans`.
4. Return `ans`.

## Visual Understanding

Think of XOR as a **magic eraser**.

Every duplicate pair cancels itself.

```text
2 ^ 2 = 0
5 ^ 5 = 0
9 ^ 9 = 0
```

Only the element without a duplicate survives.

---

## Why Does This Work?

Every duplicate contributes

```text
a ^ a = 0
```

All duplicates become zero.

Finally,

```text
0 ^ unique = unique
```

The remaining value is the required answer.

---

## Time Complexity

### O(n)

Reason:

The array is traversed exactly once.

---

## Space Complexity

### O(1)

Reason:

Only one variable (`ans`) is used.

No extra data structures are required.

---

## Comparison

| Approach | Time | Space |
|----------|------|------:|
| Dictionary | O(n) | O(n) |
| XOR | O(n) | O(1) ✅ |

---

## Common Mistakes

### Mistake 1

Using a dictionary immediately.

It works,

but it uses extra memory.

The XOR solution is more optimal.

---

### Mistake 2

Thinking XOR adds numbers.

It does **not** perform addition.

It compares bits.

---

### Mistake 3

Forgetting the XOR properties.

Always remember:

```text
a ^ a = 0
```

and

```text
0 ^ a = a
```

These two rules are enough to solve the problem.

# my method 1


````python
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        result={}
        for i in range(len(nums)):
            if nums[i] in result:
                result[nums[i]]+=1
            else:
                result[nums[i]]=1
        for key,value in result.items():
            if value==1:
                return key
````

````python
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        ans=0
        for i in nums:
            ans^=i
        
        return ans
````
