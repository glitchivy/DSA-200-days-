# LeetCode 485 - Max Consecutive Ones

## Problem

Given a binary array `nums`, return the maximum number of consecutive `1`s in the array.

---

## Pattern

**Simple Array Traversal (Running State + Best Answer)**

---

## Intuition

Traverse the array once.

- Keep track of the current streak of consecutive `1`s.
- Whenever a `0` appears, the streak breaks and resets.
- Continuously store the maximum streak encountered.

---

## Key Observation

Two variables are sufficient:

- **Current Streak** → Number of consecutive `1`s seen so far.
- **Maximum Streak** → Largest streak found during traversal.

No extra data structures are required.

---

## Algorithm

1. Initialize:
   - `count = 0`
   - `max_count = 0`
2. Traverse the array.
3. If the current element is `1`:
   - Increase `count`.
4. Otherwise:
   - Reset `count` to `0`.
5. Update the maximum streak.
6. Return the maximum streak.

---

## Dry Run

### Input

```text
[1,1,0,1,1,1]
```

| Element | Current Streak | Maximum Streak |
|---------|---------------:|---------------:|
|1|1|1|
|1|2|2|
|0|0|2|
|1|1|2|
|1|2|2|
|1|3|3|

---

### Final Output

```text
3
```

---

## Why Does This Work?

Think of walking through the array.

Every time you encounter:

### `1`

Continue the streak.

```text
1 1 1
```

Current streak keeps increasing.

---

### `0`

The streak ends.

```text
1 1 1 0
```

Reset the current streak back to zero.

---

Throughout the traversal,

keep remembering the largest streak seen so far.

---

## Time Complexity

### O(n)

Reason:

Each element is visited exactly once.

---

## Space Complexity

### O(1)

Reason:

Only two variables are maintained.

---

## Common Mistakes

### Mistake 1

Forgetting to reset the streak.

Wrong idea

```text
1 1 0 1
```

Current streak should become

```text
2 → 0 → 1
```

not

```text
2 → 3
```

---

### Mistake 2

Updating the maximum only when encountering `0`.

Always update the maximum after processing each element.

Otherwise,

an array ending with `1`s may produce an incorrect answer.

Example

```text
[1,1,1]
```

---

### Mistake 3

Using unnecessary nested loops.

Checking every possible streak separately results in

```text
O(n²)
```

A single traversal is sufficient.

````python
class Solution(object):
    def findMaxConsecutiveOnes(self, nums):
        """
        :type nums: List[int]
        :rtype: int

        """
        count=0
        value=0
        for i in nums:
            if i==1:
                count+=1
            else:
                count=0
            value=max(value,count)
        return value
            
        
        
````
