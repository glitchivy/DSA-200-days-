# LeetCode 189 - Rotate Array

## Problem

Given an integer array `nums`, rotate the array to the **right** by `k` steps.

The rotation must be performed **in-place**.

---

## Pattern

**Array Reversal + Two Pointers**

---

## Intuition

Instead of rotating the array one step at a time, think of the array as **two blocks**.

Example:

```text
nums = [1,2,3,4,5,6,7]
k = 3
```

Split the array into:

```text
A = [1,2,3,4]
B = [5,6,7]
```

Goal:

```text
A | B
```

↓

```text
B | A
```

The Reverse Technique achieves this in-place.

---

## Key Observation

Rotating one step repeatedly is inefficient.

Instead,

1. Reverse the whole array.
2. Reverse the first `k` elements.
3. Reverse the remaining elements.

---

## Why Do We Use `k % n`?

Suppose

```text
n = 7
k = 10
```

Rotating seven times returns the array to its original position.

Therefore,

```text
10 rotations
```

is the same as

```text
10 % 7 = 3 rotations
```

Always compute

```text
k = k % n
```

before performing the rotation.

---

## Reverse Technique

### Original

```text
1 2 3 4 | 5 6 7
```

---

### Step 1

Reverse the entire array.

```text
7 6 5 | 4 3 2 1
```

Notice:

The last block has moved to the front, but both blocks are reversed.

---

### Step 2

Reverse the first `k` elements.

```text
7 6 5
```

↓

```text
5 6 7
```

Array becomes

```text
5 6 7 | 4 3 2 1
```

The first block is now correct.

---

### Step 3

Reverse the remaining elements.

```text
4 3 2 1
```

↓

```text
1 2 3 4
```

Final array

```text
5 6 7 | 1 2 3 4
```

Done.

---

## Why Does This Work?

Think of the array as

```text
A | B
```

Goal

```text
B | A
```

Reverse everything

```text
Reverse(B) | Reverse(A)
```

Reverse the first block

```text
B | Reverse(A)
```

Reverse the second block

```text
B | A
```

Exactly the desired result.

---

## Reverse Helper

To reverse any part of the array:

- One pointer starts from the left.
- One pointer starts from the right.
- Swap both elements.
- Move inward.
- Continue until the pointers meet.

This is simply the **Opposite-End Two Pointer** pattern.

## Time Complexity

### O(n)

Reason:

- Reverse entire array → O(n)
- Reverse first `k` elements → O(k)
- Reverse remaining elements → O(n-k)

Overall

```text
O(n)
```

---

## Space Complexity

### O(1)

Reason:

Everything is done in-place using swaps.

---

## Common Mistakes

### Mistake 1

Rotating one step at a time.

Example

```text
Move last element to front
Repeat k times
```

Time Complexity

```text
O(n × k)
```

Too slow.

---

### Mistake 2

Forgetting

```text
k = k % n
```

Example

```text
n = 5
k = 12
```

Without modulo,

unnecessary work is performed.

---

### Mistake 3

Stopping the reverse after `k` swaps.

A reverse operation always continues until:

```text
left >= right
```

It has nothing to do with the value of `k`.

---

### Mistake 4

Thinking the three reversals are independent tricks.

Remember:

The reversals work together to transform

```text
A | B
```

into

```text
B | A
```

  **Array Reversal + Two Pointers**.

````python
class Solution(object):
    def rotate(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        
        n=len(nums)
        k=k%n

        def rev(left,right):
            while left<right:
                nums[left],nums[right]=nums[right],nums[left]
                left+=1
                right-=1
            
        rev(0,n-1)
        rev(0,k-1)
        rev(k,n-1)


        
````
