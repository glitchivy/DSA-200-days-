```python
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        if not nums:
            return 0
        write=0

        for i in range(1,len(nums)):
            if nums[i]!=nums[write]:
                write+=1
                nums[write]=nums[i]
        return write+1
```
