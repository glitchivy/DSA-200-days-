# LeetCode 283 - Move Zeroes

## Problem

Move all `0`s to the end of the array while maintaining the **relative order of the non-zero elements**.

The operation must be performed **in-place**.

---

## Pattern

**Two Pointers (Read/Write)**

---

## Intuition

Instead of moving every zero to the end,

move every **non-zero** element to the front.

The remaining positions automatically become zeros due to swapping.

---

## Key Observation

Think of the array as two regions.

```text
[Processed Non-Zero Elements][Unprocessed Elements]
```

The **write pointer** always points to the position where the next non-zero element should be placed.

The **read pointer** scans every element.

---

## Pointer Roles

### Read Pointer (`i`)

- Visits every element.
- Checks whether the current element is non-zero.

---

### Write Pointer (`w`)

- Points to where the next non-zero element should be placed.
- Moves only after writing a non-zero element.

---

## Algorithm

1. Initialize `w = 0`.
2. Traverse the array using `i`.
3. If `nums[i]` is non-zero:
   - Swap `nums[w]` and `nums[i]`.
   - Increment `w`.
4. Continue until the traversal is complete.

## Why Does This Work?

The write pointer always marks:

> **"Where should the next non-zero element go?"**

Every non-zero element is moved exactly once.

Zeros naturally shift toward the end because of the swaps.

---

## Time Complexity

### O(n)

Reason:

The array is traversed only once.

---

## Space Complexity

### O(1)

Reason:

No extra array is used.

Everything is done in-place.

---

## Common Mistakes

### Mistake 1

Starting the loop from index `1`.

Wrong

```python
for i in range(1, len(nums)):
```

This skips the first element.

Example

```text
[2,1]
```

The first non-zero (`2`) is never processed, leading to an incorrect result.

Correct

```python
for i in range(len(nums)):
```

Always process **every element**.

---

### Mistake 2

Making the write pointer represent multiple things.

For example:

- First zero position
- Current zero
- Next write position

Keep one invariant.

The write pointer should always mean:

> **"Index where the next non-zero element should be written."**

---

### Mistake 3

Thinking you need to count zeros.

You don't.

Simply process every non-zero element.

---

### Mistake 4

Creating another array.

The problem specifically asks for an **in-place** solution.

---

## Comparison with LC 26

Both problems use the same pattern.

### LC 26

Write the next **unique** element.

Condition

```text
Current element is different from the last unique element.
```

---

### LC 283

Write the next **non-zero** element.

Condition

```text
Current element is not zero.
```

The Read/Write Two Pointer pattern remains exactly the same.

Only the condition changes.

## Interview Takeaways

Whenever a problem asks you to:

- Move valid elements to the front.
- Remove unwanted elements.
- Modify an array in-place while preserving order.

Think:

> **Read Pointer + Write Pointer**

---

## Pattern Recognition

| Keyword | Pattern |
|----------|---------|
| Move Elements | Read/Write Two Pointers |
| In-place | Two Pointers |
| Preserve Order | Read/Write Pointer |
| Remove/Filter Elements | Read/Write Pointer |

```**

````python
class Solution(object):
    def moveZeroes(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        w=0
        for i in range(len(nums)):
            if  nums[i]!=0:
                nums[w],nums[i]=nums[i],nums[w]
                w+=1
        return nums
````
