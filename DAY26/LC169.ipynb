# LeetCode 169 - Majority Element

## Problem

Given an integer array `nums` of size `n`, return the element that appears **more than ⌊n/2⌋ times**.

It is guaranteed that a majority element always exists.

Example

```text
Input

[2,2,1,1,1,2,2]
```

Output

```text
2
```

---

# Pattern

**Boyer-Moore Voting Algorithm**

Category

```text
Arrays
      ↓
Greedy
      ↓
Voting Algorithm
      ↓
Boyer-Moore Voting
```

---

# Core Idea

Imagine every element is casting a vote.

- If the current element is the same as the current candidate, it gains one vote.
- If it is different, one vote is cancelled.

Whenever all votes are cancelled, we choose a new candidate.

Since the majority element appears **more than half of the array**, it can never be completely cancelled by the remaining elements.

Therefore, after all cancellations, the majority element is the only candidate left.

---

# Intuition

Think of it like an election.

```text
Candidate

↓

Gets votes from identical numbers

↓

Loses votes when different numbers appear

↓

If all votes disappear

↓

Choose a new candidate

↓

Continue until the array ends
```

The majority element always survives because it has more votes than every other element combined.

---

# Variables

### Candidate

Represents the current element believed to be the majority.

Whenever all previous votes are cancelled, a new candidate is selected.

---

### Count

Represents the current vote balance.

- Same element → increase votes.
- Different element → decrease votes.

---

# Algorithm

Traverse the array once.

For every element:

### Step 1

If there is no active candidate (vote count becomes zero),

select the current element as the new candidate.

---

### Step 2

Compare the current element with the candidate.

If both are the same,

increase the vote count.

Otherwise,

decrease the vote count.

---

### Step 3

Continue until the end of the array.

The remaining candidate is the majority element.

# Why Does It Work?

Every non-majority element cancels one occurrence of the majority element.

Since the majority element appears **more than n/2 times**, there are not enough other elements to cancel all of its occurrences.

Eventually,

only the majority element survives as the final candidate.

# Complexity

### Time

```text
O(n)
```

Only one traversal of the array.

---

### Space

```text
O(1)
```

Only two variables are maintained.

---

# Common Mistakes

## Mistake 1

Changing the candidate whenever a different number appears.

Correct approach:

Change the candidate **only when the vote count becomes zero**.

---

## Mistake 2

Thinking the candidate always has the highest frequency during traversal.

The candidate may change multiple times.

Only the final candidate is guaranteed to be the majority element.

---

## Mistake 3

Using extra space like a HashMap.

Boyer-Moore solves the problem in constant space.

---

## Mistake 4

Confusing this with frequency counting.

This algorithm never counts the actual frequency of every element.

It only maintains a **relative vote balance**.

````python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count=0
        for i in nums:
            if count==0:
                can=i
            if can==i:
                count+=1
            else:
                count-=1
        return can
````
