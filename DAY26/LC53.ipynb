# LeetCode 53 - Maximum Subarray

## Problem

Given an integer array `nums`, find the **contiguous subarray** with the largest sum and return that sum.

Example

```text
Input

[-2,1,-3,4,-1,2,1,-5,4]
```

Output

```text
6
```

Explanation

```text
Subarray = [4,-1,2,1]

Sum = 6
```

---

# Pattern

**Kadane's Algorithm**

Category:

```text
Dynamic Programming
        ↓
1D Dynamic Programming
        ↓
State Compression
        ↓
Kadane's Algorithm
```

---

# Core Idea

At every element, we have only **two choices**.

### Choice 1

Continue the current subarray.

```text
Current Sum + nums[i]
```

### Choice 2

Start a brand new subarray from the current element.

```text
nums[i]
```

We always choose the better option.

---

# Dynamic Programming Recurrence

Let

```text
dp[i]
```

represent

> Maximum subarray sum ending at index `i`.

Then

```text
dp[i] = max(
            nums[i],
            dp[i-1] + nums[i]
         )
```

Meaning

Either

- Start a new subarray

or

- Extend the previous one.

---

# Why Kadane's Algorithm Works

Instead of storing the whole DP array,

we only need

```text
dp[i-1]
```

because

```text
dp[i]
```

depends only on the previous state.

So,

```text
dp[i-1]
```

becomes

```python
current_sum
```

and

```text
Maximum of all dp values
```

becomes

```python
max_sum
```

This reduces

```text
Space

O(n)

↓

O(1)
```

---

# Variables

```python
current_sum
```

Current sum of the subarray being considered.

---

```python
max_sum
```

Largest subarray sum found so far.

---

# Algorithm

Initialize

```python
current_sum = 0
max_sum = nums[0]
```

Traverse the array.

For every number

```python
current_sum += num
```

Update answer

```python
max_sum = max(max_sum, current_sum)
```

If

```python
current_sum < 0
```

Reset

```python
current_sum = 0
```

because a negative sum will only reduce future sums.

---

# Dry Run

Example

```text
nums = [-2,1,-3,4,-1,2,1,-5,4]
```

| Element | Current Sum | Max Sum |
|---------|------------:|---------:|
| -2 | -2 → reset to 0 | -2 |
| 1 | 1 | 1 |
| -3 | -2 → reset | 1 |
| 4 | 4 | 4 |
| -1 | 3 | 4 |
| 2 | 5 | 5 |
| 1 | 6 | 6 |
| -5 | 1 | 6 |
| 4 | 5 | 6 |

Answer

```text
6
```

---

# Why Reset Current Sum?

Suppose

```text
Current Sum = -5
```

Next element

```text
7
```

Continuing gives

```text
-5 + 7 = 2
```

Starting fresh gives

```text
7
```

Obviously

```text
7 > 2
```

So carrying a negative sum is never beneficial.

Therefore

```python
if current_sum < 0:
    current_sum = 0
```

---

# Why

```python
max_sum = nums[0]
```

instead of

```python
max_sum = 0
```

Consider

```text
[-3,-2,-5]
```

Correct answer

```text
-2
```

If

```python
max_sum = 0
```

the algorithm returns

```text
0
```

which is incorrect because no subarray sums to zero.

Initializing with

```python
nums[0]
```

correctly handles arrays containing all negative numbers.

---

# Code

```python
class Solution(object):
    def maxSubArray(self, nums):

        current_sum = 0
        max_sum = nums[0]

        for num in nums:

            current_sum += num

            max_sum = max(max_sum, current_sum)

            if current_sum < 0:
                current_sum = 0

        return max_sum
```

# Common Mistakes

## Mistake 1

Initializing

```python
max_sum = 0
```

Correct

```python
max_sum = nums[0]
```

---

## Mistake 2

Resetting before updating the answer.

Wrong

```python
if current_sum < 0:
    current_sum = 0

max_sum = max(max_sum, current_sum)
```

Correct

```python
max_sum = max(max_sum, current_sum)

if current_sum < 0:
    current_sum = 0
```

Otherwise, negative-only arrays fail.

---

## Mistake 3

Using an extra DP array.

Kadane's Algorithm is simply the **space-optimized version** of the DP solution.
