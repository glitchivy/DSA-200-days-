# LeetCode 2149 - Rearrange Array Elements by Sign

## Problem

You are given an array `nums` containing an equal number of positive and negative integers.

Rearrange the array such that:

- Positive and negative integers alternate.
- The first element is positive.
- The relative order of positive integers remains the same.
- The relative order of negative integers remains the same.

Example

```text
Input

[3,1,-2,-5,2,-4]
```

Output

```text
[3,-2,1,-5,2,-4]
```

---

# Pattern

**Simulation + Index Placement**

Category

```text
Arrays
      ↓
Simulation
      ↓
Index Placement
```

---

# Core Idea

The final positions are already fixed.

- Positive numbers always go to **even indices**.
- Negative numbers always go to **odd indices**.

Instead of swapping elements, directly place each element into its correct position.

---

# Intuition

Maintain two indices.

```text
Positive Index = 0

Negative Index = 1
```

Traverse the original array once.

- If the current number is positive, place it at the current positive index.
- If the current number is negative, place it at the current negative index.

After placing,

move the corresponding index by `2`.

---

# Variables

### Positive Index

Represents the next available even position.

Initially

```text
0
```

Moves as

```text
0 → 2 → 4 → ...
```

---

### Negative Index

Represents the next available odd position.

Initially

```text
1
```

Moves as

```text
1 → 3 → 5 → ...
```

---

### Answer Array

Stores the rearranged result.

Initially

```text
[0] * n
```

---

# Algorithm

Traverse the array once.

For every number:

### Step 1

If the number is positive,

place it at the current positive index.

Move the positive index by `2`.

---

### Step 2

If the number is negative,

place it at the current negative index.

Move the negative index by `2`.

---

### Step 3

Return the answer array.

---

```text
[3,1,-2,-5,2,-4]
```

```text
ans = [_, _, _, _, _, _]

pos = 0
```

```text
[3,-2,1,-5,2,-4]
```

---

# Why This Works

The problem already tells us the exact destination of every positive and negative number.

Since we insert them in the order we encounter them,

their relative order is automatically preserved.

---

# Complexity

### Time

```text
O(n)
```

Single traversal of the array.

---

### Space

```text
O(n)
```

Extra answer array.

---

# Common Mistakes

## Mistake 1

Trying to solve it using swapping (Two Pointers).

Swapping changes the relative order of positives and negatives.

The problem explicitly asks to preserve the original order.

---

## Mistake 2

Using `append()`.

Wrong because we need elements at specific indices.

Example

```text
Positive indices

0,2,4,...
```

Appending places elements consecutively.

---

## Mistake 3

Treating `pos` and `neg` as indices of the original array.

Wrong.

They represent indices in the **answer array**, not `nums`.

Correct idea

```text
ans[pos] = current_positive

ans[neg] = current_negative
```

---

## Mistake 4

Creating an empty answer array.

Wrong

```text
ans = []
```

Correct

```text
ans = [0] * n
```

because indexed assignment requires allocated space.

- Preserve relative order
- No in-place requirement

````python
class Solution(object):
    def rearrangeArray(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        pos=0
        neg=1
        n=len(nums)
        ans=[0]*n
        for i in range(n):
            if nums[i]<0:
                ans[neg]=nums[i]
                neg+=2
            else:
                ans[pos]=nums[i]
                pos+=2
        return ans
````
