# twosum(#1)

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        n=len(nums)
        val={}
        for i in range(n):
            req=target-nums[i]
            if req in val:
                return [val[req],i]
            val[nums[i]]=i
       
        

```

# LC79 sort colours

```python
class Solution(object):
    def sortColors(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        low=0
        high=len(nums)-1
        mid=0
        while mid<=high:
            if nums[mid]==0:
                nums[low],nums[mid]=nums[mid],nums[low]
                mid+=1
                low+=1
            elif nums[mid]==2:
                nums[high],nums[mid]=nums[mid],nums[high]
                high-=1
            else:
                mid+=1
        return nums
```

# LC 41 first missing postive

```python
class Solution(object):
    def firstMissingPositive(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n=len(nums)
        for i in range(n):
            while (nums[i]>0 and nums[i]<=n and nums[i]!=nums[nums[i]-1]):
                home=nums[i]-1
                nums[i],nums[home]=nums[home],nums[i]
        for i in range(n):
            if nums[i]!=i+1:
                return i+1
        return n+1

```
