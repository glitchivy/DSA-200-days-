# LeetCode 121 - Best Time to Buy and Sell Stock

## Problem

You are given an array `prices` where `prices[i]` represents the stock price on the `iᵗʰ` day.

You may choose **one day to buy** and **one later day to sell**.

Return the **maximum profit** possible.

If no profit can be made, return `0`.

Example

```text
Input

[7,1,5,3,6,4]
```

Output

```text
5
```

Explanation

```text
Buy at 1

Sell at 6

Profit = 5
```

---

# Pattern

**Prefix Minimum + Running Maximum Profit**

Category

```text
Arrays
      ↓
Prefix Computation
      ↓
Running Minimum
      ↓
Greedy
```

---

# Core Idea

While traversing the array from left to right,

always remember the **lowest buying price** seen so far.

For every day,

pretend you sell on that day.

Calculate the profit.

Keep track of the maximum profit obtained.

---

# Intuition

At every price,

ask two questions.

### Question 1

```text
Is this the cheapest stock I've seen so far?
```

If yes,

update the minimum buying price.

---

### Question 2

```text
If I sell today,

how much profit will I make?
```

Compare this profit with the best profit found so far.

---

# Variables

### Minimum Buy Price

Stores the smallest stock price encountered while traversing.

Represents the best day to buy up to the current day.

---

### Maximum Profit

Stores the highest profit found so far.

Initially

```text
0
```

because making no transaction is always allowed.

---

# Algorithm

Traverse the array once.

For every price:

### Step 1

Update the minimum buying price seen so far.

---

### Step 2

Assume you sell today.

Calculate

```text
Today's Price − Minimum Buying Price
```

---

### Step 3

Update the maximum profit if the current profit is larger.

Continue until the end.

---

# Dry Run

Example

```text
prices = [7,1,5,3,6,4]
```

| Price | Minimum Buy | Profit Today | Maximum Profit |
|------:|------------:|-------------:|---------------:|
| 7 | 7 | 0 | 0 |
| 1 | 1 | 0 | 0 |
| 5 | 1 | 4 | 4 |
| 3 | 1 | 2 | 4 |
| 6 | 1 | 5 | 5 |
| 4 | 1 | 3 | 5 |

Answer

```text
5
```

# Why Not Just Find Minimum and Maximum?

Consider

```text
[7,1,5,3,6,4]
```

Minimum price

```text
1
```

Maximum price

```text
7
```

Profit would be

```text
7 - 1 = 6
```

But

```text
7
```

comes **before**

```text
1
```

You cannot sell before buying.

Therefore,

finding the overall minimum and maximum does **not** solve the problem.

---

# Why This Works

At every day,

you already know the cheapest buying price seen before.

Therefore,

you can instantly compute the best possible profit if you sell today.

Repeating this for every day guarantees finding the optimal answer.

---

# Complexity

### Time

```text
O(n)
```

Only one traversal of the array.

---

### Space

```text
O(1)
```

Only two variables are maintained.

---

# Common Mistakes

## Mistake 1

Finding the overall minimum and maximum.

Wrong because

the selling day must always come **after** the buying day.

---

## Mistake 2

Updating profit without remembering the minimum buying price.

The minimum buying price is required to calculate the best possible profit for each day.

---

## Mistake 3

Allowing negative profit.

If every price decreases,

the answer should be

```text
0
```

because making no transaction is allowed.

---

## Mistake 4

Thinking you need two nested loops.

Brute Force

```text
O(n²)
```

Optimal

```text
O(n)
```

using one traversal.

---

# Pattern Recognition

Whenever a problem contains

- Buy once
- Sell once
- Maximum Profit
- Prices over time
- Buy before Sell

Think

```text
Running Minimum

↓

Profit = Current Price − Minimum Price

↓

Update Maximum Profit
```

````python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        min_buy=prices[0]
        max_buy=0
        for i in prices:
            min_buy=min(min_buy,i)
            profit=i-min_buy
    
            max_buy=max(max_buy,profit)
        return max_buy
````
