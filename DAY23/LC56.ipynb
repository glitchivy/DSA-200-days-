# LeetCode 56 - Merge Intervals

## Problem

Given an array of intervals where

```text
[start, end]
```

merge all overlapping intervals and return the merged intervals.

Example

```text
Input

[[1,3],[2,6],[8,10],[15,18]]
```

Output

```text
[[1,6],[8,10],[15,18]]
```

Because

```text
[1,3]

and

[2,6]
```

overlap.

---

# Pattern

**Sorting + Greedy**

---

# Intuition

The intervals may not be in order.

So first,

sort them by their starting value.

Example

```text
Before

[[8,10],[1,3],[2,6],[15,18]]
```

↓

```text
After Sorting

[[1,3],[2,6],[8,10],[15,18]]
```

Now simply move from left to right.

For every interval,

either

- Merge

or

- Add a new interval.

---

# Step 1 : Sort

```python
intervals.sort()
```

or

```python
intervals.sort(key=lambda x: x[0])
```

Both sort according to the starting value.

---

# Step 2 : Initialize Result

The first interval is always valid.

```python
result = [intervals[0]]
```

Example

```text
result

↓

[[1,3]]
```

---

# Step 3 : Traverse Remaining Intervals

```python
for current in intervals[1:]:
```

Skip the first interval because it is already inside `result`.

---

# Step 4 : Get Last Merged Interval

```python
last = result[-1]
```

Example

```text
result

↓

[[1,6],[8,10]]
```

Then

```python
last
```

contains

```text
[8,10]
```

---

# Step 5 : Check Overlap

Condition

```python
if current[0] <= last[1]:
```

Meaning

```text
Current Start

<=

Last End
```

If true,

the intervals overlap.

---

# Step 6 : Merge

Example

```text
Last

[1,3]

Current

[2,6]
```

Merged Interval

```text
[1,6]
```

Only the end changes.

```python
last[1] = max(last[1], current[1])
```

Notice

We **do not append**.

We simply extend the previous interval.

---

# Step 7 : No Overlap

Example

```text
Last

[1,6]

Current

[8,10]
```

Since

```text
8 > 6
```

They do not overlap.

Append it.

```python
result.append(current)
```

---

# Complete Algorithm

```python
intervals.sort()

result = [intervals[0]]

for current in intervals[1:]:

    last = result[-1]

    if current[0] <= last[1]:
        last[1] = max(last[1], current[1])

    else:
        result.append(current)

return result
```

# Why Do We Compare With `result[-1]`?

Suppose

```text
[[1,3],[2,6],[5,8]]
```

After merging

```text
[1,3]

↓

[1,6]
```

The next interval

```text
[5,8]
```

must be compared with

```text
[1,6]
```

not

```text
[2,6]
```

Therefore

```python
last = result[-1]
```

is always the correct interval to compare.

# Common Mistakes

## Mistake 1

Forgetting to sort.

Without sorting,

adjacent intervals may not overlap even though others do.

---

## Mistake 2

Appending after merging.

Wrong

```python
last[1] = max(last[1], current[1])
result.append(current)
```

This creates duplicate intervals.

Correct

```python
last[1] = max(last[1], current[1])
```

Only update.

---

## Mistake 3

Comparing with the previous interval instead of the last merged interval.

Wrong

```python
intervals[i-1]
```

Correct

```python
result[-1]
```

---

## Mistake 4

Using

```python
min()
```

for the ending point.

Always use

```python
max()
```

because the merged interval should extend as far as possible.

# Personal Learning Notes

- Learned the **Sorting + Greedy** pattern.
- Always sort intervals before processing.
- Compare the current interval with the **last merged interval**.
- If intervals overlap, extend the previous interval.
- If they don't overlap, append a new interval.
- Learned that merging updates only the ending value.
- Added a new DSA pattern:

**Sorting + Greedy (Merge Intervals)**.

````python
class Solution(object):
    def merge(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: List[List[int]]
        """
        intervals.sort()
        result=[intervals[0]]
        for i in intervals[1:]:
            last=result[-1]
            if i[0]<=last[1]:
                last[1] = max(last[1], i[1])
            else:
                result.append(i)

        return result
````
