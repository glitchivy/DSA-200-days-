# LeetCode 48 - Rotate Image

## Problem

Given an `n × n` matrix, rotate the image **90° clockwise**.

### Constraints

- Rotate **in-place**
- Do **not** create another matrix.

Example

```text
Input

1 2 3
4 5 6
7 8 9
```

Output

```text
7 4 1
8 5 2
9 6 3
```

---

# Pattern

**Matrix Transformation**

Uses two techniques:

- Transpose
- Reverse Every Row

---

# Intuition

Instead of rotating every element individually,

break the problem into two simpler operations.

```
Original

↓

Transpose

↓

Reverse Every Row

↓

90° Clockwise Rotation
```

---

# Step 1 : Transpose

Swap rows and columns.

Example

```text
Original

1 2 3
4 5 6
7 8 9
```

becomes

```text
Transpose

1 4 7
2 5 8
3 6 9
```

---

## Observation

Element

```text
(0,1)
```

swaps with

```text
(1,0)
```

If we traverse the whole matrix,

we would swap them twice.

Example

```text
(0,1)

↓

(1,0)

↓

(0,1)
```

Back to original.

---

# Triangle Traversal

Visit only the **upper triangle**.

```text
X ✓ ✓
X X ✓
X X X
```

Where

```text
✓ = Swap

X = Ignore
```

Loop

```python
for i in range(n):
    for j in range(i + 1, n):
        matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
```

---

# Step 2 : Reverse Every Row

After transpose

```text
1 4 7
2 5 8
3 6 9
```

Reverse each row.

```text
7 4 1
8 5 2
9 6 3
```

---

## Two Pointer Technique

For every row

```python
left = 0
right = n - 1
```

Swap

```python
matrix[i][left], matrix[i][right] = matrix[i][right], matrix[i][left]
```

Move pointers

```python
left += 1
right -= 1
```

Continue until

```python
left < right
```

---

# Algorithm

## Step 1

Transpose

```python
for i in range(n):
    for j in range(i + 1, n):
        matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
```

---

## Step 2

Reverse every row

```python
for i in range(n):

    left = 0
    right = n - 1

    while left < right:

        matrix[i][left], matrix[i][right] = matrix[i][right], matrix[i][left]

        left += 1
        right -= 1
```

# Why Does This Work?

Transpose

```text
Rows

↓

Columns
```

Reverse Rows

```text

↓

```

Together,

these two operations are equivalent to

```text
90° Clockwise Rotation
```

# Common Mistakes

## Mistake 1

Traversing the whole matrix during transpose.

Wrong

```python
for i in range(n):
    for j in range(n):
```

This swaps every pair twice.

Correct

```python
for j in range(i + 1, n):
```

---

## Mistake 2

Reversing columns instead of rows.

Wrong

```python
matrix[left][i]
```

Correct

```python
matrix[i][left]
```

The row stays fixed.

---

## Mistake 3

Keeping the reverse loop inside the transpose loop.

Wrong

```python
for i:
    transpose

    reverse
```

Correct

```python
Transpose Entire Matrix

↓

Reverse Entire Matrix
```

Two separate phases.

---

## Mistake 4

Using

```python
while left <= right
```

Works,

but performs one unnecessary self-swap.

Better

```python
while left < right
```

- Understood that reversing every row after transpose gives a 90° clockwise rotation.
- Reused the **Two Pointer** technique to reverse each row.
- Learned to separate the solution into two phases:
  - Transpose
  - Reverse Rows
- Added a new DSA pattern:
  **Transpose + Reverse Rows (Matrix Rotation)**.

````python
class Solution(object):
    def rotate(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        n=len(matrix)
        for i in range(n):
            for j in range(i+1,n):
                matrix[i][j],matrix[j][i]=matrix[j][i],matrix[i][j]
        for i in range(n):
            left=0
            right=n-1
            while left<right:
                matrix[i][left],matrix[i][right]=matrix[i][right],matrix[i][left]
                left+=1
                right-=1
````
