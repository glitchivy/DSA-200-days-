# LeetCode 54 - Spiral Matrix

## Problem

Given an `m × n` matrix, return all elements of the matrix in **spiral order**.

Example:

```text
1 2 3
4 5 6
7 8 9
```

Output

```text
[1,2,3,6,9,8,7,4,5]
```

---

# Pattern

**Boundary Traversal**

---

# Intuition

Imagine the matrix has four boundaries.

```text
top
bottom
left
right
```

Instead of marking cells as visited,

we continuously shrink these boundaries.

Every iteration consists of:

1. Traverse Top Row →
2. Traverse Right Column ↓
3. Traverse Bottom Row ←
4. Traverse Left Column ↑

After one full round,

move the boundaries inward.

---

# Initial Boundaries

```python
top = 0
bottom = rows - 1
left = 0
right = cols - 1
```

---

# Traversal Order

```text
→ → → →

      ↓

← ← ← ←

↑
```

Repeat until boundaries cross.

---

# Step 1 : Traverse Top Row

Move

```text
Left → Right
```

```python
for j in range(left, right + 1):
    ans.append(matrix[top][j])

top += 1
```

---

# Step 2 : Traverse Right Column

Move

```text
Top ↓ Bottom
```

```python
for i in range(top, bottom + 1):
    ans.append(matrix[i][right])

right -= 1
```

---

# Step 3 : Traverse Bottom Row

Only if a row still exists.

```python
if top <= bottom:
```

Move

```text
Right ← Left
```

```python
for j in range(right, left - 1, -1):
    ans.append(matrix[bottom][j])

bottom -= 1
```

---

# Step 4 : Traverse Left Column

Only if a column still exists.

```python
if left <= right:
```

Move

```text
Bottom ↑ Top
```

```python
for i in range(bottom, top - 1, -1):
    ans.append(matrix[i][left])

left += 1
```

---

# Stopping Condition

Continue while a valid rectangle still exists.

```python
while top <= bottom and left <= right:
```

Stop when

```text
top > bottom

or

left > right
```

# Why Do We Need Boundary Checks?

## Bottom Row

Suppose

```text
1 2 3
```

After traversing the top row

```python
top += 1
```

Now

```text
top = 1
bottom = 0
```

There is no bottom row.

Without

```python
if top <= bottom:
```

the bottom row would be processed again.

---

## Left Column

Suppose

```text
1
2
3
```

After traversing the right column

```python
right -= 1
```

Now

```text
left = 0
right = -1
```

No left column exists.

Without

```python
if left <= right:
```

elements would be duplicated.

---

# Algorithm

```text
Initialize

↓

While boundaries are valid

↓

Traverse Top

↓

Move Top

↓

Traverse Right

↓

Move Right

↓

If row exists

    Traverse Bottom

    Move Bottom

↓

If column exists

    Traverse Left

    Move Left

↓

Repeat
```

```

O(1)
```

(Output array not counted.)

# Common Mistakes

## Mistake 1

Forgetting

```python
top += 1
```

Results in traversing the same row repeatedly.

---

## Mistake 2

Forgetting

```python
left += 1
```

Results in an infinite loop or duplicate elements.

---

## Mistake 3

Writing

```python
range(right, left)
```

instead of

```python
range(right, left - 1, -1)
```

Need a negative step.

---

## Mistake 4

Not checking

```python
if top <= bottom
```

Duplicates the bottom row.

---

## Mistake 5

Not checking

```python
if left <= right
```

Duplicates the left column.

- Learned to shrink boundaries after each traversal.

- Learned that every matrix element is visited exactly once.
- Added a new DSA pattern:
  **Boundary Traversal / Spiral Matrix**.

````python
class Solution(object):
    def spiralOrder(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: List[int]
        """
        row=len(matrix)
        col=len(matrix[0])
        top=0
        left=0
        right=col-1
        bottom=row-1
        ans=[]
        while top<=bottom and left<=right:
            for j in range(left,right+1):
                ans.append(matrix[top][j])
            top+=1
            for i in range(top,bottom+1):
                ans.append(matrix[i][right])
            right-=1
            if top <= bottom:
                for j in range(right,left-1,-1):
                    ans.append(matrix[bottom][j])
                bottom-=1
            if left <= right:
                for i in range(bottom,top-1,-1):
                    ans.append(matrix[i][left])
                left+=1

        return ans
````
