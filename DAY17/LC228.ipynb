# LeetCode 228 - Summary Ranges

## Problem

Given a **sorted unique integer array** `nums`, return the smallest list of ranges that covers every number exactly once.

If a range contains:

- Only one number → `"7"`
- Multiple consecutive numbers → `"4->6"`

---

## Pattern

**Range Traversal (One Pass)**

---

## Intuition

Traverse the array from left to right.

A range continues as long as the numbers are consecutive.

The moment the sequence breaks, the current range ends.

Store the completed range and begin a new one.

---

## Key Observation

A range continues if:

```text
nums[i] + 1 == nums[i + 1]
```

A range ends if:

```text
nums[i] + 1 != nums[i + 1]
```

---

## Algorithm

1. Handle the empty array.
2. Store the beginning of the current range using `start`.
3. Traverse the array.
4. Whenever the sequence breaks:
   - If the range has only one element, add that number.
   - Otherwise, add `"start->end"`.
   - Update `start`.
5. After the loop, process the final remaining range.

---

## Why Do We Need `start`?

`start` stores the beginning of the current range.

Example:

```text
0 1 2 4 5 7
```

Initially

```text
start = 0
```

When the sequence breaks after `2`,

we know the range is

```text
0 -> 2
```

Now begin a new range.

```text
start = 4
```

---

## Why Do We Process the Last Range Separately?

The algorithm only detects a range when the sequence **breaks**.

Example:

```text
0 1 2 4 5 7
```

The loop detects

```text
0->2
```

Then

```text
4->5
```

But

```text
7
```

never causes a break because the loop ends.

Therefore, after the loop, one unfinished range is always left.

Process it separately.

## Common Mistakes

### Mistake 1

Forgetting the empty array.

Example

```text
[]
```

Always return

```text
[]
```

before accessing any element.

---

### Mistake 2

Forgetting the last range.

The loop only detects ranges when they break.

The final range never breaks.

Always process it after the loop.

---

### Mistake 3

Returning integers instead of strings.

Wrong

```text
[7]
```

Correct

```text
["7"]
```

---

### Mistake 4

Returning a list instead of a formatted string.

Wrong

```text
[0,"->",2]
```

Correct

```text
"0->2"
```

---

### Mistake 5

Updating `start` only for multi-element ranges.

Whenever a range ends,

always update

```text
start = i + 1
```

regardless of whether the range contained one element or many.

## Interview Takeaways

Whenever you read a problem involving:

- Consecutive values
- Groups
- Intervals
- Runs

Ask yourself:

> **Can I mark where a group starts and detect where it ends?**

If yes,

a one-pass range traversal is often enough.

| Keyword | Pattern |
|----------|---------|

| Group consecutive values | One Pass |
| Build intervals | Range Detection |

````python
class Solution(object):
    def summaryRanges(self, nums):
        """
        :type nums: List[int]
        :rtype: List[str]
        """
        if not nums:
            return []
        start=0
        n=len(nums)
        result=[]
        for i in range(n-1):
            if nums[i]+1!=nums[i+1]:
                if nums[start]==nums[i]:
                    result.append(str(nums[i]))
                else:
                    result.append(str(nums[start]) + "->" + str(nums[i]))
                start=i+1
        if nums[start] == nums[-1]:
            result.append(str(nums[-1]))
        else:
            result.append(str(nums[start]) + "->" + str(nums[-1]))
        return result
            
            
````
