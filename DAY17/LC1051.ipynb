# LeetCode 1051 - Height Checker

## Problem

Given an array `heights`, return the number of indices where the students are **not standing in the order they would be if the array were sorted in non-decreasing order**.

---

## Pattern

**Sorting + Comparison**

---

## Intuition

The expected order is simply the **sorted version** of the array.

Instead of rearranging the original array, create a sorted copy and compare both arrays index by index.

Every mismatch represents a student standing in the wrong position.

---

## Key Observation

We don't need to physically rearrange the students.

We only need to know:

> **How many positions differ between the original array and the sorted array?**

---

## Algorithm

1. Create a sorted copy of the array.
2. Traverse both arrays together.
3. Compare elements at every index.
4. If they are different, increment the answer.
5. Return the mismatch count.

---

### Input

```text
heights = [1,1,4,2,1,3]
```

Sorted Copy

```text
[1,1,1,2,3,4]

```text

```

## Time Complexity

### O(n log n)

Reason:

- Sorting takes **O(n log n)**.
- Comparing both arrays takes **O(n)**.

Overall:

```text
O(n log n)
```

---

## Space Complexity

### O(n)

Reason:

A new sorted copy of the array is created.

---

## Common Mistakes

### Mistake 1

Using

```python
heights.sort()
```

instead of

```python
sorted(heights)
```

Problem:

`.sort()` modifies the original array.

You lose the original ordering needed for comparison.

---

### Mistake 2

Writing

```python
sorted = heights.sort()
```

Problem:

`.sort()` returns

```text
None
```

not the sorted array.

---

### Mistake 3

Comparing the sorted array with itself.

If the original array is modified before comparison,

every element matches,

giving an incorrect answer.

## Can This Be Optimized?

### General Case

No.

Sorting requires

```text
O(n log n)
```

which is optimal for comparison-based sorting.

---

### Special Case (LeetCode Constraint)

The problem guarantees:

```text
1 <= heights[i] <= 100
```

Because the values are limited,

**Counting Sort** can solve it in

```text
Time: O(n)

Space: O(1)
```

(auxiliary space, since the frequency array has fixed size 101).

- Recognized that limited value ranges may allow **Counting Sort** instead of comparison sorting.
```****

````python
class Solution(object):
    def heightChecker(self, heights):
        """
        :type heights: List[int]
        :rtype: int
        """
        sort= sorted(heights)
        n=len(heights)
        count=0
        for i in range(n):
            if heights[i]!=sort[i]:
                count+=1
        return count
````
