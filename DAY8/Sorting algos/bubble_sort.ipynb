# BUBBLE SORT

### bubble sort is an sorting algo where adjancent elements are being compared and swapped if they are in the wrong order 

### lets code in ascending order

````python
arr=list(map(int,input().split()))
n=len(arr)
for i in range(n-2,-1,-1):
    for j in range(i+1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
print(arr)
````

# Question 1 — Optimized Bubble Sort

---

# Problem Statement

Modify Bubble Sort so that:

```python
if array already becomes sorted,
stop early
```

instead of running unnecessary passes.

---

# Why Optimization?

Normal Bubble Sort always runs:

\[
O(N^2)
\]

even if array is already sorted.

Example:

```python
[1,2,3,4,5]
```

This array is already sorted,
but normal Bubble Sort still performs all passes.

This is unnecessary work.

---

# Goal

Detect:
```python
whether any swap happened
```

during a pass.

If NO swaps happen:
```python
array is already sorted
```

so stop algorithm early.

---

# Example

## Input

```python
[1,2,3,4,5]
```

---

# Pass 1

Compare:
- 1 and 2
- 2 and 3
- 3 and 4
- 4 and 5

No swaps happen.

This means:
```python
array already sorted
```

So:
```python
break loop
```

---

# Hint

Use a variable:

```python
swapped = False
```

Whenever swap happens:

```python
swapped = True
```

After every pass:

If:

```python
swapped == False
```

then:

```python
break
```

---

# Expected Learning

This problem teaches:

- Bubble Sort optimization
- unnecessary traversal detection
- best case optimization
- swap tracking
- early stopping conditions

---

# Important Observation

Optimized Bubble Sort:

## Best Case

\[
O(N)
\]

because only one pass needed.

---

## Worst Case

\[
O(N^2)
\]

when array completely unsorted.

---

# Dry Run

## Input

```python
[1,2,3,4,5]
```

---

# Initial

```python
swapped = False
```

---

# Compare Adjacent Elements

No swaps occur.

So:

```python
swapped remains False
```

Meaning:

```python
array already sorted
```

Break loop immediately.

---

# Final Output

```python
[1,2,3,4,5]
```

---

# Important Concept Learned

Optimization often means:

```python
detect unnecessary work
```

This idea appears later in:
- Sliding Window
- Greedy
- Dynamic Programming
- Graph algorithms
- Binary Search optimizations

````python
arr=list(map(int,input().split()))
n=len(arr)

for i in range(n-2,-1,-1):
    swapped=False
    for j in range(i+1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
            swapped=True 
    if swapped==False:
        break 
print(arr)
````

# Question 2 — Count Total Swaps in Bubble Sort

---

# Problem Statement

Modify Bubble Sort to count:

```python
total number of swaps
```

performed during sorting.

---

# Example

## Input

```python
[5,4,3]
```

---

# Output

```python
Sorted Array: [3,4,5]

Total Swaps: 3
```

---

# Explanation

## Initial Array

```python
[5,4,3]
```

---

# Pass 1

## Compare 5 and 4

```python
5 > 4
```

Swap.

Array becomes:

```python
[4,5,3]
```

Swap count:

```python
1
```

---

## Compare 5 and 3

Swap.

Array becomes:

```python
[4,3,5]
```

Swap count:

```python

```

---

# Pass 2

## Compare 4 and 3

Swap.

Array becomes:

```python
[3,4,5]
```

Swap count:

```python
3
```

---

# Final Output

```python
[3,4,5]
```

Total swaps:

```python
3
```

---

# Hint

Create a variable:

```python
swap_count = 0
```

Whenever swap happens:

```python
swap_count += 1
```

---

# Important Observation

More swaps usually indicate:
```python
array is more unsorted
```

---

# What This Problem Teaches

- swap tracking
- Bubble Sort internals
- adjacent correction logic
- operation counting
- inversion intuition

---

# Important Concept

Bubble Sort works by:
```python
repeatedly fixing local disorder
```

using adjacent swaps.

---

# Time Complexity

## Worst Case

\[
O(N^2)
\]

---

# Space Complexity

\[
O(1)
\]

---

# Bonus Thought

Which array causes:
```python
maximum swaps
```

in Bubble Sort?

Think about:
```python
completely reverse sorted array
```

Example:

```python
[5,4,3,2,1]
```

This creates maximum disorder.

````python
arr=list(map(int,input().split()))
count=0
n=len(arr)
for i in range(n-2,-1,-1):
    swap=False 
    for j in range(i+1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
            count+=1
            swap=True 
    if swap==False:
        break
print(count)
print(arr)
````
