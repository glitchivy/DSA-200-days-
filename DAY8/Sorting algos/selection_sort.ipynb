# SELECTION SORT 

### selection sort is an type of sorting algo in which we find the smallest element and place it to its correct position 

### lets code it in ascending order first 

````python
arr=list(map(int,input().split()))
n=len(arr)

for i in range(n):
    min_index=i
    for j in range(i+1,n):
        if arr[j]<arr[min_index]:
            min_index=j 
    arr[i],arr[min_index]=arr[min_index],arr[i]
print(arr)

````

### Lets code it in descending order now 

````python
arr=list(map(int,input().split()))
n=len(arr)
for i in range(n):
    max_index=i
    for j in range(i+1,n):
        if arr[j]>arr[max_index]:
            max_index=j
    arr[i],arr[max_index]=arr[max_index],arr[i]
print(arr)
````

# Question 1 — Find Minimum Element Using Selection Sort Logic

---

# Problem Statement

Given an array:

```python
arr = [5,2,8,1,9]
```

Find the:
```python
smallest element
```

WITHOUT using:
```python
min()
```

---

# Goal

Use:
```python
candidate tracking
```

similar to Selection Sort.

---

# Expected Output

```python
1
```

---

# Important Condition

Do NOT:
- sort the array
- use built-in functions like:
  - `min()`
  - `sorted()`

---

# Hint

Assume first element is minimum.

Example:

```python
minimum = arr[0]
```

Now traverse remaining array.

If smaller element found:

```python
arr[i] < minimum
```

update minimum.

---

# Dry Run

## Input

```python
[5,2,8,1,9]
```

---

# Step 1

Assume:

```python
minimum = 5
```

---

# Step 2

Compare with:
```python
2
```

Since:
```python
2 < 5
```

Update:

```python
minimum = 2
```

---

# Step 3

Compare with:
```python
8
```

No update.

---

# Step 4

Compare with:
```python
1
```

Update:

```python
minimum = 1
```

---

# Step 5

Compare with:
```python
9
```

No update.

---

# Final Answer

```python
1
```

---

# Concept Learned

This problem teaches:

- candidate tracking
- comparison logic
- minimum maintenance
- core intuition behind Selection Sort

---

# Time Complexity

\[
O(N)
\]

because array is traversed once.

---

# Space Complexity

\[
O(1)
\]

because no extra space is used.

### brute 

````python
arr=list(map(int,input().split()))
n=len(arr)
for i in range(n):
    min_index=i
    for j in range(i+1,n):
        if arr[j]<arr[min_index]:
            min_index=j
    arr[i],arr[min_index]=arr[min_index],arr[i]
print(arr[0])
````

### optimal

````python
arr=list(map(int,input().split()))
n=len(arr)
min_arr=arr[0]
for i in range(n):
    if arr[i]<min_arr:
        min_arr=i
print(min_arr)
````

# Question 2 — Find Second Smallest Element

---

# Problem Statement

Given an array:

```python
arr = [4,1,6,2,7]
```

Find the:
```python
second smallest element
```

WITHOUT:
- sorting the array
- using:
  - `sort()`
  - `sorted()`

---

# Expected Output

```python
2
```

---

# Important Observation

You do NOT need full sorting.

You only need:
- smallest element
- second smallest element

---

# Goal

Practice:
```python
candidate tracking
```

This is the same thinking used in:
- Selection Sort
- Greedy algorithms
- Optimization problems

---

# Hint

Maintain TWO variables:

```python
smallest
second_smallest
```

---

# Logic

## Step 1

Assume:

```python
smallest = infinity
second_smallest = infinity
```

---

## Step 2

Traverse array.

---

# Case 1

If current element is smaller than:
```python
smallest
```

Then:

```python
second_smallest = smallest
smallest = current element
```

---

# Case 2

Else if current element is:
- greater than smallest
- smaller than second_smallest

Update:

```python
second_smallest = current element
```

---

# Dry Run

## Input

```python
[4,1,6,2,7]
```

---

# Initial

```python
smallest = inf
second_smallest = inf
```

---

# Element = 4

```python
smallest = 4
second_smallest = inf
```

---

# Element = 1

```python
second_smallest = 4
smallest = 1
```

---

# Element = 6

```python
6 > 1
```

No update to smallest.

```python
second_smallest = 4
```

still smaller than 6.

---

# Element = 2

```python
2 > 1
2 < 4
```

Update:

```python
second_smallest = 2
```

---

# Final Answer

```python
2
```

---

# Important Edge Case

Input:

```python
[1]
```

No second smallest exists.

---

# Another Edge Case

Input:

```python
[1,1,1]
```

No distinct second smallest exists.

---

# Concepts Learned

This problem teaches:

- candidate tracking
- maintaining multiple best values
- conditional updates
- optimization without sorting

---

# Time Complexity

\[
O(N)
\]

because array is traversed once.

---

# Space Complexity

\[
O(1)
\]

because no extra space is used.

````python
arr=list(map(int,input().split()))
first_small=float("inf")
second_small=float("inf")
for i in arr:
    if i<first_small:
            second_small = first_small
            first_small=i
            
    elif i<second_small and second_small!=first_small:
          second_small=i
print(second_small)
    
````
