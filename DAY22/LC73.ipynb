# LeetCode 73 - Set Matrix Zeroes

## Problem

Given an `m × n` matrix.

If any cell contains `0`, set its **entire row** and **entire column** to `0`.

The modification must be done **in-place**.

---

# Brute Force

Traverse the matrix.

Whenever a `0` is found:

- Immediately make its row `0`.
- Immediately make its column `0`.

### Why it fails

New zeroes created during traversal are treated as original zeroes.

Example

```
1 1 1
1 0 1
1 1 1
```

Immediately modifying gives

```
1 0 1
0 0 0
1 0 1
```

Now while traversing,

```
(0,1)
```

is also `0`.

If we process it again,

its entire row becomes zero.

Eventually,

```
0 0 0
0 0 0
0 0 0
```

Incorrect.

---

# Key Observation

Do **not** modify the matrix while searching.

Instead,

split the problem into **two passes**.

---

# Pass 1

Only collect information.

Whenever a zero is found,

store:

- Row number
- Column number

Example

```
1 0 3
4 5 6
0 8 9
```

Zeroes found at

```
(0,1)
(2,0)
```

Store

```
Rows

{0,2}

Columns

{0,1}
```

Do **not** modify the matrix.

---

# Pass 2

Traverse the matrix again.

For every cell ask:

```
Is my row marked?

OR

Is my column marked?
```

If yes,

make the cell zero.

---

# Why Use a Set?

We only care whether a row or column should become zero.

Example

```
Row 2 has three zeroes.
```

We still only need to know

```
Row 2 → Yes
```

Not

```
How many zeroes?
```

A set automatically removes duplicates.

---

# Algorithm

### Step 1

Create two sets.

```
rows = set()
cols = set()
```

---

### Step 2

Traverse the matrix.

If

```
matrix[i][j] == 0
```

Store

```
rows.add(i)

cols.add(j)
```

---

### Step 3

Traverse the matrix again.

If

```
i in rows

OR

j in cols
```

then

```
matrix[i][j] = 0
```

---

# Example

Input

```
1 1 1
1 0 1
1 1 1
```

---

## Pass 1

Found

```
(1,1)
```

Store

```
Rows

{1}

Columns

{1}
```

Matrix remains unchanged.

---

## Pass 2

Every cell checks

```
Is my row marked?

OR

Is my column marked?
```

Result

```
1 0 1
0 0 0
1 0 1
```

---

# Time Complexity

First traversal

```
O(m × n)
```

Second traversal

```
O(m × n)
```

Overall

```
O(m × n)
```

---

# Space Complexity

Two sets store

- Rows
- Columns

Worst case

```
O(m + n)
```

---

# Pattern Learned

**Matrix Traversal + Two Pass Technique**

Whenever modifying data during traversal may affect future decisions,

think:

> **Collect information first, modify later.**

---

# Common Mistakes

### 1. Modifying immediately

Wrong because newly created zeroes are treated as original zeroes.

---

### 2. Using a List

Lists allow duplicate row and column indices.

A set is better because we only need to know whether a row or column should be zeroed.

---

### 3. Forgetting the Second Pass

The first traversal only identifies rows and columns.

Actual modification happens in the second traversal.

---

# Interview Follow-up

The interviewer may ask:

> Can you solve it using **O(1)** extra space?

The optimization is to use:

- First row as the row marker.
- First column as the column marker.

This eliminates the need for the two sets while keeping the same time complexity.

---

# Interview Takeaway

Whenever you see a matrix problem where changing values can affect future decisions, ask yourself:

> **Should I separate information gathering from modification?**

If yes, a **two-pass algorithm** is often the cleanest solution.

````python
class Solution(object):
    def setZeroes(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """
        rows=set()
        col=set()
        m=len(matrix)
        n=len(matrix[0])
        for i in range(m):
            for j in range(n):
                if matrix[i][j]==0:
                    rows.add(i)
                    col.add(j)

        for i in range(m):
            for j in range(n):
                if i in rows or j in col:
                    matrix[i][j]=0

````
