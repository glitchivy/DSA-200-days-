```python
class Solution(object):
    def maxSubArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        c_s=0
        max_s=nums[0]
        n=len(nums)
        for i in range(n):
            c_s+=nums[i]
            max_s=max(c_s,max_s)
            if c_s<0:
                c_s=0
        return max_s
```

```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        #bayes mooyer  method
        count=0
        for i in nums:
            if count<=0:
                can=i
                count=0
            if can==i:
                count+=1
            else:
                count-=1
        return can
            
```

```python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        min_buy=prices[0]
        profit=0
        max_buy=0
        for i in range(len(prices)):
            min_buy=min(min_buy,prices[i])
            profit=prices[i]-min_buy
            max_buy=max(max_buy,profit)
        return max_buy
     
```

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        noticed={}
        n=len(nums)
        for i in range(n):
            req=target-nums[i]
            if req in noticed:
                return [noticed[req],i]
            noticed[nums[i]]=i

```
