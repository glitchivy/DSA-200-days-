# LeetCode 31 - Next Permutation

## Problem

Given an array of integers, rearrange it into the **next lexicographically greater permutation**.

If such a permutation does not exist, rearrange it into the **lowest possible order** (ascending).

The modification must be done **in-place**.

---

# Brute Force

Generate all possible permutations.

Sort them lexicographically.

Find the current permutation.

Return the next permutation.

### Time Complexity

```
O(n! × n)
```

### Space Complexity

```
O(n!)
```

Not practical.

---

# Key Observation

A permutation increases only when we modify the **rightmost possible position**.

Notice that every permutation has a suffix that is already in **decreasing order**.

Example

```
1 2 5 4 3
```

Here,

```
5 4 3
```

is decreasing.

The first element before this decreasing suffix is where we can make the smallest possible increase.

This element is called the **pivot**.

---

# Optimal Approach

The algorithm has four steps.

---

## Step 1 — Find the Pivot

Traverse from right to left.

Find the first index where

```
nums[i] < nums[i + 1]
```

This index is called the **pivot**.

Example

```
1 2 5 4 3
```

Pivot is

```
2
```

because

```
2 < 5
```

---

## Step 2 — Handle the Largest Permutation

If no pivot exists,

the entire array is in decreasing order.

Example

```
5 4 3 2 1
```

This is already the largest permutation.

The next permutation is simply

```
1 2 3 4 5
```

So reverse the whole array.

---

## Step 3 — Find the Next Greater Element

Search from the end of the array.

Find the **first element greater than the pivot**.

Since the suffix is decreasing,

the first greater element from the right is automatically the **smallest possible greater value**.

Swap it with the pivot.

Example

Before swap

```
1 2 5 4 3
```

Swap

```
2
```

with

```
3
```

After swap

```
1 3 5 4 2
```

---

## Step 4 — Reverse the Suffix

After swapping,

everything after the pivot is still in decreasing order.

Reverse it.

Example

Before

```
1 3 5 4 2
```

After reversing the suffix

```
1 3 2 4 5
```

This becomes the **next smallest permutation**.

---

# Dry Run

Input

```
1 2 5 4 3
```

### Find Pivot

```
5 > 4

4 > 3

2 < 5
```

Pivot

```
2
```

---

### Find Next Greater Element

Search from the right.

```
3 > 2
```

Swap.

```
1 3 5 4 2
```

---

### Reverse Suffix

Reverse

```
5 4 2
```

Result

```
2 4 5
```

Final answer

```
1 3 2 4 5
```

---

# Why Search From the Right?

The suffix is already in decreasing order.

Searching from the right guarantees finding the **smallest element greater than the pivot**, producing the immediate next permutation.

---

# Why Reverse the Suffix?

The suffix is decreasing.

Reversing it makes it increasing.

An increasing suffix is the **smallest possible arrangement**, ensuring we get the immediate next permutation instead of skipping several permutations.

---

# Common Mistakes

### 1. Forgetting to Initialize the Pivot

Always initialize

```
pivot = -1
```

Otherwise, arrays like

```
3 2 1
```

cause an error.

---

### 2. Forgetting to Break

After finding

- the pivot
- the swap element

stop searching immediately.

---

### 3. Searching From the Left

Always search from the **right**.

The rightmost valid position gives the next lexicographical permutation.

---

### 4. Forgetting to Reverse the Suffix

Swapping alone is not enough.

The suffix must be reversed to obtain the smallest possible permutation.

---

### 5. Reversing the Entire Array

Only reverse the suffix after the pivot.

Reverse the entire array **only when no pivot exists**.

---

# Time Complexity

```
O(n)
```

- One pass to find the pivot.
- One pass to find the swap element.
- One pass to reverse the suffix.

Overall

```
O(n)
```

---

# Space Complexity

```
O(1)
```

The array is modified in-place.

---

# Pattern Learned

**Lexicographical Next Permutation**

---

# Pattern Recognition

Whenever a problem asks for

- next permutation
- next lexicographical arrangement
- immediate greater arrangement

think about

- finding the pivot,
- swapping with the next greater element,
- reversing the suffix.

---

# Mental Model

Think of the array as having two parts.

```
| Fixed Prefix | Decreasing Suffix |
```

The decreasing suffix is already the **largest arrangement**.

To get the next permutation:

1. Increase the prefix by the smallest amount possible.
2. Reset the suffix to its smallest possible order.

This produces the immediate next lexicographical permutation.

````python
class Solution(object):
    def nextPermutation(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        pivot=-1
        n=len(nums)
        for i in range(n-2,-1,-1):
            if nums[i]<nums[i+1]:
                pivot=i
                break
        if pivot==-1:
            nums.reverse()
            return 
        for i in range(n-1,pivot,-1):
            if nums[i]>nums[pivot]:
                nums[i],nums[pivot]=nums[pivot],nums[i]
                break 
        nums[pivot + 1:] = nums[pivot + 1:][::-1]
        return
````
