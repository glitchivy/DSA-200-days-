# LeetCode 128 - Longest Consecutive Sequence

## Problem

Given an unsorted array of integers, return the length of the **longest consecutive sequence**.

The algorithm must run in **O(n)** time.

### Example

```
Input:
[100,4,200,1,3,2]

Output:
4

Explanation:
The longest consecutive sequence is

1 в†’ 2 в†’ 3 в†’ 4

Length = 4
```

---

# Brute Force

For every number in the array:

- Keep checking if the next consecutive number exists.
- Count the sequence length.
- Store the maximum length.

### Time Complexity

```
O(nВІ)
```

Reason:

For every element, searching for the next element in the array takes O(n).

---

# Better Approach

Sort the array.

Traverse once and count consecutive numbers.

Skip duplicates.

### Time Complexity

```
O(n log n)
```

### Space Complexity

```
O(1)
```

(or O(log n) depending on the sorting algorithm)

---

# Key Observation

We are repeatedly checking whether numbers exist.

Searching inside an array is slow.

A **HashSet** gives

```
O(1)
```

average lookup.

Another important observation:

Do **not** start counting from every number.

Only start from the **beginning** of a sequence.

Example

```
1 2 3 4
```

Start from

```
1
```

Do **not** start from

```
2
3
4
```

because they belong to an already counted sequence.

---

# Optimal Approach

## Step 1

Insert every number into a HashSet.

```
numSet = set(nums)
```

Now every lookup becomes

```
O(1)
```

---

## Step 2

Traverse every unique number.

```
for num in numSet
```

---

## Step 3

Check whether this number is the start of a sequence.

A number is the start only if

```
num - 1
```

does **not** exist.

```
if num - 1 not in numSet
```

Example

```
100
```

Previous number

```
99
```

doesn't exist.

Start.

Example

```
3
```

Previous number

```
2
```

exists.

Do not start.

---

## Step 4

Initialize

```
current = num
count = 1
```

Now expand the sequence.

While

```
current + 1
```

exists,

move forward.

```
current += 1
count += 1
```

---

## Step 5

After finishing one sequence,

update

```
longest = max(longest, count)
```

Continue until every sequence has been processed.

Return

```
longest
```

# Why Check `num - 1`?

Suppose

```
1 2 3 4
```

Without this condition,

you would count

```
1в†’2в†’3в†’4

2в†’3в†’4

3в†’4

4
```

The same sequence is processed multiple times.

Checking

```
num - 1 not in numSet
```

ensures every sequence is counted exactly once.

---

# Common Mistakes

### 1. Wrong Start Condition

Incorrect

```
if num not in numSet
```

Correct

```
if num - 1 not in numSet
```

---

### 2. Initializing Count Outside

Incorrect

```
count = 0
```

outside the loop.

Correct

Initialize

```
count = 1
```

only when a new sequence starts.

---

### 3. Wrong Indentation

The following must be inside the `if` block:

- while loop
- updating longest

Otherwise `current` may contain stale values.

---

### 4. Iterating Over the Original Array

Prefer

```
for num in numSet
```

instead of

```
for num in nums
```

This automatically removes duplicates.

---

### 5. Forgetting to Return

Always

```
return longest
```

# Mental Model

Imagine every number is a domino.

```
1 2 3 4

100

200
```

Only the **leftmost domino** starts the chain.

How do you know it's the leftmost?

Its previous domino is missing.

```
num - 1 not in numSet
```

Start there.

Walk forward until the chain ends.

Count its length.

Repeat for every independent chain.

````python
class Solution(object):
    def longestConsecutive(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        num_set=set(nums)
        longest=0
        count=0
        for x in num_set:
            if x-1 not in num_set:
                current=x
                count=1
                while current+1 in num_set:
                    current+=1
                    count+=1
            longest=max(longest,count)
        return longest
````
