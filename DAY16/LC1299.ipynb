# LeetCode 1299 - Replace Elements with Greatest Element on Right Side

## Problem

Given an integer array `arr`, replace every element with the **greatest element among the elements to its right**.

The last element has no elements on its right, so replace it with **-1**.

---

## Pattern

**Reverse Traversal + Running Maximum**

---

## Intuition

Each element needs to know the **largest value on its right**.

A brute-force approach would search the entire right side for every element, resulting in repeated work.

Instead, traverse the array **from right to left** while maintaining the maximum value seen so far.

---

## Brute Force Approach

For every element:

- Traverse all elements to its right.
- Find the maximum.
- Replace the current element.

### Time Complexity

```text
O(n²)
```

---

## Optimal Idea

Maintain a variable called **`max_right`**.

Initially:

```text
max_right = -1
```

because the last element always becomes `-1`.

Now traverse the array from **right to left**.

For every index:

1. Save the current value.
2. Replace the current element with `max_right`.
3. Update `max_right` using the saved value.

---

## Why Save the Current Value?

Suppose the current element is:

```text
6
```

If you replace it immediately, the original `6` is lost.

But you still need it to update the running maximum.

Therefore, always save the original value before overwriting it.

## Time Complexity

### O(n)

Reason:

- The array is traversed exactly once.
- Each operation inside the loop takes constant time.

---

## Space Complexity

### O(1)

Reason:

Only a single variable (`max_right`) and one temporary variable are used.

---

## Common Mistakes

### Mistake 1

Using nested loops to search the right side for every element.

Result:

```text
O(n²)
```

---

### Mistake 2

Traversing from left to right.

Problem:

The maximum value on the right has not been seen yet.

---

### Mistake 3

Comparing adjacent elements.

This problem is **not** about comparing neighbors.

It is about maintaining the **maximum value seen while moving from right to left**.

---

### Mistake 4

Overwriting the current element before saving it.

Once overwritten, the original value is lost and cannot be used to update the running maximum.

---

## Interview Takeaways

- Whenever every element needs information from the **right side**, consider traversing **from right to left**.
- Maintain a running value instead of repeatedly searching.
- Always ask yourself:
  > **Can I carry information from the previous iteration instead of recomputing it?**

---

## Pattern Recognition

| Keyword | Pattern |
|----------|---------|
| Greatest element to the right | Reverse Traversal |
| Running maximum | Running Maximum |
| Right-side information | Right-to-Left Traversal |
| Avoid repeated searching | Running Maximum |

````python
class Solution(object):
    def replaceElements(self, arr):
        """
        :type arr: List[int]
        :rtype: List[int]
        """
        n=len(arr)
        max_right=-1

        for i in range(n-1,-1,-1):
        
            current = arr[i]
            arr[i] = max_right
            max_right = max(max_right, current)
           
        return arr


````
