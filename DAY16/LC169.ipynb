# LeetCode 169 - Majority Element

## Problem
Given an array `nums` of size `n`, return the majority element.

The majority element is the element that appears **more than ⌊n / 2⌋ times**.

---

## Approach
Use a **HashMap (Dictionary)** to count the frequency of each element.

- Traverse the array.
- If an element is not present in the dictionary, initialize its count as `1`.
- Otherwise, increment its count.
- As soon as the count of an element becomes greater than `n // 2`, return it immediately.

This avoids traversing the dictionary again.

---

## Pattern
- HashMap
- Frequency Counting

---

## Time Complexity
**O(n)**

- We traverse the array only once.

---

## Space Complexity
**O(n)**

- In the worst case, the dictionary stores all unique elements.

---

## What I Learned
- How to use a dictionary for frequency counting.
- Returning early can avoid unnecessary work.
- Majority Element can be solved efficiently using a HashMap.

---

## Better Solution
- **Boyer-Moore Voting Algorithm**
- Time Complexity: **O(n)**
- Space Complexity: **O(1)**

This is the optimal solution and is commonly expected in interviews.

```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n=len(nums)
        result=dict()
        for i in range(n):
            if nums[i] not in result:
                result[nums[i]]=1
            else:
                result[nums[i]]+=1
            if result[nums[i]]>n//2:
                return nums[i]
```

```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        can=None
        count=0
        for i in nums:
            if count==0:
                can=i
            if i==can:
                count+=1
            else:
                count-=1
        return can
```
