# LeetCode 605 - Can Place Flowers

## Problem

You are given a flowerbed represented as an array.

- `1` → Flower already planted.
- `0` → Empty plot.

You cannot plant flowers in **adjacent** plots.

Given an integer `n`, determine whether it is possible to plant `n` new flowers without violating the rule.

---

## Pattern

**Greedy**

---

## Intuition

At every position, ask one question:

> **Can I safely plant a flower here?**

A flower can be planted only if:

- The current plot is empty.
- The left plot is empty (or doesn't exist).
- The right plot is empty (or doesn't exist).

If all three conditions are satisfied, plant the flower immediately.

This is a **Greedy** decision because planting at the earliest valid position never decreases the maximum number of flowers that can be planted later.

---

## Algorithm

1. Initialize `count = 0`.
2. Traverse the flowerbed from left to right.
3. For every index:
   - Check whether the left side is empty.
   - Check whether the right side is empty.
   - If both sides and the current position are empty:
     - Plant a flower.
     - Increment `count`.
4. Return `count >= n`.

---

## Why Do We Use `left_empty` and `right_empty`?

### Left Side

```python
left_empty = (i == 0) or (flowerbed[i-1] == 0)
```

Meaning:

- If we are at the first position, there is no left neighbor.
- Therefore, the left side is automatically considered empty.
- Otherwise, check whether the left neighbor is `0`.

---

### Right Side

```python
right_empty = (i == k-1) or (flowerbed[i+1] == 0)
```

Meaning:

- If we are at the last position, there is no right neighbor.
- Therefore, the right side is automatically considered empty.
- Otherwise, check whether the right neighbor is `0`.

## Common Mistakes

### Mistake 1

Only checking the middle elements.

```python
for i in range(1, k-1):
```

Problem:

The first and last positions are ignored.

---

### Mistake 2

Counting possible positions without actually planting.

Wrong:

```python
count += 1
```

Correct:

```python
flowerbed[i] = 1
count += 1
```

The flowerbed must be updated because future decisions depend on previous plantings.

---

### Mistake 3

Accessing

```python
flowerbed[i-1]
```

or

```python
flowerbed[i+1]
```

without checking boundaries.

Always treat missing neighbors as empty.

---

### Mistake 4

Forgetting edge cases.

Examples:

```text
[0]

[1]

[0,0,1]

[1,0,0]
```

## Interview Takeaways

- Greedy problems ask:
  > "If I can make a safe choice now, should I take it?"

- Here, the answer is **Yes**.

- Planting as early as possible never reduces the total number of flowers that can be planted.

- Always handle boundary conditions before accessing neighboring indices.

- Learned my first **Greedy** pattern and how local optimal decisions can lead to the correct overall solution.

````python

class Solution(object):
    def canPlaceFlowers(self, flowerbed, n):
        count = 0
        k = len(flowerbed)

        for i in range(k):

            left_empty = (i == 0) or (flowerbed[i - 1] == 0)
            right_empty = (i == k - 1) or (flowerbed[i + 1] == 0)

            if left_empty and flowerbed[i] == 0 and right_empty:
                flowerbed[i] = 1
                count += 1

        return count >= n

````
