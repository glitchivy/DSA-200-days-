# LeetCode 209 - Minimum Size Subarray Sum

## Problem

Given an array of positive integers `nums` and an integer `target`, return the **minimum length** of a contiguous subarray whose sum is **greater than or equal to** `target`.

If no such subarray exists, return `0`.

---

# Brute Force

Generate every possible subarray.

For each subarray:

- Calculate its sum.
- If the sum ≥ target, update the minimum length.

### Time Complexity

```
O(n²)
```

or

```
O(n³)
```

if the sum is recalculated every time.

---

# Key Observation

Since all numbers are **positive**, two important properties hold:

### If the current sum is too small

```
sum < target
```

The only way to increase it is to **expand** the window.

Move:

```
right++
```

---

### If the current sum is large enough

```
sum >= target
```

The window is valid.

Now try to make it **smaller**.

Move:

```
left++
```

---

This gives the Sliding Window pattern.

---

# Sliding Window

Maintain two pointers.

```
left
right
```

Current Window

```
[left ........ right]
```

Maintain:

- Current Window Sum
- Minimum Window Length

---

# Algorithm

1. Initialize

- Left Pointer
- Running Sum
- Minimum Length

2. Traverse using the right pointer.

3. Add the current element to the running sum.

4. While

```
Current Sum >= Target
```

- Update the minimum length.
- Remove the leftmost element.
- Move the left pointer.

5. Continue until the right pointer reaches the end.

6. If no valid window was found, return `0`.

---

# Time Complexity

```
O(n)
```

Reason:

- Left pointer moves at most `n` times.
- Right pointer moves at most `n` times.

Each element enters and leaves the window only once.

---

# Space Complexity

```
O(1)
```

---

# Sliding Window Pattern

```
Window too small

↓

Expand

↓

Window becomes valid

↓

Update Answer

↓

Shrink

↓

Still valid?

Yes → Shrink Again

No → Expand Again
```

---

# Pattern Recognition

Ask yourself:

### Is the window too small?

```
Current Sum < Target
```

Expand.

---

### Is the window valid?

```
Current Sum >= Target
```

Shrink.

---

# Mistakes I Made

## 1. Used `if` instead of `while`

Wrong idea:

```
If the window becomes valid,
shrink only once.
```

Correct idea:

```
Keep shrinking while the window is still valid.
```

A valid window can often be reduced multiple times.

---

## 2. Resetting the count

I wrote:

```
count = 0
```

after finding a valid window.

Wrong because:

The window does **not** restart.

It simply becomes

```
[left+1 .... right]
```

The window still exists.

---

## 3. Maintaining a separate count variable

I manually tracked:

```
count
```

This is unnecessary.

Window length can always be calculated as:

```
right - left + 1
```

---

## 4. Wrong initialization of answer

I initialized the answer with

```
len(nums)
```

If no valid subarray exists, this would return the wrong answer.

Better approach:

Initialize with

```
Infinity
```

and return `0` if it never changes.

# Interview Takeaway

Whenever you see:

- Contiguous Subarray
- Minimum/Maximum Length
- Sum
- Positive Numbers

Think:

> Can I maintain a window instead of recomputing every subarray?

# Sliding Window Checklist

- Two pointers (`left`, `right`)
- Running sum
- Expand with the right pointer
- Shrink with the left pointer
- Update the answer while shrinking
- Continue until the right pointer reaches the end

````python
class Solution(object):
    def minSubArrayLen(self, target, nums):
        """
        :type target: int
        :type nums: List[int]
        :rtype: int
        """
        left = 0
        curr_sum = 0
        min_len = float('inf')
        for i in range(len(nums)):
            curr_sum += nums[i]
            while curr_sum >= target:
                min_len = min(min_len, i - left + 1)
                curr_sum -= nums[left]
                left+=1
        if min_len == float('inf'):
            return 0
        return min_len



````
