# LeetCode 31 - Next Permutation

## Problem
Given an array of integers, rearrange the numbers into the **next lexicographically greater permutation**.

If no such permutation exists, rearrange it into the **lowest possible order**.

---

# Intuition

The goal is **not** to find any larger permutation.

The goal is to find the **smallest possible permutation that is larger than the current one.**

---

## Observation 1

Example:

```
123 → 132
132 → 213
213 → 231
231 → 312
312 → 321
321 → 123
```

We always make the **smallest possible increase**.

---

## Observation 2

Traverse from the **right**.

Example:

```
1 2 7 4 3 1
```

The suffix

```
7 4 3 1
```

is in **decreasing order**.

A decreasing sequence is already the **largest possible permutation**.

Therefore, changing only this suffix cannot produce the next permutation.

---

## Pivot

Find the first index from the right where

```
nums[i] < nums[i+1]
```

Example:

```
1 2 7 4 3 1
  ↑

Pivot = 2
```

---

## Why choose the Pivot?

Everything after the pivot is already the largest arrangement.

So we must increase something before it.

The pivot is the **rightmost digit that can still be increased**.

---

## Successor

Find the **smallest element greater than the pivot**.

Because the suffix is decreasing,

**the first element greater than the pivot while scanning from the right is automatically the smallest greater element.**

Example:

```
1 2 7 4 3 1
        ↑

Successor = 3
```

---

## Swap

```
Before

1 2 7 4 3 1

↓

After Swap

1 3 7 4 2 1
```

---

## Reverse the Suffix

Current suffix:

```
7 4 2 1
```

It is still decreasing.

Reverse it:

```
1 2 4 7
```

Final answer:

```
1 3 1 2 4 7
```

---

# Edge Case

```
3 2 1
```

There is no pivot because the entire array is decreasing.

This means the current permutation is already the largest.

Reverse the whole array.

```
3 2 1

↓

1 2 3
```

---

# Algorithm

1. Traverse from right to left.
2. Find the first index where

```
nums[i] < nums[i+1]
```

3. If no pivot exists
   - Reverse the whole array.
   - Return.
4. Traverse from the right again.
5. Find the first element greater than the pivot.
6. Swap pivot and successor.
7. Reverse everything after the pivot.

---

# Time Complexity

- Finding pivot → **O(n)**
- Finding successor → **O(n)**
- Reversing suffix → **O(n)**

Overall:

```
O(n)
```

---

# Space Complexity

```
O(1)
```

---

# Common Mistakes

### 1. Wrong pivot loop

❌

```python
for i in range(n-1,-1,-1):
```

✅

```python
for i in range(n-2,-1,-1):
```

---

### 2. Returning reverse()

❌

```python
return nums.reverse()
```

✅

```python
nums.reverse()
return
```

---

### 3. Reversing only one element

❌

```python
nums[pivot+1] = nums[pivot+1][::-1]
```

✅

```python
nums[pivot+1:] = nums[pivot+1:][::-1]
```

---

### 4. Returning the array

LeetCode modifies the array **in-place**.

Do **not** return `nums`.

---

# Pattern

This problem teaches:

- Greedy
- Array Manipulation
- Lexicographical Order
- Two Pointer Reversal

````python
class Solution(object):
    def nextPermutation(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        pivot=-1
        n=len(nums)
        if n<=2:
            return nums.reverse()
        for i in range(n-2,-1,-1):
            if nums[i]<nums[i+1]:
                pivot=i
                break
        if pivot==-1:
            nums.reverse()
            return
        for i in range(n-1,pivot-1,-1):
            if nums[pivot]<nums[i]:
                nums[pivot],nums[i]=nums[i],nums[pivot]
                break
        nums[pivot+1:]=nums[pivot+1:][::-1]
        return 
````
