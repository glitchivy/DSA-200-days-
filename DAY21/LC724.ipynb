# LeetCode 724 - Find Pivot Index

## Problem

Given an integer array `nums`, return the **pivot index**.

A pivot index is an index where:

```
Sum of elements on the left == Sum of elements on the right
```

The pivot element itself is **not included** in either sum.

If no pivot exists, return `-1`.

---

# Brute Force

For every index:

1. Calculate the left sum.
2. Calculate the right sum.
3. Compare them.

```python
for i in range(n):
    left = sum(nums[:i])
    right = sum(nums[i+1:])

    if left == right:
        return i
```

### Complexity

Time:

```
O(n²)
```

because `sum()` takes `O(n)` and is called for every index.

---

# Observation

Instead of calculating the left and right sums every time, compute the total sum once.

```
Total = Left + Current + Right
```

Rearranging:

```
Right = Total - Left - Current
```

Now we only need:

- Total Sum
- Running Left Sum

The Right Sum is computed in **O(1)**.

---

# Example

```
nums = [1,7,3,6,5,6]
```

Total Sum

```
28
```

---

### i = 0

```
        i
        ↓
[1,7,3,6,5,6]
```

Left Sum

```
0
```

Right Sum

```
28 - 0 - 1 = 27
```

---

### i = 1

```
          i
          ↓
[1,7,3,6,5,6]
```

Left Sum

```
1
```

Right Sum

```
28 - 1 - 7 = 20
```

---

### i = 3

```
              i
              ↓
[1,7,3,6,5,6]
```

Left Sum

```
1 + 7 + 3 = 11
```

Right Sum

```
28 - 11 - 6 = 11
```

Since

```
11 == 11
```

Answer:

```
3
```

---

# Algorithm

1. Compute the total sum.
2. Initialize `left_sum = 0`.
3. Traverse the array.
4. Compute

```
right_sum = total_sum - left_sum - nums[i]
```

5. If

```
left_sum == right_sum
```

return `i`.

6. Otherwise update

```
left_sum += nums[i]
```

7. If no pivot exists, return `-1`.

---

# Time Complexity

```
O(n)
```

---

# Space Complexity

```
O(1)
```

# Pattern Learned

This is a **Prefix Sum** problem.

Key identity:

```
Total = Left + Current + Right
```

Therefore,

```
Right = Total - Left - Current
```

Instead of recalculating sums repeatedly, maintain a running `left_sum`.

---

# Common Mistakes

### 1. Calculating sums inside the loop

❌

```python
left = sum(nums[:i])
right = sum(nums[i+1:])
```

Time Complexity becomes:

```
O(n²)
```

---

### 2. Including the current element

Wrong:

```
Left = elements before + current
```

Correct:

```
Left = elements before only
```

and

```
Right = elements after only
```

The pivot element is **never included**.

---

### 3. Updating `left_sum` before comparison

❌

```python
left_sum += nums[i]
```

Then calculate `right_sum`.

This includes the current element in the left sum.

Correct order:

```python
right_sum = total_sum - left_sum - nums[i]

if left_sum == right_sum:
    return i

left_sum += nums[i]
```

---

# Interview Takeaway

Whenever you see:

- Left Sum
- Right Sum
- Prefix
- Running Sum

Think:

> "Can I compute the total once and derive one side using a running prefix sum?"

This is the core Prefix Sum pattern and appears frequently in array interview problems.

````python
class Solution(object):
    def pivotIndex(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n=len(nums)
        left_sum=0
        total_sum=sum(nums)
        for i in range(n):
            right_sum=total_sum-left_sum-nums[i]
            if right_sum==left_sum:
                return i
            left_sum+=nums[i]
        return -1 
````
