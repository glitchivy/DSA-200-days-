# LeetCode 977 - Squares of a Sorted Array

## Problem
Given a sorted integer array `nums`, return an array containing the squares of each number, also sorted in non-decreasing order.

---

## Intuition

- Squaring negative numbers can make them larger than positive numbers.
- The largest square will always come from either:
  - the leftmost element, or
  - the rightmost element.
- Use two pointers to compare the squares from both ends.
- Place the larger square at the end of the result array.

---

## Algorithm

1. Initialize:
   - `left = 0`
   - `right = n - 1`
   - `k = n - 1`
   - `result = [0] * n`
2. Compare:
   - `nums[left]²`
   - `nums[right]²`
3. Place the larger square at `result[k]`.
4. Move the corresponding pointer.
5. Decrement `k`.
6. Repeat until `left > right`.

---

## Dry Run

Input:
nums = [-4, -1, 0, 3, 10]

Initial:

left = 0
right = 4
k = 4

| left² | right² | Place | Result |
|-------:|--------:|------|--------|
| 16 | 100 | 100 | [0,0,0,0,100] |
| 16 | 9 | 16 | [0,0,0,16,100] |
| 1 | 9 | 9 | [0,0,9,16,100] |
| 1 | 0 | 1 | [0,1,9,16,100] |
| 0 | 0 | 0 | [0,1,9,16,100] |

Output:

[0,1,9,16,100]

---

## Why Two Pointers?

The array is already sorted.

The largest absolute value is always at one of the two ends.

Instead of sorting after squaring (`O(n log n)`), compare both ends directly and build the answer in one pass.

---

## Time Complexity

**O(n)**

Reason:
- Each element is processed exactly once.

---

## Space Complexity

**O(n)**

Reason:
- An additional result array of size `n` is used.

---

## Pattern Learned

- Two Pointers
- Compare values from both ends.
- Fill the answer array from the end.
- Useful when the input is already sorted.

---

## Edge Cases

- All positive numbers
- All negative numbers
- Contains zero
- Single element
- Duplicate values

---

## Mistakes I Made

- Initially squared every element and tried to sort it.
- Tried a single Bubble Sort pass, which doesn't fully sort the array.
- Forgot that the largest square always comes from one of the two ends.
- Realized that using two pointers gives an `O(n)` solution.

# code

```python
class Solution(object):
    def sortedSquares(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        n=len(nums)
        left=0
        right=n-1
        k=n-1
        result=n*[0]
        while left<=right:
            left_value=nums[left]*nums[left]
            right_value=nums[right]*nums[right]
            if left_value>right_value:
                result[k]=left_value
                left+=1
                
            else:
                result[k]=right_value
                right-=1
            k-=1
        return result

        
```
