# Check if a Number is Prime or Not

## Problem Statement

Given an integer `N`, check whether it is prime or not.

A prime number is a number that:
- is divisible only by `1` and itself
- has exactly `2` divisors

---

# Example

## Example 1

### Input

```python
N = 2
```

### Output

```python
True
```

### Explanation

`2` has exactly two divisors:
```python
1 and 2
```

So it is a prime number.

---

## Example 2

### Input

```python
N = 10
```

### Output

```python
False
```

### Explanation

Divisors of `10`:

```python
1, 2, 5, 10
```

Since it has more than `2` divisors, it is not prime.

---

# Brute Force Approach

## Idea

Check every number from:

```python
2 → n-1
```

If any number divides `n` completely:

```python
n % i == 0
```

then the number is NOT prime.

---

# Brute Force Code

```python
n = int(input())

val = True

if n <= 1:
    val = False

else:

    for i in range(2, n):

        if n % i == 0:
            val = False
            break

print(val)
```

---

# Time Complexity

\[
O(N)
\]

Because in worst case the loop checks almost every number from:

```python
2 → n-1
```

---

# Optimized Approach

## Observation

Divisors occur in pairs.

Example for `36`:

```python
1 × 36
2 × 18
3 × 12
4 × 9
6 × 6
```

If we find one divisor:

```python
2
```

we automatically get:

```python
18
```

So we only need to check till:

\[
\sqrt{N}
\]

---

# Why Only Till √N ?

Suppose:

\[
a \times b = N
\]

If BOTH divisors become greater than:

\[
\sqrt{N}
\]

Then:

\[
a \times b > N
\]

which is impossible.

So at least one divisor must always be less than or equal to:

\[
\sqrt{N}
\]

That is why checking till `√N` is enough.

---

# Optimized Code

```python
import math

n = int(input())

val = True

if n <= 1:
    val = False

else:

    for i in range(2, int(math.sqrt(n)) + 1):

        if n % i == 0:
            val = False
            break

print(val)
```

---

# Why Do We Use `+1` ?

Example:

```python
n = 49
```

\[
\sqrt{49} = 7
\]

Without `+1`:

```python
range(2, 7)
```

Loop stops at `6`.

So `7` is never checked.

But:

```python
49 % 7 == 0
```

So `49` would incorrectly become prime.

That is why we use:

```python
int(math.sqrt(n)) + 1
```

to include the square root itself.
