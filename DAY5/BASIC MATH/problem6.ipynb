# Print All Divisors of a Given Number

## Problem Statement

Given an integer `N`, return all divisors of `N`.

A divisor of an integer `N` is a positive integer that divides `N` without leaving a remainder.

---

# Brute Force Approach

## Idea

Check every number from `1 → N`.

If:

```python
N % i == 0
```

then `i` is a divisor.

---

## Brute Force Code

```python
n = int(input())

result = []

for i in range(1, n + 1):

    if n % i == 0:
        result.append(i)

print(result)
```

---

## Time Complexity

\[
O(N)
\]

Because the loop runs from:

```python
1 → N
```

---

# Optimized Approach

## Observation

Divisors always occur in pairs.

Example for `36`:

```python
1 × 36
2 × 18
3 × 12
4 × 9
6 × 6
```

If we find:

```python
2
```

we automatically get:

```python
36 // 2 = 18
```

---

# Why Iterate Only Till √N ?

For `36`:

\[
\sqrt{36} = 6
\]

Suppose both divisors become greater than `6`.

Then:

\[
a > 6,\quad b > 6
\]

So:

\[
a \times b > 36
\]

But divisor pair must multiply to exactly `36`.

Impossible.

So at least one divisor in every pair must be less than or equal to:

\[
\sqrt{N}
\]

That is why we only iterate till `√N`.

---

# Optimized Code

```python
import math

n = int(input())

result = []

for i in range(1, int(math.sqrt(n)) + 1):

    if n % i == 0:

        result.append(i)

        if i != n // i:
            result.append(n // i)

result.sort()

print(result)
```
