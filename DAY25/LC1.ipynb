# LeetCode 1 - Two Sum

## Problem

Given an array of integers `nums` and an integer `target`.

Return the indices of the two numbers such that:

```
nums[i] + nums[j] = target
```

You may assume there is exactly one solution.

---

# Brute Force

For every element,

check every other element.

If

```
nums[i] + nums[j] == target
```

return their indices.

---

# Time Complexity

```
O(n²)
```

because every pair is checked.

---

# Key Observation

Instead of searching the array every time,

ask:

> **Have I already seen the number I need?**

---

# Main Idea

For every number,

calculate its complement.

```
Need = Target - Current Number
```

If the complement has already appeared,

the answer is found.

Otherwise,

store the current number for future use.

---

# Why HashMap?

We want to answer:

> **Have I seen this number before?**

A HashMap gives

- Fast lookup
- Fast insertion

Both in

```
O(1)
```

average time.

---

# What Do We Store?

Store

```
Number → Index
```

Example

```
nums = [2,7,11,15]
```

HashMap after reading the first number

```
2 → 0
```

Meaning

```
Number 2 is at index 0.
```

---

# Algorithm

Traverse the array.

For every number:

### Step 1

Find the number you need.

```
Need = Target - Current Number
```

---

### Step 2

Check whether the needed number already exists in the HashMap.

If yes,

return

```
Previous Index

Current Index
```

---

### Step 3

If not,

store the current number and its index in the HashMap.

Continue.

# Why Do We Store Number → Index?

Because we search using the **number**.

Example

Current

```
7
```

Need

```
2
```

We ask

```
Have I seen 2?
```

Not

```
Have I seen index 2?
```

Therefore

```
Number → Index
```

is the correct mapping.

---

# Time Complexity

```
O(n)
```

Every element is processed exactly once.

---

# Space Complexity

```
O(n)
```

The HashMap stores previously seen numbers.

---

# Common Mistakes

### 1. Using `list.index()`

Wrong because

- It is `O(n)`.
- It always returns the first occurrence.
- It fails for duplicate values.

---

### 2. Storing

```
Number → 1
```

Wrong.

Store

```
Number → Index
```

because the problem asks for indices.

---

### 3. Returning the indices incorrectly

Return

```
[Previous Index, Current Index]
```

not

```
[Current Index, Previous Index]
```

if you are storing previous numbers first.

---

### 4. Storing before checking

Always:

```
Check first

↓

Store later
```

Otherwise,

Example

```
nums = [3,3]
target = 6
```

you may match the same element with itself.

---

# Pattern Learned

**HashMap + Complement Search**

---

# Pattern Recognition

Whenever a problem asks:

- Find two numbers with a given sum
- Check whether a complement exists
- Return indices or values quickly

Think:

> **HashMap**

Store previously seen numbers and, for each current number, search for its complement.

---

# Interview Takeaway

Instead of asking:

> **Which two numbers add up to the target?**

Ask:

> **What number do I need to complete the target, and have I already seen it?**

This small change in thinking reduces the solution from

```
O(n²)
```

to

```
O(n)
```

````python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        n=len(nums)
        val={}
        for i in range(n):
            hidden=target-nums[i]
            if hidden in val:
                return [val[hidden],i]
            val[nums[i]]=i
            
        

````
