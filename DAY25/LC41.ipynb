# LeetCode 41 - First Missing Positive

## Problem

Given an unsorted integer array `nums`, return the **smallest missing positive integer**.

Constraints:

- Time Complexity: **O(n)**
- Extra Space: **O(1)**

---

# Brute Force

Check every positive number starting from `1`.

```
1
2
3
...
n+1
```

For every number, check whether it exists in the array.

The first missing number is the answer.

### Time Complexity

```
O(n²)
```

---

# Key Observation 1

We only care about numbers in the range

```
1 ... n
```

Ignore:

- Negative numbers
- Zero
- Numbers greater than `n`

Example

```
nums = [100,200,300]
```

Answer

```
1
```

Large numbers can never be the smallest missing positive.

---

# Key Observation 2

Every valid number has a **correct home**.

```
Value → Index

1 → 0
2 → 1
3 → 2
4 → 3
...
```

Formula

```
Correct Index = Value - 1
```

Instead of sorting the array,

place every valid number into its correct home.

---

# Main Idea

Use the array itself as a hash map.

Every valid number tries to move into its own home.

Example

```
nums = [3,4,-1,1]
```

```
3 belongs at index 2

4 belongs at index 3

1 belongs at index 0
```

---

# Algorithm

## Phase 1 - Place Every Number

Traverse every index.

For the current value ask:

### Question 1

Is the value valid?

```
1 <= value <= n
```

If not,

ignore it.

---

### Question 2

Where is its home?

```
home = value - 1
```

---

### Question 3

Does the home already contain this value?

Example

```
Current = 3

Home = -1
```

Different

↓

Swap.

---

Example

```
Current = 1

Home = 1
```

Same

↓

Do not swap.

Move to the next index.

---

### Important

After swapping,

**stay on the same index.**

A new value has arrived.

Try placing that value as well.

This is why the algorithm uses

```
while
```

instead of

```
if
```

---

## Phase 2 - Find the Missing Number

After every valid number reaches its correct home,

scan the array.

Expected values:

```
Index 0 → 1

Index 1 → 2

Index 2 → 3

Index 3 → 4
```

If

```
Actual Value ≠ Expected Value
```

then

```
Answer = Expected Value
```

If every position is correct,

return

```
n + 1
```

# Why Use `while`?

One swap may bring another valid number to the same index.

Example

```
Index 1

4

↓

1

↓

-1
```

The value keeps changing.

So we keep fixing the same index until:

- The value is invalid.
- The value is already in its home.
- A duplicate already exists at its home.

Only then do we move to the next index.

---

# Why Compare the Current Value with the Home Value?

Ask one question:

> **Will this swap actually change the array?**

Example

```
Current = 3

Home = -1
```

Different

↓

Swap.

---

Example

```
Current = 1

Home = 1
```

Same

↓

Don't swap.

This also prevents infinite loops.

Example

```
[1,1]
```

Without this check,

the algorithm would keep swapping

```
1 ↔ 1

1 ↔ 1

1 ↔ 1
```

forever.

---

# Common Mistakes

### 1. Using `if` instead of `while`

Wrong.

One swap may bring another valid number to the current index.

---

### 2. Comparing the index instead of the value

Wrong idea:

```
Is the number at the correct index?
```

Correct idea:

```
Does the destination already contain the same value?
```

---

### 3. Swapping duplicates

Example

```
[1,1]
```

Always make sure the destination does **not** already contain the same value before swapping.

---

### 4. Forgetting the final scan

After rearranging,

scan from left to right.

The first mismatch gives the answer.

---

# Time Complexity

```
O(n)
```

Each valid number is moved to its correct home at most once.

---

# Space Complexity

```
O(1)
```

Everything is done in-place.

---

# Pattern Learned

**Cyclic Sort / Index Placement**

---

# Pattern Recognition

Whenever:

- Values are in the range `1...n`
- Every value has a natural correct index
- Constant extra space is required

Think:

> **Place every value into its correct home using swaps.**

Don't think of it as sorting.

Think of it as:

> **Every number is trying to reach its own home.**

````python
class Solution(object):
    def firstMissingPositive(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n=len(nums)
        for i in range(n):
            while 1<=nums[i]<=n:
                home=nums[i]-1
                if nums[home]!=nums[i]:
                    nums[i],nums[home]=nums[home],nums[i]
                else:
                    break
        for i in range(n):
            if nums[i]!=i+1:
                return i+1
        return n+1
        
````
