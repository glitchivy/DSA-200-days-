# LeetCode 75 - Sort Colors

## Problem

Given an array containing only

```
0
1
2
```

Sort the array **in-place** without using the built-in sorting function.

---

# Brute Force

Sort the array using any sorting algorithm.

Example:

- Bubble Sort
- Merge Sort
- Quick Sort

### Time Complexity

```
O(n log n)
```

or

```
O(n²)
```

depending on the sorting algorithm.

---

# Better Approach

Count the frequency of

```
0
1
2
```

Then overwrite the array.

Example

Input

```
2 0 2 1 1 0
```

Count

```
0 → 2

1 → 2

2 → 2
```

Rewrite

```
0 0 1 1 2 2
```

### Time Complexity

```
O(n)
```

### Space Complexity

```
O(1)
```

Since only three counters are used.

---

# Optimal Approach

## Dutch National Flag Algorithm

Instead of counting,

divide the array into **four regions**.

```
| 0s | 1s | Unknown | 2s |
```

Initially

```
| Unknown Unknown Unknown Unknown ... |
```

As we process the array,

the unknown region keeps shrinking.

---

# Three Pointers

## low

Represents the position where the next

```
0
```

should be placed.

---

## mid

Current element being processed.

Everything between

```
mid → high
```

is still unknown.

---

## high

Represents the position where the next

```
2
```

should be placed.

---

# Algorithm

While there are still unknown elements:

---

## Case 1

Current value is

```
0
```

Swap it with

```
low
```

Then

```
low++

mid++
```

Reason:

The element swapped from the left has already been processed.

---

## Case 2

Current value is

```
1
```

It already belongs in the middle.

Simply

```
mid++
```

---

## Case 3

Current value is

```
2
```

Swap it with

```
high
```

Then

```
high--
```

Do **NOT**

```
mid++
```

Reason:

A new unknown element has arrived from the end.

It still needs to be processed.

# Why Use `while` Instead of `for`?

A swap with

```
high
```

brings a **new unknown element** to the current position.

It must be processed before moving forward.

A `for` loop would automatically move to the next index,

possibly skipping that element.

---

# Why Doesn't `mid` Move After Seeing a `2`?

Example

```
2 1 0
```

Swap

```
2
```

with

```
0
```

Result

```
0 1 2
```

The new

```
0
```

has just arrived.

It has never been processed.

So

```
mid
```

must stay.

---

# Why Does `mid` Move After Seeing a `0`?

When swapping with

```
low
```

the incoming element has already been processed.

Therefore,

it is safe to move forward.

---

# Common Mistakes

### 1. Wrong Loop Condition

Wrong

```
low <= high
```

Correct

```
mid <= high
```

The unknown region exists between

```
mid

↓

high
```

---

### 2. Moving `mid` After Swapping With `high`

Wrong.

The new element from the end is still unknown.

Process it first.

---

### 3. Using `for` Loop

Wrong.

The current index may need multiple operations.

Use

```
while
```

instead.

---

### 4. Using `==` Instead of `=` During Swap

Remember:

```
=
```

assigns.

```
==
```

compares.

# Time Complexity

```
O(n)
```

Each iteration either

- moves `mid`, or
- moves `high`.

The unknown region continuously shrinks.

Every element is processed a constant number of times.

---

# Space Complexity

```
O(1)
```

Only three pointers are used.

No extra array is created.

---

# Pattern Learned

**Dutch National Flag Algorithm**

---

# Pattern Recognition

Whenever:

- Array contains only a few distinct values.
- Sorting must be done in-place.
- Constant extra space is required.

Think:

> **Dutch National Flag Algorithm**

---

# Mental Model

Instead of thinking about sorting,

think about maintaining four regions.

```
| 0s | 1s | Unknown | 2s |
```

Your job is simply to shrink the **Unknown** region until nothing remains.

That's why the algorithm finishes in one pass.

````python
class Solution(object):
    def sortColors(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        low=0
        high=len(nums)-1
        mid=0
        while mid<=high:
            if nums[mid]==0:
                nums[low],nums[mid]=nums[mid],nums[low]
                low+=1
                mid+=1
            elif nums[mid]==2:
                nums[high],nums[mid]=nums[mid],nums[high]
                high-=1
            else:
                mid+=1
        return nums
       
````
