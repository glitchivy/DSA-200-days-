# LeetCode 1752 - Check if Array Is Sorted and Rotated

## Problem

Given an array `nums`, determine whether it was originally sorted in **non-decreasing order** and then **rotated** some number of times (including zero).

Return:

- `True` if possible.
- `False` otherwise.

---

## Pattern

**Circular Array Traversal + Observation**

---

## Intuition

A sorted array has numbers in increasing order.

When a sorted array is rotated, the increasing order is broken **exactly once**.

That break is called a **drop**.

If the order breaks more than once, the array cannot be obtained by rotating a sorted array.

---

## Key Observation

A rotation is simply **one cut** in a sorted array.

Example:

Sorted array

```text
1 2 3 4 5
```

Rotate after `3`

```text
3 4 5 1 2
```

Notice only one place where the order decreases.

```text
5 → 1
```

Only **one break** exists.

---

## What is a Drop?

A **drop** occurs whenever

```text
Current Element > Next Element
```

Example

```text
3 4 5 1 2
```

Comparisons

```text
3 → 4 ✅
4 → 5 ✅
5 → 1 ❌  (Drop)
1 → 2 ✅
2 → 3 ✅ (Circular comparison)
```

Drops

```text
1
```

Valid.

---

## Why Circular Comparison?

The array is rotated.

A rotated array behaves like a **circle**.

Example

```text
3 4 5 1 2
↑       ↓
└───────┘
```

The last element is followed by the first element.

Therefore, we must compare

```text
Last → First
```

---

## Why Do We Use Modulo?

Suppose

```text
n = 5
```

Indices

```text
0 1 2 3 4
```

At

```text
i = 4
```

Normally,

```text
i + 1 = 5
```

which does not exist.

Using modulo

```text
(4 + 1) % 5

= 0
```

automatically wraps back to the first index.

This makes the traversal circular.

---

## Algorithm

1. Initialize `drops = 0`.
2. Traverse every element.
3. Compare the current element with the next element (circularly).
4. If the current element is greater than the next element:
   - Increment `drops`.
5. After traversal:
   - If `drops <= 1`, return `True`.
   - Otherwise, return `False`.

## Why Does One Drop Work?

Imagine cutting a sorted array once.

```text
1 2 3 | 4 5
```

Swap both parts.

```text
4 5 1 2 3
```

Only one place becomes unsorted.

```text
5 → 1
```

A single rotation can create **only one break**.

If there are two or more breaks, the array would require multiple cuts, which is impossible with one rotation.

---

## Time Complexity

### O(n)

Reason:

The array is traversed only once.

---

## Space Complexity

### O(1)

Reason:

Only one extra variable (`drops`) is used.

---

## Common Mistakes

### Mistake 1

Forgetting the circular comparison.

Wrong

```text
3 4 5 1 2
```

Only checking

```text
5 → 1
```

and forgetting

```text
2 → 3
```

Always compare the last element with the first.

---

### Mistake 2

Comparing only adjacent elements.

Remember:

The array is **circular**, not linear.

---

### Mistake 3

Thinking rotation creates multiple breaks.

One rotation means:

- One cut.
- One break.

Never more than one.

---

### Mistake 4

Using sorting.

Sorting changes the original array.

The problem only asks us to **verify** whether the property holds.

No modification is required.

````python
class Solution(object):
    def check(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        drops=0
        n=len(nums)
        
        for i in range(n):
            if nums[i]>nums[(i+1)%n]:
                drops+=1
        return drops<=1
````
