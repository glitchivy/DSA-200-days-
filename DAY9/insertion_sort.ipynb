# INSERTION SORT

insertion sort is basically taking an element and inserting it into an sorted part

code in ascending order

````python
arr=list(map(int,input().split()))
n=len(arr)
for i in range(1,n): # i am taking one cause i will use j for left side iteration 
    key=arr[i]
    j=i-1 
    while j>=0 and arr[j]>key:
        arr[j+1]=arr[j]
        j-=1
    arr[j+1]=key 
print(arr)
````

# What is Insertion Sort?

Insertion Sort works by:

```text
Taking one element at a time
and inserting it into its correct position
in the already sorted part of the array.
```

---

# Real Life Analogy

Think of arranging cards in your hand.

Suppose you already have:

```text
3 5 8
```

Now you get:

```text
4
```

You don't rearrange everything.

You simply insert:

```text
3 4 5 8
```

at the correct position.

This is exactly how Insertion Sort works.

---

# Core Idea

At every iteration:

```text
Left side is already sorted.
```

Take current element:

```python
key = arr[i]
```

and insert it into the correct position.

# Algorithm

## Step 1

Start from:

```python
i = 1
```

because first element is already sorted.

---

## Step 2

Store current element.

```python
key = arr[i]
```

---

## Step 3

Move left.

```python
j = i - 1
```

---

## Step 4

While larger elements exist:

```python
while j >= 0 and arr[j] > key
```

shift them right.

---

## Step 5

Insert key at correct position.

```python
arr[j+1] = key
```

---

# Code

```python
arr = list(map(int, input().split()))

n = len(arr)

for i in range(1, n):

    key = arr[i]

    j = i - 1

    while j >= 0 and arr[j] > key:

        arr[j + 1] = arr[j]

        j -= 1

    arr[j + 1] = key

print(arr)
```

---

# Important Variables

## key

```python
key = arr[i]
```

Stores the current element.

Without it, the value would be lost during shifting.

---

## j

```python
j = i - 1
```

Used to traverse the sorted portion.

---

# Shifting vs Swapping

Insertion Sort uses:

```python
arr[j+1] = arr[j]
```

This is:

```text
Shifting
```

NOT:

```text
Swapping
```

---

# Why Is It Called Insertion Sort?

Because every element is:

```text
Inserted into its correct position
inside the already sorted portion.
```

---

# Time Complexity

## Best Case (Already Sorted)

\[
O(N)
\]

No shifting required.

---

## Average Case

\[
O(N^2)
\]

---

## Worst Case (Reverse Sorted)

\[
O(N^2)
\]

Every element must be shifted.

---

# Space Complexity

\[
O(1)
\]

In-place sorting.

---

# Why Is Insertion Sort Faster On Nearly Sorted Arrays?

Because:

```text
Only misplaced elements are shifted.
```

Very little work is required.

Example:

```python
[1,2,3,5,4]
```

Only:

```text
5
```

needs to move.
