# MERGE SORT

it works on the principle of divide and conqeur method

````python
def merge(left, right):

    result = []

    i = 0
    j = 0

    while i < len(left) and j < len(right):

        if left[i] <= right[j]:

            result.append(left[i])
            i += 1

        else:

            result.append(right[j])
            j += 1

    while i < len(left):

        result.append(left[i])
        i += 1

    while j < len(right):

        result.append(right[j])
        j += 1

    return result


def merge_sort(arr):

    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2

    left = merge_sort(arr[:mid])

    right = merge_sort(arr[mid:])

    return merge(left, right)


arr = list(map(int, input().split()))

print(merge_sort(arr))
````

# What is Merge Sort?

Merge Sort is a:

```text
Divide and Conquer Algorithm
```

It works by:

1. Splitting the array into smaller halves.
2. Recursively sorting both halves.
3. Merging the sorted halves back together.

---

# Core Idea

```text
Big Problem
    ↓
Smaller Problems
    ↓
Even Smaller Problems
    ↓
Single Elements
    ↓
Merge Back in Sorted Order
```

---

# Example

Input:

```python
[5,3,8,1]
```

---

## Divide Phase

```text
[5,3,8,1]

      ↓

[5,3]    [8,1]

  ↓         ↓

[5] [3]  [8] [1]
```

---

## Merge Phase

```text
[5] + [3]
   ↓
[3,5]

[8] + [1]
   ↓
[1,8]

[3,5] + [1,8]
      ↓

[1,3,5,8]
```

---

# Important Observation

Merge Sort:

```text
DOES NOT sort while splitting.
```

It only:

```text
Splits while going down.
Sorts while coming back up.
```

---

## 2. merge()

# Why Better Than Bubble/Selection/Insertion?

| Algorithm | Time Complexity |
|------------|----------------|
| Bubble Sort | O(N²) |
| Selection Sort | O(N²) |
| Insertion Sort | O(N²) |
| Merge Sort | O(N log N) |
